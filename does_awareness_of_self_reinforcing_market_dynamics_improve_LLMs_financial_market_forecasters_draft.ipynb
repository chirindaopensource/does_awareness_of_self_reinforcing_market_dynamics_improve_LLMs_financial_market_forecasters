## Introduction

This project is an independent, professional-grade implementation of the ideas, methodologies, and experimental protocols from the paper titled **"Reflexivity as Prompt: Does Awareness of Self-Reinforcing Market Dynamics Improve LLMs as Financial Market Forecasters?"** by the author:
*   **Eugene Park**

This repository provides a complete, end-to-end computational framework for replicating the paper's findings. It delivers a highly optimized, crash-recoverable pipeline that evaluates how frontier Large Language Models (GPT-5, Claude Sonnet 4.6, Gemini 3 Pro) behave as financial forecasters during historical boom-bust cycles when injected with George Soros's theory of reflexivity. By treating the market not as an exogenous statistical process, but as a complex adaptive system governed by endogenous feedback between participant beliefs and fundamental realities, this codebase establishes a rigorous mechanism-aware forecasting architecture.

## Theoretical Background

The implemented methods bridge qualitative financial economics with quantitative prompt engineering and portfolio theory.

**1. The Theory of Reflexivity:**
The pipeline operationalizes Soros's framework, which posits two simultaneous, bidirectional functions. The *cognitive function* maps reality to participant beliefs, which are inherently biased. The *participative function* maps those biased beliefs back to reality, as capital allocation directly alters corporate fundamentals (e.g., cost of capital, balance sheet strength).

**2. The Prevailing Bias and P/E Dynamics:**
This two-way causality generates a *prevailing bias*, mathematically operationalized in this study as the systematic deviation of the actual Price-to-Earnings (P/E) ratio from its long-run historical mean. The pipeline feeds normalized trailing twelve-month EPS and adjusted closing prices to the LLMs to allow them to detect this bias.

**3. The Seven-Stage Boom-Bust Cycle:**
The theoretical scaffold forces the LLM to classify the market into one of seven stages: Unrecognized, InitialPhase, Testing, Acceleration, Twilight, Crash, and Recovery. A critical diagnostic distinction enforced by the prompt is the transition from Twilight (prices falling, but EPS intact) to Crash (prices and EPS falling reflexively).

**4. Structural vs. Narrative Cycles:**
The pipeline evaluates models across two historically distinct episodes: the dot-com bubble (1996–2001), representing an equity P/E-driven narrative cycle, and the Global Financial Crisis (2004–2009), representing a credit-channel structural cycle.

## Features

-   **True Idempotency and Crash-Recoverability:** Utilizes durable JSONL logging with `os.fsync` to reconstruct the exact mathematical state of the 48-cell experimental grid. The 3,456-call pipeline can be interrupted and resumed infinitely without duplicating API calls or corrupting the accumulators.
-   **Strict Data Version Calibration:** Implements an upfront buy-and-hold Sharpe ratio gate (0.65, 0.02, 0.35) that mathematically guarantees the Yahoo Finance data has not been retroactively revised, ensuring perfect reproducibility before expending API budget.
-   **Ratio-Invariant Normalization:** Executes a dynamic rolling-window normalization protocol ($P'_k = 100 \cdot P_{t_0+k-1}/P_0$) that preserves the economically essential P/E signal while stripping absolute levels and calendar dates to prevent LLM memorization.
-   **Jittered Exponential Backoff:** Provides robust handling of API rate limits across three distinct provider SDKs (OpenAI, Anthropic, Google) using a $2^{a-1} \cdot (1 + u)$ backoff schedule.
-   **Diagnostic Taxonomy:** Automatically classifies deviations from the reported results into Data Version Discrepancy, Model Version Drift, or Expected Stochasticity (specifically isolating GPT-5's temperature=1.0 behavior).

## Methodology Implemented

1.  **Data Ingestion & Alignment:** Reconciles Yahoo Finance's last-trading-day convention with Shiller's first-of-month convention using pandas `Period` indices, creating a canonical 288-month dataset.
2.  **Rolling Normalization:** Applies the base-100 normalization equations at each evaluation step $\tau$, advancing the base $P_0$ by one month iteratively.
3.  **Factorial Prompt Assembly:** Dynamically concatenates theoretical Blocks 1-5 based on a progressive-disclosure ablation design (Conditions A through D).
4.  **LLM Inference & Parsing:** Executes zero-shot calls and robustly extracts the ternary directional forecast (`up`, `neutral`, `down`) using a three-tier JSON parsing strategy with a conservative `"neutral"` fallback mechanism.
5.  **Metric Computation:** Maps forecasts to long/cash positions ($s_t \in \{1.0, 0.5, 0.0\}$), computes portfolio returns, and derives the annualized Sharpe ratio with Bessel correction ($n-1$). It strictly prohibits the mathematical error of averaging per-episode Sharpe ratios for the combined metric.

### **Step 1: Synthetic Data Generation (`df_yahoo_raw` and `df_shiller_raw`)**

The first requirement is a high-fidelity synthetic representation of the two empirical inputs. These datasets must adhere strictly to the schemas defined in the study.

**Methodology:**
1.  **Date Generation:** We generate a sequence of 288 month-end business days spanning the study period (January 1986 to December 2009) for the Yahoo dataset, and a corresponding sequence of decimal-year floats for the Shiller dataset.
2.  **Time Series Simulation:** We simulate a price path ($P_t$) using a geometric random walk with drift, and an earnings path ($\text{EPS}_t$) that roughly tracks the price to maintain a plausible P/E ratio. *Crucial Note:* While this synthetic data perfectly matches the required schema and types, it will mathematically fail the strict buy-and-hold Sharpe ratio calibration gate (0.65, 0.02, 0.35) enforced by the pipeline. In a production environment, the exact historical data must be used. For this example, the pipeline will correctly catch this divergence and halt safely, demonstrating the fail-fast architecture.
3.  **Schema Enforcement:** We explicitly cast columns to `datetime64[ns]`, `float64`, and `int64` to match the strict validation requirements.

```python
# Import the pandas library for data manipulation and DataFrame construction.
import pandas as pd
# Import the numpy library for numerical operations and array generation.
import numpy as np
# Import the os module for filesystem operations and path validation.
import os
# Import the yaml module for parsing the configuration file.
import yaml
# Import the Faker library to generate synthetic volume data.
from faker import Faker
# Import typing modules for strict type hinting.
from typing import Dict, Any, Tuple

# Initialize the Faker instance to generate realistic large integers.
fake = Faker()
# Set the Faker seed to ensure reproducibility of the synthetic volume data.
Faker.seed(42)
# Set the NumPy random seed to ensure reproducibility of the simulated price paths.
np.random.seed(42)

# Define the function to generate and save the synthetic datasets.
def generate_and_save_synthetic_datasets(
    # The start date of the study period.
    start_date: str = "1986-01-01",
    # The end date of the study period.
    end_date: str = "2009-12-31",
    # The output filename for the Yahoo CSV.
    yahoo_filename: str = "synthetic_yahoo.csv",
    # The output filename for the Shiller Excel file.
    shiller_filename: str = "synthetic_shiller.xlsx"
) -> Tuple[str, str]:
    """
    Generates synthetic Yahoo Finance and Shiller datasets and saves them to disk.

    Purpose:
        To create high-fidelity mock datasets that mimic the exact schema, data types,
        and temporal alignment of the empirical data required by the study. This allows
        the pipeline's ingestion and validation logic to be tested.

    Inputs:
        start_date (str): The start date of the simulation in 'YYYY-MM-DD' format.
        end_date (str): The end date of the simulation in 'YYYY-MM-DD' format.
        yahoo_filename (str): The target filename for the Yahoo CSV.
        shiller_filename (str): The target filename for the Shiller Excel workbook.

    Processes:
        1. Date Generation: Creates 288 month-end business days and decimal years.
        2. Simulation: Generates a random walk for prices and a correlated EPS path.
        3. Schema Enforcement: Constructs DataFrames with exact column names and types.
        4. File I/O: Saves the DataFrames to CSV and Excel formats respectively.

    Outputs:
        Tuple[str, str]: The absolute file paths to the generated Yahoo and Shiller files.

    Raises:
        ValueError: If the generated date range does not produce exactly 288 months.
        IOError: If the files cannot be written to disk.
    """
    # 1. Generate Trading Days (Month-End Business Days)
    # Use 'BM' frequency to get the last business day of each month, mimicking Yahoo.
    dates: pd.DatetimeIndex = pd.date_range(start=start_date, end=end_date, freq='BM')
    
    # Calculate the total number of months generated.
    n_months: int = len(dates)
    # Assert that exactly 288 months were generated, as required by the study.
    if n_months != 288:
        # Raise a ValueError if the temporal span is incorrect.
        raise ValueError(f"Expected exactly 288 months, generated {n_months}.")

    # 2. Simulate Price and EPS Paths
    # Generate random normal returns with a slight positive drift.
    returns: np.ndarray = np.random.normal(loc=0.005, scale=0.04, size=n_months)
    # Calculate the cumulative sum of returns to create a price path.
    cumulative_returns: np.ndarray = np.cumsum(returns)
    # Exponentiate and scale to create a realistic starting price level (~200).
    simulated_prices: np.ndarray = 200.0 * np.exp(cumulative_returns)
    
    # Simulate EPS to roughly track prices, maintaining a P/E ratio around 15.
    # Add some random noise to the EPS path.
    simulated_eps: np.ndarray = (simulated_prices / 15.0) + np.random.normal(0, 1, size=n_months)
    # Ensure EPS remains strictly positive.
    simulated_eps = np.abs(simulated_eps) + 1.0

    # 3. Construct Yahoo Finance DataFrame
    # Initialize the dictionary with the exact required column names.
    yahoo_data: Dict[str, Any] = {
        # The Date column, formatted as strings to mimic the raw CSV.
        "Date": dates.strftime('%Y-%m-%d'),
        # Mock Open prices.
        "Open": simulated_prices * 0.99,
        # Mock High prices.
        "High": simulated_prices * 1.02,
        # Mock Low prices.
        "Low": simulated_prices * 0.98,
        # Mock Close prices.
        "Close": simulated_prices,
        # The critical Adj Close column (P_t).
        "Adj Close": simulated_prices,
        # Generate random large integers for Volume using Faker.
        "Volume": [fake.random_int(min=10000000, max=900000000) for _ in range(n_months)]
    }
    # Create the pandas DataFrame.
    df_yahoo_raw: pd.DataFrame = pd.DataFrame(yahoo_data)

    # 4. Construct Shiller DataFrame
    # Calculate decimal years: Year + (Month - 1) / 12.
    decimal_years: np.ndarray = dates.year + (dates.month - 1) / 12.0
    # Initialize the dictionary with the exact required column names.
    shiller_data: Dict[str, Any] = {
        # The decimal-year Date column.
        "Date": decimal_years,
        # Mock Price column.
        "Price": simulated_prices,
        # Mock Dividend column.
        "Dividend": simulated_prices * 0.02,
        # The critical Earnings column (EPS_t).
        "Earnings": simulated_eps,
        # Mock CPI column.
        "CPI": np.linspace(100, 215, n_months),
        # Mock Long_Rate column.
        "Long_Rate": np.linspace(8.0, 3.0, n_months),
        # Mock Real_Price column.
        "Real_Price": simulated_prices * 1.5,
        # Mock Real_Dividend column.
        "Real_Dividend": simulated_prices * 0.03,
        # Mock Real_Earnings column.
        "Real_Earnings": simulated_eps * 1.5,
        # Mock CAPE column.
        "CAPE": np.random.uniform(10, 30, size=n_months)
    }
    # Create the pandas DataFrame.
    df_shiller_raw: pd.DataFrame = pd.DataFrame(shiller_data)

    # 5. Save to Disk
    # Get the absolute path for the current working directory.
    cwd: str = os.getcwd()
    # Construct the full path for the Yahoo CSV.
    yahoo_path: str = os.path.join(cwd, yahoo_filename)
    # Construct the full path for the Shiller Excel file.
    shiller_path: str = os.path.join(cwd, shiller_filename)

    # Write the Yahoo DataFrame to CSV without an index column.
    try:
        df_yahoo_raw.to_csv(yahoo_path, index=False)
    # Catch any I/O errors during the write process.
    except IOError as e:
        # Raise a descriptive IOError.
        raise IOError(f"Failed to write Yahoo data to {yahoo_path}: {e}")

    # Write the Shiller DataFrame to Excel without an index column.
    # Note: Requires 'openpyxl' engine.
    try:
        df_shiller_raw.to_excel(shiller_path, index=False, sheet_name="Data", engine='openpyxl')
    # Catch any I/O errors during the write process.
    except IOError as e:
        # Raise a descriptive IOError.
        raise IOError(f"Failed to write Shiller data to {shiller_path}: {e}")

    # Print a success message to the console.
    print(f"Synthetic datasets generated successfully.")
    # Return the absolute file paths.
    return yahoo_path, shiller_path

# Execute the generation function and store the file paths.
yahoo_file_path, shiller_file_path = generate_and_save_synthetic_datasets()
```

### **Step 2: Loading the Configuration (`config.yaml`)**

The study relies on a deterministic configuration dictionary (`STUDY_CONFIG`) that defines all hyperparameters, prompt blocks, and evaluation metrics. We assume this file exists in the working directory as `config.yaml`.

**Methodology:**
1.  **File I/O:** Open `config.yaml` in read mode with explicit UTF-8 encoding.
2.  **Parsing:** Use `yaml.safe_load` to convert the YAML structure into a nested Python dictionary.
3.  **Validation:** Catch file existence errors and parsing errors, raising explicit exceptions to halt execution if the configuration is unavailable.

```python
# Define the function to load the YAML configuration file.
def load_study_configuration(
    # The filesystem path to the YAML configuration file.
    filepath: str = "config.yaml"
) -> Dict[str, Any]:
    """
    Loads the study configuration parameters from a YAML file into a Python dictionary.

    Purpose:
        To ingest the deterministic hyperparameters, prompt blocks, and evaluation
        metrics defined in the external configuration file, ensuring reproducibility.

    Inputs:
        filepath (str): The relative or absolute path to the YAML configuration file.

    Processes:
        1. File Access: Attempts to open the specified file in read mode.
        2. Parsing: Uses PyYAML's safe_load to parse the YAML structure securely.
        3. Validation: Catches and handles FileNotFoundError and YAMLError.

    Outputs:
        Dict[str, Any]: A nested dictionary containing the complete study configuration.

    Raises:
        TypeError: If filepath is not a string.
        FileNotFoundError: If the config.yaml file does not exist.
        yaml.YAMLError: If the file contains invalid YAML syntax.
    """
    # Validate that the filepath argument is a string.
    if not isinstance(filepath, str):
        # Raise a TypeError if the input is invalid.
        raise TypeError(f"Expected str for filepath, got {type(filepath).__name__}.")

    # Attempt to open and parse the file within a try-except block.
    try:
        # Open the file stream with explicit UTF-8 encoding.
        with open(filepath, "r", encoding="utf-8") as file:
            # Parse the YAML content safely into a Python dictionary.
            config: Dict[str, Any] = yaml.safe_load(file)
        
        # Print a success message to the console.
        print(f"Successfully loaded configuration from {filepath}")
        # Return the parsed configuration dictionary.
        return config

    # Catch the specific error if the file is missing.
    except FileNotFoundError:
        # Raise a descriptive FileNotFoundError.
        raise FileNotFoundError(f"Fatal Error: Configuration file not found at '{filepath}'.")
    # Catch any YAML parsing errors.
    except yaml.YAMLError as e:
        # Raise a descriptive YAMLError.
        raise yaml.YAMLError(f"Fatal Error: Failed to parse YAML file {filepath}: {e}")

# Load the configuration into memory.
# Note: This assumes 'config.yaml' is present in the working directory.
study_config: Dict[str, Any] = load_study_configuration("config.yaml")
```

## Research Question

The paper poses a single focused empirical question: does making a frontier large language model progressively aware of Soros's theory of reflexivity improve its directional forecasting accuracy during boom-bust market cycles?

The motivating premise is that standard AI-assisted forecasting treats the market as exogenous — a system the model observes from outside. Reflexivity theory rejects this. Every forecaster who forms beliefs about a market and acts on them is a participant in the very feedback loop that generates price dynamics. The paper asks whether giving the model this self-awareness of market structure improves its predictive performance.

## Theoretical Grounding: Soros's Reflexivity Framework

The paper operationalizes reflexivity through two simultaneous causal functions.

The cognitive function runs from reality to beliefs: participants observe prices and fundamentals and form views, but those views are always incomplete and systematically biased. The participative function runs from beliefs back to reality: participants act on their biased views, and those actions — buying, selling, capital allocation — alter the very fundamentals the views were supposed to reflect. A rising price improves a firm's access to capital, which genuinely strengthens its balance sheet, which seems to validate the higher price. The self-reinforcing loop is real, not illusory.

The aggregate gap between the actual price-to-earnings ratio and its long-run historical mean is defined as the prevailing bias — the measurable signature of this feedback in the data. Booms occur when the bias and the underlying trend reinforce each other; busts occur when the accumulated deviation becomes unsustainable and the feedback reverses. Busts are characteristically faster and steeper than the preceding booms because the bias must fully unwind.

The paper operationalizes Soros's boom-bust sequence as seven named stages, each defined by the observable dynamics of price and EPS:

**Boom arc:** Unrecognized → InitialPhase → Testing → Acceleration
**Bust arc:** Twilight → Crash → Recovery

Each stage has specific signal-based criteria. In Testing, for instance, a price correction of at least 5% occurs while EPS continues growing — a price-only correction that, if followed by a recovery to new highs, strengthens the self-reinforcing loop. In Twilight, price has reversed from its peak but EPS has not yet declined — the critical diagnostic that separates Twilight from Crash. In Crash, falling prices feed back into fundamentals and EPS begins to deteriorate, producing the characteristic rapid collapse.

## Data and Evaluation Episodes

The data consists of two publicly available time series:

Monthly adjusted closing prices for the S&P 500, sourced from Yahoo Finance. Monthly trailing-twelve-month earnings-per-share (EPS), sourced from Robert Shiller's publicly available dataset. Both series begin in January 1986 to provide a ten-year pre-episode history for computing the rolling P/E z-score that underlies the stage definitions.

Two evaluation episodes are selected, each spanning 72 months:

The dot-com bubble (January 1996 through December 2001) represents a speculative cycle driven primarily by narrative excess and extreme P/E expansion in technology. The prevailing bias in equity valuations is the central signal, making this the textbook reflexivity test case.

The global financial crisis (January 2004 through December 2009) represents a structural credit cycle in which reflexive feedback operated through the financial system's fundamentals rather than equity P/E directly. This episode tests whether reflexivity-grounded reasoning generalizes beyond the canonical speculative bubble to a structurally different cycle type.

## Normalization and Anti-Memorization Protocol

Because both evaluation episodes predate all three models' training cutoffs, the paper implements a normalization procedure designed to prevent retrieval of memorized outcomes.

Let $P_0$ denote the price at the first month of the context window. The normalized price series is $P'_t = 100 \cdot P_t / P_0$ and the normalized EPS series is $\text{EPS}'_t = 100 \cdot \text{EPS}_t / P_0$ — both divided by the same base price. This construction preserves the true P/E ratio exactly at every point, since $P'_t / \text{EPS}'_t = P_t / \text{EPS}_t$, while removing all absolute level information. Calendar dates are replaced by $t = 1, \ldots, T$ and the index is relabeled "equity market index," stripping all identifying calendar and name information.

## Experimental Design: Four Prompting Conditions

The core intervention is a four-condition zero-shot ablation in which theoretical content accumulates across conditions. Each condition builds on the previous one:

**Condition A (Naive):** The model receives the normalized price series and the seven-stage boom-bust framework as definitional context, but no earnings data and no theoretical explanation of why the stages arise.

**Condition B (+ EPS):** Adds the normalized EPS series, enabling the model to compute the P/E ratio and observe the divergence between price and fundamentals.

**Condition C (+ Reflexivity):** Adds a formal description of the reflexivity mechanism — the cognitive function, the participative function, and the concept of the prevailing bias as the P/E gap from its long-run mean.

**Condition D (Full Theory):** Adds a stage-level mapping that explains precisely how the cognitive and participative functions operate at each of the seven stages — for example, that in Stage 4 (Acceleration) the participative function is at maximum power as higher prices sustain EPS growth and justify ever-higher P/E ratios, and that in Stage 6 (Crash) falling prices destroy credit availability and directly reduce EPS through reflexive fundamental damage.

The approximate prompt sizes are 1.5 kB, 1.6 kB, 4.3 kB, and 10.0 kB for Conditions A through D respectively, reflecting the substantial increase in theoretical content at Conditions C and D.

## Models, Temperature, and Context Windows

Three frontier models are evaluated under zero-shot prompting: GPT-5 at temperature 1.0 (the API default, which the authors note cannot be adjusted), Claude Sonnet 4.6 at temperature 0, and Gemini 3 Pro at temperature 0.

Two context window lengths are evaluated: $W = 36$ months and $W = 60$ months. This yields a total experimental design of 3 models × 4 conditions × 2 context windows × 2 episodes, with $n = 72$ monthly forecast observations per episode cell and $n = 144$ combined.

## Tasks and Metrics

At each evaluation step, the model receives the most recent $W$ months of normalized price and EPS history and is asked to forecast three quantities for the next monthly close: the boom-bust cycle stage, the direction of the price return (up, neutral, or down), and the return magnitude as a signed percentage.

The primary evaluation metric is directional accuracy: the fraction of months for which the predicted direction matches the realized direction. The secondary metric is the annualized Sharpe ratio of an implied long/cash strategy, defined as:

$$\text{SR} = \sqrt{12} \cdot \frac{\mathbb{E}[r^{\text{port}}]}{\text{std}[r^{\text{port}}]}$$

where position $s_t \in \{1.0, 0.5, 0.0\}$ maps up/neutral/down to portfolio return $r^{\text{port}}_t = s_t \cdot r_{t+1}$.

The authors note an important distinction between these two metrics: directional accuracy treats all months equally, whereas the Sharpe ratio is return-weighted and therefore reflects whether correct calls are concentrated in the economically most significant months — a distinction that matters greatly in markets characterized by rare but extreme moves.

The buy-and-hold benchmark Sharpe ratios are 0.65 for the dot-com episode, 0.02 for the GFC, and 0.35 combined.

## Results by Model

Below is the diagram of the experimental architecture, followed by the result-by-result breakdown.**Gemini 3 Pro** is the most consistent beneficiary of theoretical scaffolding across both context windows and both episodes. At $W = 36$, Condition B alone lifts GFC accuracy to 62.5% from a naive baseline of 56.9%, and Conditions C and D maintain gains in both episodes. The combined Sharpe rises monotonically from −0.024 under Condition A to +0.277 under Condition D. At $W = 60$, the GFC accuracy reaches 59.7% under both C and D, and the combined Sharpe under Condition C is +0.357. Gemini is the only model for which reflexivity awareness translates into consistent improvement in both directional accuracy and risk-adjusted returns across the full experimental grid.

**GPT-5** presents the starkest context-window dependence in the dataset. At $W = 36$, directional accuracy declines monotonically from Condition A (54.9% combined) to Condition D (50.7%), the only result pattern in the dataset that sits consistently at or below the 50% random baseline. The Sharpe trajectory at $W = 36$ is also negative or near-zero across most conditions. At $W = 60$, the pattern reverses completely: combined accuracy rises from 52.8% under Condition A to 59.0% under Condition D, and the Sharpe ratio reaches +0.440 under Condition D, with an exceptional +0.674 in the GFC under Condition C — the highest single-cell Sharpe in the entire dataset. The authors' interpretation is that the reflexivity framework requires sufficient P/E accumulation history to be actionable: with 36 months of context, GPT-5 cannot observe the full build-up of the prevailing bias, causing the theory to induce premature bearish positions; 60 months provides enough signal for the framework to serve as productive scaffolding.

**Claude Sonnet 4.6** does not benefit consistently from reflexivity awareness. At $W = 36$, Condition B raises dot-com accuracy to 58.3% — a dataset high for that episode — but the Sharpe trajectory is non-monotone: +0.454 at B, declining to +0.128 at C, and only partially recovering to +0.256 at D. At $W = 60$, adding EPS or theory reduces both accuracy and Sharpe relative to the naive baseline. The authors identify the core difficulty as a timing problem: the reflexivity mechanism (Condition C) correctly identifies overvaluation but provides no information about when the reversal will occur, inducing over-bearish positions during extended Mania phases. The stage-level mapping (Condition D) partially mitigates this by conveying that the Acceleration stage can persist with sustained upward momentum, but the correction is incomplete.

## Two Cross-Cutting Findings

**Episode-type dependence.** Across all three models, reflexivity conditions produce more consistent gains in the GFC than in the dot-com bubble. In the GFC, the reflexive feedback between falling asset prices and deteriorating bank fundamentals directly matches the mechanism described in the theory, giving the model concrete traction. In the dot-com, the Mania phase was exceptionally prolonged and narrative-driven, making overvaluation signals systematically misleading for timing. The authors draw an explicit parallel with human traders: understanding that a market is reflexively overvalued is not sufficient to time the reversal. LLMs equipped with economic theory exhibit analogous context-specific strengths and weaknesses.

**The accuracy–Sharpe divergence.** The paper identifies a key methodological asymmetry between its two metrics. Higher directional accuracy does not guarantee better risk-adjusted returns, because accuracy treats all months equally while economic performance is return-weighted. A model that correctly calls direction in low-magnitude months while missing large Mania-phase rallies can achieve high directional accuracy but negative Sharpe. This divergence is most pronounced for Claude at $W = 36$, where Condition A achieves the highest dot-com Sharpe (+0.559) precisely by remaining fully long during the bull market — a strategy that is correct in economic terms even if not always directionally accurate month-by-month.

## Conclusions and Implications

The paper's central conclusion is that reflexivity-aware prompting is not a uniformly beneficial intervention. Its value depends on the interaction between the theoretical scaffold and each model's internalized prior about financial markets. The same theoretical content that substantially improves GPT-5 at $W = 60$ produces no consistent benefit for Claude and causes mild harm for GPT-5 at $W = 36$.

The paper makes three methodological contributions. First, it establishes context window length as a substantive design variable rather than a robustness check — the reflexivity framework describes a multi-year process of bias accumulation, and shorter windows may provide insufficient history for the framework to be actionable. Second, it demonstrates the accuracy–Sharpe divergence, arguing that economic evaluation metrics must accompany directional accuracy in any financially motivated LLM forecasting study. Third, it introduces the normalization protocol as an operational approach to guarding against memorization when evaluating on historically identified episodes.

The principal limitation acknowledged by the authors is statistical power: two 72-month episodes provide limited basis for cross-episode generalization, and several reported differences between conditions are small enough to warrant caution. Future work is identified as requiring additional boom-bust episodes, a broader sweep of context window lengths, and investigation of whether the model-level heterogeneity tracks training data composition differences.

````python
# Task 1 — Validate the STUDY_CONFIG Dictionary Structure and Internal Consistency

# ==============================================================================
# Task 1: Validate the STUDY_CONFIG Dictionary Structure and Internal Consistency
# ==============================================================================

# Define authoritative module-level constants for episode start indices.
# These constants represent the 1-based t-index for the start of each episode.
TAU_START_DOTCOM: int = 121
TAU_START_GFC: int = 217

# -------------------------------------------------------------------------------------------------------------------------------
# Helper Function: Safely retrieve nested dictionary values using a dotted path
# -------------------------------------------------------------------------------------------------------------------------------
def get_by_path(config: Dict[str, Any], path: str) -> Any:
    """
    Retrieves a value from a nested dictionary using a dot-separated string path.

    Parameters
    ----------
    config : Dict[str, Any]
        The configuration dictionary to traverse.
    path : str
        The dot-separated path to the desired key (e.g., "sharpe_params.annualization_factor").

    Returns
    -------
    Any
        The value located at the specified path.

    Raises
    ------
    KeyError
        If any key in the path does not exist or if an intermediate value is not a dictionary.
    """
    # Split the dot-separated path into individual keys
    keys: List[str] = path.split(".")
    # Initialize the current level to the root configuration dictionary
    current: Any = config

    # Iterate through each key in the path
    for key in keys:
        # Check if the current level is a dictionary
        if not isinstance(current, dict):
            # Raise an error if we cannot traverse further
            raise KeyError(f"Path traversal failed at '{key}' in path '{path}': Not a dictionary.")
        # Check if the key exists in the current dictionary
        if key not in current:
            # Raise a precise KeyError naming the missing key and the full path
            raise KeyError(f"Missing required key: '{path}' (failed at '{key}')")
        # Advance to the next level in the nested dictionary
        current = current[key]

    # Return the final resolved value
    return current

# -------------------------------------------------------------------------------------------------------------------------------
# Task 1, Step 1: Validate the presence, structural completeness, and type-correctness of keys
# -------------------------------------------------------------------------------------------------------------------------------
def validate_config_structure(config: Dict[str, Any]) -> None:
    """
    Validates the presence and type-correctness of all 13 required top-level keys
    and specific depth-2 critical sub-keys in the configuration dictionary.

    Parameters
    ----------
    config : Dict[str, Any]
        The master study configuration dictionary.

    Raises
    ------
    KeyError
        If any required key or sub-key is missing.
    TypeError
        If any key contains a value of an incorrect type.
    ValueError
        If a numeric value is not finite or a string is empty when it shouldn't be.
    """
    # Define the validation manifest: a list of tuples containing:
    # (dotted_path, expected_type, is_none_permitted)
    manifest: List[Tuple[str, type, bool]] = [
        # Top-level keys (existence and type check)
        ("paper_metadata", dict, False),
        ("raw_data_schemas", dict, False),
        ("data_provenance", dict, False),
        ("normalization_params", dict, False),
        ("prompt_blocks", dict, False),
        ("experiment_config", dict, False),
        ("llm_api_config", dict, False),
        ("user_message_params", dict, False),
        ("rolling_eval_params", dict, False),
        ("sharpe_params", dict, False),
        ("accuracy_params", dict, False),
        ("results_table_params", dict, False),
        ("reported_results", dict, False),

        # Depth-2 critical sub-keys governing equations and logic
        ("sharpe_params.annualization_factor", int, False),
        ("sharpe_params.std_denominator_per_episode", int, False),
        ("sharpe_params.std_denominator_combined", int, False),
        ("sharpe_params.sqrt_annualization_factor", float, False),
        ("normalization_params.scale_factor", int, False),
        ("accuracy_params.random_baseline", float, False),
        ("experiment_config.n_models", int, False),
        ("experiment_config.n_conditions", int, False),
        ("experiment_config.n_context_windows", int, False),
        ("experiment_config.n_episodes", int, False),
        ("experiment_config.n_cells", int, False),
        ("experiment_config.n_eval_months_per_cell", int, False),
        ("experiment_config.total_api_calls", int, False),

        # Episode sub-dictionaries and their critical bounds
        ("experiment_config.episodes.dotcom", dict, False),
        ("experiment_config.episodes.dotcom.t_start", int, False),
        ("experiment_config.episodes.dotcom.t_end", int, False),
        ("experiment_config.episodes.dotcom.n_months", int, False),
        ("experiment_config.episodes.gfc", dict, False),
        ("experiment_config.episodes.gfc.t_start", int, False),
        ("experiment_config.episodes.gfc.t_end", int, False),
        ("experiment_config.episodes.gfc.n_months", int, False)
    ]

    # Iterate through the predefined manifest to validate each path
    for path, expected_type, none_permitted in manifest:
        # Retrieve the value using the helper function; raises KeyError if missing
        value: Any = get_by_path(config, path)

        # Check if the value is None and if None is permitted for this path
        if value is None:
            if none_permitted:
                # If None is permitted, skip further type checking for this key
                continue
            else:
                # If None is not permitted, raise a TypeError
                raise TypeError(f"Value at '{path}' cannot be None.")

        # Perform strict type checking based on the expected type
        if expected_type is bool:
            # Prevent the boolean-as-integer trap by using strict identity checking
            if type(value) is not bool:
                raise TypeError(f"Value at '{path}' must be strictly bool, got {type(value).__name__}.")
        elif expected_type is int:
            # Ensure the value is an integer and strictly NOT a boolean
            if not isinstance(value, int) or type(value) is bool:
                raise TypeError(f"Value at '{path}' must be strictly int, got {type(value).__name__}.")
        elif expected_type is float:
            # Allow integers to be evaluated as floats, but strictly exclude booleans
            if not isinstance(value, (int, float)) or type(value) is bool:
                raise TypeError(f"Value at '{path}' must be float or int, got {type(value).__name__}.")
            # Verify that the numeric value is finite (rejects NaN and infinity)
            if not math.isfinite(value):
                raise ValueError(f"Numeric value at '{path}' must be finite, got {value}.")
        elif expected_type is str:
            # Ensure the value is a string
            if not isinstance(value, str):
                raise TypeError(f"Value at '{path}' must be str, got {type(value).__name__}.")
            # Ensure the string is not empty
            if len(value.strip()) == 0:
                raise ValueError(f"String value at '{path}' cannot be empty.")
        elif expected_type is list:
            # Ensure the value is a list
            if not isinstance(value, list):
                raise TypeError(f"Value at '{path}' must be list, got {type(value).__name__}.")
            # Ensure the list is not empty
            if len(value) == 0:
                raise ValueError(f"List at '{path}' cannot be empty.")
        else:
            # For other types (like dict), perform a standard isinstance check
            if not isinstance(value, expected_type):
                raise TypeError(f"Value at '{path}' must be {expected_type.__name__}, got {type(value).__name__}.")

# -------------------------------------------------------------------------------------------------------------------------------
# Task 1, Step 2: Validate the seven numerically critical scalar parameters
# -------------------------------------------------------------------------------------------------------------------------------
def validate_critical_scalars(config: Dict[str, Any]) -> None:
    """
    Validates that the numerically critical scalar parameters carry the exact
    VERBATIM values stated in the paper.

    Parameters
    ----------
    config : Dict[str, Any]
        The master study configuration dictionary.

    Raises
    ------
    ValueError
        If any critical scalar parameter deviates from its expected exact value.
    """
    # Initialize a list to accumulate validation error messages
    errors: List[str] = []

    # Define validation rules for integer parameters requiring exact equality
    # Format: (dotted_path, expected_value)
    int_rules: List[Tuple[str, int]] = [
        ("sharpe_params.annualization_factor", 12),
        ("normalization_params.scale_factor", 100),
        ("sharpe_params.std_denominator_per_episode", 71),
        ("sharpe_params.std_denominator_combined", 143)
    ]

    # Iterate through the integer rules
    for path, expected in int_rules:
        # Retrieve the actual value from the configuration
        actual: int = get_by_path(config, path)
        # Enforce exact mathematical equality for integers
        if actual != expected:
            # Append a detailed error message if the check fails
            errors.append(f"{path} = {actual}, expected exactly {expected}")

    # Define validation rules for float parameters requiring tolerance-based equality
    # Format: (dotted_path, expected_value)
    float_rules: List[Tuple[str, float]] = [
        ("sharpe_params.bah_sharpe_dotcom", 0.65),
        ("sharpe_params.bah_sharpe_gfc", 0.02),
        ("sharpe_params.bah_sharpe_combined", 0.35),
        ("accuracy_params.random_baseline", 0.5)
    ]

    # Iterate through the float rules
    for path, expected in float_rules:
        # Retrieve the actual value from the configuration
        actual: float = get_by_path(config, path)
        # Enforce equality within floating-point representation limits
        # rel_tol=1e-9 and abs_tol=1e-12 accommodate binary representation artifacts
        if not math.isclose(actual, expected, rel_tol=1e-9, abs_tol=1e-12):
            # Append a detailed error message if the check fails
            errors.append(f"{path} = {actual}, expected approximately {expected}")

    # If any errors were accumulated, raise a single comprehensive ValueError
    if errors:
        # Join all error messages with newlines for readability
        error_msg: str = "Critical scalar validation failed:\n" + "\n".join(errors)
        raise ValueError(error_msg)

# -------------------------------------------------------------------------------------------------------------------------------
# Task 1, Step 3: Verify internal arithmetic self-consistency
# -------------------------------------------------------------------------------------------------------------------------------
def verify_arithmetic_consistency(config: Dict[str, Any]) -> None:
    """
    Verifies internal arithmetic self-consistency among the nested configuration values.

    Parameters
    ----------
    config : Dict[str, Any]
        The master study configuration dictionary.

    Raises
    ------
    ValueError
        If any derived quantity contradicts its primitive operands or boundary conditions.
    """
    # Initialize a list to accumulate arithmetic inconsistency error messages
    errors: List[str] = []

    # Extract primitive operands for grid dimensions
    n_models: int = get_by_path(config, "experiment_config.n_models")
    n_conditions: int = get_by_path(config, "experiment_config.n_conditions")
    n_windows: int = get_by_path(config, "experiment_config.n_context_windows")
    n_episodes: int = get_by_path(config, "experiment_config.n_episodes")
    n_months_per_cell: int = get_by_path(config, "experiment_config.n_eval_months_per_cell")

    # 1. Verify Cell Count Identity
    # Expected: n_models * n_conditions * n_context_windows * n_episodes
    expected_cells: int = n_models * n_conditions * n_windows * n_episodes
    actual_cells: int = get_by_path(config, "experiment_config.n_cells")
    if actual_cells != expected_cells:
        errors.append(f"Arithmetic inconsistency in n_cells: computed {expected_cells}, stored {actual_cells}")

    # 2. Verify Total API Calls Identity
    # Expected: n_cells * n_eval_months_per_cell
    expected_api_calls: int = actual_cells * n_months_per_cell
    actual_api_calls: int = get_by_path(config, "experiment_config.total_api_calls")
    if actual_api_calls != expected_api_calls:
        errors.append(f"Arithmetic inconsistency in total_api_calls: computed {expected_api_calls}, stored {actual_api_calls}")

    # 3. Verify Combined Observation Count Identity
    # Expected: n_episodes * n_eval_months_per_cell
    expected_combined_months: int = n_episodes * n_months_per_cell
    actual_combined_months: int = get_by_path(config, "experiment_config.n_eval_months_combined")
    if actual_combined_months != expected_combined_months:
        errors.append(f"Arithmetic inconsistency in n_eval_months_combined: computed {expected_combined_months}, stored {actual_combined_months}")

    # Extract episode boundary parameters
    dotcom_start: int = get_by_path(config, "experiment_config.episodes.dotcom.t_start")
    dotcom_end: int = get_by_path(config, "experiment_config.episodes.dotcom.t_end")
    gfc_start: int = get_by_path(config, "experiment_config.episodes.gfc.t_start")
    gfc_end: int = get_by_path(config, "experiment_config.episodes.gfc.t_end")

    # 4. Verify Dot-com Episode Span
    # Expected: t_end - t_start + 1 == 72
    expected_dotcom_span: int = dotcom_end - dotcom_start + 1
    if expected_dotcom_span != 72:
        errors.append(f"Arithmetic inconsistency in dotcom span: computed {expected_dotcom_span}, expected 72")

    # 5. Verify GFC Episode Span
    # Expected: t_end - t_start + 1 == 72
    expected_gfc_span: int = gfc_end - gfc_start + 1
    if expected_gfc_span != 72:
        errors.append(f"Arithmetic inconsistency in gfc span: computed {expected_gfc_span}, expected 72")

    # Define the window lengths to check
    windows: List[int] = [36, 60]
    # Define the maximum allowed dataset index
    max_dataset_index: int = 288

    # 6. Verify Window Start Indices and Upper Boundaries for Dot-com
    for w in windows:
        # Formula: t_0(1) = tau_start - W + 1 - 1 = tau_start - W
        computed_t0: int = dotcom_start - w
        # Retrieve the stored t0 value for tau=1
        stored_t0: int = get_by_path(config, f"normalization_params.window_boundaries.dotcom_W{w}.t0_tau1")
        if computed_t0 != stored_t0:
            errors.append(f"Arithmetic inconsistency in dotcom W={w} t0_tau1: computed {computed_t0}, stored {stored_t0}")

        # Formula for upper boundary safety: t_0(72) + W <= 288
        # t_0(72) = tau_start - W + 72 - 1 = tau_start - W + 71
        # t_0(72) + W = tau_start + 71
        upper_bound: int = dotcom_start + 71
        if upper_bound > max_dataset_index:
            errors.append(f"Upper boundary violation for dotcom W={w}: forecast target {upper_bound} exceeds {max_dataset_index}")

    # 7. Verify Window Start Indices and Upper Boundaries for GFC
    for w in windows:
        # Formula: t_0(1) = tau_start - W + 1 - 1 = tau_start - W
        computed_t0: int = gfc_start - w
        # Retrieve the stored t0 value for tau=1
        stored_t0: int = get_by_path(config, f"normalization_params.window_boundaries.gfc_W{w}.t0_tau1")
        if computed_t0 != stored_t0:
            errors.append(f"Arithmetic inconsistency in gfc W={w} t0_tau1: computed {computed_t0}, stored {stored_t0}")

        # Formula for upper boundary safety: t_0(72) + W <= 288
        # t_0(72) = tau_start - W + 72 - 1 = tau_start - W + 71
        # t_0(72) + W = tau_start + 71
        upper_bound: int = gfc_start + 71
        if upper_bound > max_dataset_index:
            errors.append(f"Upper boundary violation for gfc W={w}: forecast target {upper_bound} exceeds {max_dataset_index}")

    # If any arithmetic inconsistencies were found, raise a comprehensive ValueError
    if errors:
        # Join all error messages with newlines for readability
        error_msg: str = "Arithmetic consistency checks failed:\n" + "\n".join(errors)
        raise ValueError(error_msg)

# -------------------------------------------------------------------------------------------------------------------------------
# Task 1, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def validate_study_config(config: Dict[str, Any]) -> bool:
    """
    Orchestrates the complete validation of the STUDY_CONFIG dictionary.

    This function sequentially executes structural validation, critical scalar
    validation, and arithmetic consistency verification to ensure the configuration
    is perfectly sound before any downstream processing occurs.

    Parameters
    ----------
    config : Dict[str, Any]
        The master study configuration dictionary to validate.

    Returns
    -------
    bool
        True if all validation checks pass successfully.

    Raises
    ------
    KeyError
        If required keys are missing.
    TypeError
        If values have incorrect types.
    ValueError
        If numerical values are incorrect or arithmetic inconsistencies exist.
    """
    # Step 1: Validate the presence, structural completeness, and type-correctness
    validate_config_structure(config)

    # Step 2: Validate the exact values of numerically critical scalar parameters
    validate_critical_scalars(config)

    # Step 3: Verify internal arithmetic self-consistency among nested values
    verify_arithmetic_consistency(config)

    # If no exceptions were raised during the steps, the configuration is valid
    return True
````

````python
# Task 2 — Ingest and Validate the Yahoo Finance Raw Price Series

# ==============================================================================
# Task 2: Ingest and Validate the Yahoo Finance Raw Price Series
# ==============================================================================

@dataclass
class YahooDataContainer:
    """
    A structured container for the Yahoo Finance dataset and its provenance metadata.

    Attributes
    ----------
    df : pd.DataFrame
        The pandas DataFrame containing the filtered and validated time series.
    price_array : np.ndarray
        The 0-based float64 NumPy array of the Adj Close prices (P_t).
    metadata : Dict[str, Any]
        Provenance metadata including download date, source, and divergence flags.
    """
    df: pd.DataFrame = field(default_factory=pd.DataFrame)
    price_array: np.ndarray = field(default_factory=lambda: np.array([]))
    metadata: Dict[str, Any] = field(default_factory=dict)

# -------------------------------------------------------------------------------------------------------------------------------
# Task 2, Step 1: Load df_yahoo_raw, confirm schema, parse dates, and initialize metadata
# -------------------------------------------------------------------------------------------------------------------------------
def load_and_parse_yahoo_data(file_path: str, download_date: str) -> YahooDataContainer:
    """
    Loads the raw Yahoo Finance CSV, verifies the exact column schema, parses the Date
    column to datetime64[ns], and initializes the reproducibility metadata.

    Parameters
    ----------
    file_path : str
        The filesystem path to the raw Yahoo Finance CSV file.
    download_date : str
        The ISO 8601 formatted string representing the date the data was downloaded.

    Returns
    -------
    YahooDataContainer
        A container holding the loaded DataFrame and initialized metadata.

    Raises
    ------
    KeyError
        If any of the exactly required 7 columns are missing from the CSV.
    """
    # Define the exact expected column headers, noting the space in "Adj Close"
    expected_columns: List[str] = ["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]

    # Load the CSV without parsing dates automatically to maintain explicit control
    df_raw: pd.DataFrame = pd.read_csv(file_path, parse_dates=False, index_col=False)

    # Verify that all expected columns are present in the loaded DataFrame
    missing_cols = [col for col in expected_columns if col not in df_raw.columns]
    if missing_cols:
        # Fail loudly if the schema deviates from the strict expectation
        raise KeyError(f"Raw Yahoo Finance data is missing required columns: {missing_cols}")

    # Parse the Date column explicitly using the YYYY-MM-DD format
    # This prevents locale-dependent parsing errors and ensures datetime64[ns] dtype
    df_raw["Date"] = pd.to_datetime(df_raw["Date"], format="%Y-%m-%d")

    # Ensure the Adj Close column is cast to float64
    df_raw["Adj Close"] = df_raw["Adj Close"].astype("float64")

    # Initialize the provenance metadata dictionary
    metadata: Dict[str, Any] = {
        "source": "Yahoo Finance",
        "ticker": "^GSPC",
        "column_used": "Adj Close",
        "download_date": download_date,
        "version_divergence_flag": False
    }

    # Return the structured container with the parsed DataFrame and metadata
    return YahooDataContainer(df=df_raw, metadata=metadata)

# -------------------------------------------------------------------------------------------------------------------------------
# Task 2, Step 2: Filter to study window, assert completeness, and validate chronology
# -------------------------------------------------------------------------------------------------------------------------------
def filter_and_validate_yahoo_chronology(container: YahooDataContainer) -> YahooDataContainer:
    """
    Filters the DataFrame to the 1986-2009 study window, asserts exact row counts,
    checks for missing values, and validates strict chronological monotonicity.

    Parameters
    ----------
    container : YahooDataContainer
        The container holding the parsed DataFrame.

    Returns
    -------
    YahooDataContainer
        The container with the filtered and chronologically validated DataFrame.

    Raises
    ------
    ValueError
        If row counts, NaN checks, or chronological assertions fail.
    """
    df: pd.DataFrame = container.df

    # Filter to the inclusive calendar interval [1986-01-01, 2009-12-31]
    # This safely captures end-of-month trading days regardless of weekends/holidays
    start_date = pd.Timestamp("1986-01-01")
    end_date = pd.Timestamp("2009-12-31")
    mask = (df["Date"] >= start_date) & (df["Date"] <= end_date)
    df_filtered: pd.DataFrame = df.loc[mask].copy()

    # Sort explicitly by Date to guarantee chronological order for downstream array extraction
    df_filtered.sort_values("Date", inplace=True, ignore_index=True)

    # Assert exactly 288 rows remain (12 months * 24 years)
    if len(df_filtered) != 288:
        raise ValueError(f"Expected exactly 288 rows after filtering, got {len(df_filtered)}.")

    # Assert zero NaN values in the critical Adj Close column
    if df_filtered["Adj Close"].isna().sum() > 0:
        raise ValueError("The 'Adj Close' column contains NaN values within the study window.")

    # Handle the Volume column tolerantly: if not all NaN, ensure non-negative
    if not df_filtered["Volume"].isna().all():
        # Check that all non-NaN volume values are >= 0
        if not (df_filtered["Volume"].dropna() >= 0).all():
            raise ValueError("The 'Volume' column contains negative values.")

    # Validate strict monotonic increase of the Date column
    # diff() > 0 ensures no dates go backwards and no exact duplicate days exist
    if not (df_filtered["Date"].diff().dropna() > pd.Timedelta(0)).all():
        raise ValueError("The 'Date' sequence is not strictly monotonically increasing.")

    # Validate absence of duplicate calendar months
    # Convert to monthly periods and check uniqueness
    monthly_periods = df_filtered["Date"].dt.to_period("M")
    if not monthly_periods.is_unique:
        raise ValueError("Duplicate calendar months detected in the filtered dataset.")

    # Update the container with the filtered DataFrame
    container.df = df_filtered
    return container

# -------------------------------------------------------------------------------------------------------------------------------
# Task 2, Step 3: Extract P_t array, verify positivity, and check boundary anchors
# -------------------------------------------------------------------------------------------------------------------------------
def extract_and_verify_price_array(container: YahooDataContainer) -> YahooDataContainer:
    """
    Extracts the Adj Close series as a 0-based float64 NumPy array, verifies strict
    positivity, and checks approximate boundary values to detect version divergence.

    Parameters
    ----------
    container : YahooDataContainer
        The container holding the filtered DataFrame.

    Returns
    -------
    YahooDataContainer
        The container updated with the extracted price_array and divergence flags.

    Raises
    ------
    ValueError
        If any price in the array is not strictly positive.
    """
    df: pd.DataFrame = container.df

    # Extract P_t as a contiguous 0-based NumPy float64 array
    # The paper's 1-based index t maps to Python array index t-1
    price_array: np.ndarray = df["Adj Close"].to_numpy(dtype="float64")

    # Verify strict positivity: S&P 500 index levels must be > 0
    if not (price_array > 0).all():
        raise ValueError("Fatal data corruption: Non-positive prices detected in the array.")

    # Define the approximate boundary anchors for version divergence checking
    # Format: (1-based t_index, expected_approximate_value, description)
    anchors: List[Tuple[int, float, str]] = [
        (121, 636.0, "Jan 1996 (Dot-com start)"),
        (192, 1148.0, "Dec 2001 (Dot-com end)"),
        (217, 1131.0, "Jan 2004 (GFC start)"),
        (288, 1115.0, "Dec 2009 (GFC end)")
    ]

    divergence_detected: bool = False

    # Check each anchor against a +/- 10% tolerance band
    for t_index, expected_val, desc in anchors:
        # Apply the t-1 indexing convention to access the 0-based array
        actual_val: float = price_array[t_index - 1]

        # Calculate the 10% tolerance bounds
        lower_bound: float = expected_val * 0.90
        upper_bound: float = expected_val * 1.10

        # If the actual value falls outside the bounds, log a warning
        if not (lower_bound <= actual_val <= upper_bound):
            logger.warning(
                f"Data version divergence at {desc} (t={t_index}): "
                f"Actual P_t = {actual_val:.2f}, Expected ≈ {expected_val:.2f} (±10%)."
            )
            divergence_detected = True

    # Update the metadata flag if any divergence was detected
    if divergence_detected:
        container.metadata["version_divergence_flag"] = True
        logger.warning("Yahoo Finance data may be version-divergent from the original study.")

    # Update the container with the extracted array
    container.price_array = price_array
    return container

# -------------------------------------------------------------------------------------------------------------------------------
# Task 2, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def ingest_and_validate_yahoo_finance_series(file_path: str, download_date: str) -> YahooDataContainer:
    """
    Orchestrates the complete ingestion, filtering, and validation of the raw
    Yahoo Finance price series.

    This function sequentially loads the CSV, filters it to the 1986-2009 study
    window, validates chronological integrity, extracts the numerical price array,
    and checks for data version divergence.

    Parameters
    ----------
    file_path : str
        The filesystem path to the raw Yahoo Finance CSV file.
    download_date : str
        The ISO 8601 formatted string representing the date the data was downloaded.

    Returns
    -------
    YahooDataContainer
        A structured container holding the validated DataFrame, the 0-based
        float64 price array, and provenance metadata.
    """
    # Step 1: Load the CSV, confirm the schema, and parse dates
    container: YahooDataContainer = load_and_parse_yahoo_data(file_path, download_date)

    # Step 2: Filter to the study window and validate chronological integrity
    container = filter_and_validate_yahoo_chronology(container)

    # Step 3: Extract the numerical array, verify positivity, and check anchors
    container = extract_and_verify_price_array(container)

    # Return the fully populated and validated container
    return container
````

````python
# Task 3 — Ingest, Convert and Validate the Shiller EPS Series

# ==============================================================================
# Task 3: Ingest, Convert and Validate the Shiller EPS Series
# ==============================================================================

@dataclass
class ShillerDataContainer:
    """
    A structured container for the Shiller EPS dataset and its provenance metadata.

    Attributes
    ----------
    df : pd.DataFrame
        The pandas DataFrame containing the filtered, decoded, and validated time series.
    eps_array : np.ndarray
        The 0-based float64 NumPy array of the Earnings series (EPS_t).
    metadata : Dict[str, Any]
        Provenance metadata including access date, source, and divergence flags.
    """
    df: pd.DataFrame = field(default_factory=pd.DataFrame)
    eps_array: np.ndarray = field(default_factory=lambda: np.array([]))
    metadata: Dict[str, Any] = field(default_factory=dict)

# -------------------------------------------------------------------------------------------------------------------------------
# Task 3, Step 1: Load df_shiller_raw, confirm schema, and decode decimal-year dates
# -------------------------------------------------------------------------------------------------------------------------------
def load_and_decode_shiller_data(file_path: str) -> ShillerDataContainer:
    """
    Loads the raw Shiller Excel workbook, confirms the exact column schema, drops
    non-numeric junk rows, and decodes the decimal-year Date column into calendar dates.

    Parameters
    ----------
    file_path : str
        The filesystem path to the raw Shiller Excel workbook (.xls or .xlsx).

    Returns
    -------
    ShillerDataContainer
        A container holding the decoded DataFrame and initialized metadata.

    Raises
    ------
    KeyError
        If any of the exactly required 10 columns are missing from the worksheet.
    ValueError
        If the Excel file cannot be read or the sheet is missing.
    """
    # Attempt to load the specific "Data" sheet commonly used in Shiller's workbook
    try:
        df_raw: pd.DataFrame = pd.read_excel(file_path, sheet_name="Data")
    except (ValueError, KeyError):
        # Fallback to the first sheet (index 0) if "Data" is not found, logging a warning
        logger.warning("Sheet 'Data' not found in Shiller workbook. Falling back to sheet index 0.")
        df_raw = pd.read_excel(file_path, sheet_name=0)

    # Strip leading and trailing whitespace from all column headers to ensure exact matching
    df_raw.columns = df_raw.columns.str.strip()

    # Coerce the 'Date' column to numeric, turning text-based headers/footers into NaN
    df_raw["Date"] = pd.to_numeric(df_raw["Date"], errors="coerce")

    # Drop any rows where the 'Date' column is NaN, effectively removing junk rows
    df_raw.dropna(subset=["Date"], inplace=True)

    # Define the exact expected column headers for the Shiller dataset
    expected_columns: List[str] = [
        "Date", "Price", "Dividend", "Earnings", "CPI",
        "Long_Rate", "Real_Price", "Real_Dividend", "Real_Earnings", "CAPE"
    ]

    # Verify that all expected columns are present in the loaded DataFrame
    missing_cols = [col for col in expected_columns if col not in df_raw.columns]
    if missing_cols:
        # Fail loudly if the schema deviates from the strict expectation
        raise KeyError(f"Raw Shiller data is missing required columns: {missing_cols}")

    # Extract the float64 decimal-year Date column for mathematical decoding
    decimal_dates: pd.Series = df_raw["Date"].astype("float64")

    # Apply the inverse decimal-year formula: year = floor(Date)
    # Use numpy.floor to ensure exact mathematical fidelity
    df_raw["year"] = np.floor(decimal_dates).astype("int64")

    # Apply the inverse decimal-year formula: month = round((Date - floor(Date)) * 12) + 1
    # Use numpy.round to prevent floating-point truncation errors (e.g., 1986.0833 -> month 2)
    df_raw["month"] = np.round((decimal_dates - df_raw["year"]) * 12).astype("int64") + 1

    # Construct a proper datetime64[ns] column for downstream alignment
    # Explicitly set the day to 1 (first of the month) as per Shiller's convention
    df_raw["calendar_date"] = pd.to_datetime({
        "year": df_raw["year"],
        "month": df_raw["month"],
        "day": 1
    })

    # Initialize the provenance metadata dictionary with the verbatim access date
    metadata: Dict[str, Any] = {
        "source": "Robert Shiller",
        "dataset_title": "U.S. Stock Markets 1871-Present and CAPE Ratio",
        "column_used": "Earnings",
        "access_date": "2025-05",  # VERBATIM from the study citation
        "version_divergence_flag": False
    }

    # Return the structured container with the decoded DataFrame and metadata
    return ShillerDataContainer(df=df_raw, metadata=metadata)

# -------------------------------------------------------------------------------------------------------------------------------
# Task 3, Step 2: Filter to study window, extract EPS, and assert chronology
# -------------------------------------------------------------------------------------------------------------------------------
def filter_and_validate_shiller_chronology(container: ShillerDataContainer) -> ShillerDataContainer:
    """
    Filters the DataFrame to the 1986-2009 study window, asserts exact row counts,
    checks for missing EPS values, and validates strict chronological monotonicity.

    Parameters
    ----------
    container : ShillerDataContainer
        The container holding the decoded DataFrame.

    Returns
    -------
    ShillerDataContainer
        The container with the filtered and chronologically validated DataFrame.

    Raises
    ------
    ValueError
        If row counts, NaN checks, or chronological assertions fail.
    """
    df: pd.DataFrame = container.df

    # Filter to the inclusive window of January 1986 through December 2009
    # Filtering on the decoded integer 'year' column ensures exact calendar matching
    mask = (df["year"] >= 1986) & (df["year"] <= 2009)
    df_filtered: pd.DataFrame = df.loc[mask].copy()

    # Sort explicitly by the constructed calendar_date to guarantee chronological order
    df_filtered.sort_values("calendar_date", inplace=True, ignore_index=True)

    # Assert exactly 288 rows remain (12 months * 24 years)
    if len(df_filtered) != 288:
        raise ValueError(f"Expected exactly 288 rows after filtering Shiller data, got {len(df_filtered)}.")

    # Ensure the Earnings column is explicitly cast to float64
    df_filtered["Earnings"] = df_filtered["Earnings"].astype("float64")

    # Assert zero NaN values in the critical Earnings column
    if df_filtered["Earnings"].isna().sum() > 0:
        raise ValueError("The 'Earnings' column contains NaN values within the study window.")

    # Convert the calendar_date to monthly periods for strict chronological validation
    monthly_periods = df_filtered["calendar_date"].dt.to_period("M")

    # Validate absence of duplicate calendar months
    if not monthly_periods.is_unique:
        raise ValueError("Duplicate calendar months detected in the filtered Shiller dataset.")

    # Validate that the sequence is contiguous without any missing months
    # The difference between consecutive monthly periods must be exactly 1 month
    period_diffs = monthly_periods.diff().dropna()
    # In pandas, subtracting periods yields an integer representing the number of units (months)
    if not (period_diffs == 1).all():
        raise ValueError("The Shiller monthly sequence is not contiguous; missing months detected.")

    # Update the container with the filtered DataFrame
    container.df = df_filtered
    return container

# -------------------------------------------------------------------------------------------------------------------------------
# Task 3, Step 3: Verify EPS positivity and check boundary anchors
# -------------------------------------------------------------------------------------------------------------------------------
def verify_eps_positivity_and_anchors(container: ShillerDataContainer) -> ShillerDataContainer:
    """
    Extracts the Earnings series as a 0-based float64 NumPy array, verifies strict
    positivity, and checks approximate boundary values using (year, month) lookups.

    Parameters
    ----------
    container : ShillerDataContainer
        The container holding the filtered DataFrame.

    Returns
    -------
    ShillerDataContainer
        The container updated with the extracted eps_array and divergence flags.

    Raises
    ------
    ValueError
        If any EPS value in the array is not strictly positive.
    """
    df: pd.DataFrame = container.df

    # Extract EPS_t as a contiguous 0-based NumPy float64 array
    eps_array: np.ndarray = df["Earnings"].to_numpy(dtype="float64")

    # Verify strict positivity: Trailing twelve-month EPS must be > 0
    if not (eps_array > 0).all():
        raise ValueError("Fatal data corruption: Non-positive EPS values detected in the array.")

    # Define the approximate boundary anchors for version divergence checking
    # Format: (year, month, expected_approximate_value, description)
    # Lookups are done by (year, month) to avoid positional indexing risks prior to the Task 4 merge
    anchors: List[Tuple[int, int, float, str]] = [
        (1993, 1, 19.43, "Jan 1993 (Dot-com W=36 tau=1 start)"),
        (1996, 1, 34.11, "Jan 1996 (Dot-com start)"),
        (2001, 12, 36.84, "Dec 2001 (Dot-com end)"),
        (2004, 1, 54.22, "Jan 2004 (GFC start)"),
        (2009, 12, 56.86, "Dec 2009 (GFC end)")
    ]

    divergence_detected: bool = False

    # Check each anchor against a +/- 10% tolerance band
    for year, month, expected_val, desc in anchors:
        # Locate the exact row using boolean masking on the decoded integer columns
        mask = (df["year"] == year) & (df["month"] == month)

        # Ensure exactly one row matches the year/month combination
        if mask.sum() != 1:
            raise ValueError(f"Failed to locate unique row for anchor {desc} ({year}-{month}).")

        # Extract the actual Earnings value for the matched row
        actual_val: float = df.loc[mask, "Earnings"].values[0]

        # Calculate the 10% tolerance bounds
        lower_bound: float = expected_val * 0.90
        upper_bound: float = expected_val * 1.10

        # If the actual value falls outside the bounds, log a warning
        if not (lower_bound <= actual_val <= upper_bound):
            logger.warning(
                f"Data version divergence at {desc}: "
                f"Actual EPS_t = {actual_val:.2f}, Expected ≈ {expected_val:.2f} (±10%)."
            )
            divergence_detected = True

    # Update the metadata flag if any divergence was detected due to Shiller revisions
    if divergence_detected:
        container.metadata["version_divergence_flag"] = True
        logger.warning("Shiller EPS data may be version-divergent from the original study.")

    # Update the container with the extracted array
    container.eps_array = eps_array
    return container

# -------------------------------------------------------------------------------------------------------------------------------
# Task 3, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def ingest_and_validate_shiller_eps_series(file_path: str) -> ShillerDataContainer:
    """
    Orchestrates the complete ingestion, decoding, filtering, and validation of the
    raw Shiller EPS series.

    This function sequentially loads the Excel workbook, decodes the decimal-year
    dates, filters to the 1986-2009 study window, validates chronological integrity,
    extracts the numerical EPS array, and checks for data version divergence.

    Parameters
    ----------
    file_path : str
        The filesystem path to the raw Shiller Excel workbook (.xls or .xlsx).

    Returns
    -------
    ShillerDataContainer
        A structured container holding the validated DataFrame, the 0-based
        float64 EPS array, and provenance metadata.
    """
    # Step 1: Load the Excel file, confirm the schema, and decode decimal-year dates
    container: ShillerDataContainer = load_and_decode_shiller_data(file_path)

    # Step 2: Filter to the study window, extract EPS, and assert chronological integrity
    container = filter_and_validate_shiller_chronology(container)

    # Step 3: Extract the numerical array, verify positivity, and check boundary anchors
    container = verify_eps_positivity_and_anchors(container)

    # Return the fully populated and validated container
    return container
````

````python
# Task 4 — Align, Integrate and Cleanse the Merged Dataset

# ==============================================================================
# Task 4: Align, Integrate and Cleanse the Merged Dataset
# ==============================================================================

# Define authoritative module-level constants for episode start indices
TAU_START_DOTCOM: int = 121
TAU_START_GFC: int = 217

@dataclass
class MergedDataContainer:
    """
    A structured container for the aligned, integrated, and cleansed dataset.

    Attributes
    ----------
    df : pd.DataFrame
        The canonical DataFrame containing columns 't', 'P_t', and 'EPS_t'.
    price_array : np.ndarray
        The 0-based float64 NumPy array of the adjusted closing prices (P_t).
    eps_array : np.ndarray
        The 0-based float64 NumPy array of the earnings per share (EPS_t).
    metadata : Dict[str, Any]
        Combined provenance metadata including the SHA-256 dataset fingerprint.
    """
    df: pd.DataFrame = field(default_factory=pd.DataFrame)
    price_array: np.ndarray = field(default_factory=lambda: np.array([]))
    eps_array: np.ndarray = field(default_factory=lambda: np.array([]))
    metadata: Dict[str, Any] = field(default_factory=dict)

# -------------------------------------------------------------------------------------------------------------------------------
# Task 4, Step 1: Inner-join price and EPS series and compute SHA-256 fingerprint
# -------------------------------------------------------------------------------------------------------------------------------
def align_and_integrate_series(
    yahoo_df: pd.DataFrame,
    shiller_df: pd.DataFrame
) -> Tuple[pd.DataFrame, str]:
    """
    Performs a month-level inner join between the Yahoo and Shiller DataFrames,
    verifies exact row matching, and computes an immutable SHA-256 fingerprint.

    Parameters
    ----------
    yahoo_df : pd.DataFrame
        The filtered Yahoo Finance DataFrame containing the 'Date' and 'Adj Close' columns.
    shiller_df : pd.DataFrame
        The filtered Shiller DataFrame containing the 'calendar_date' and 'Earnings' columns.

    Returns
    -------
    Tuple[pd.DataFrame, str]
        A tuple containing the merged DataFrame and the SHA-256 hex digest string.

    Raises
    ------
    ValueError
        If the inner join results in fewer or more than exactly 288 rows, indicating unmatched months.
    """
    # Create a common month-level key for the Yahoo data (abstracts away last-trading-day)
    yahoo_keyed: pd.DataFrame = yahoo_df.copy()
    yahoo_keyed["ym_key"] = yahoo_keyed["Date"].dt.to_period("M")

    # Create a common month-level key for the Shiller data (abstracts away first-of-month)
    shiller_keyed: pd.DataFrame = shiller_df.copy()
    shiller_keyed["ym_key"] = shiller_keyed["calendar_date"].dt.to_period("M")

    # Execute the inner join on the standardized monthly period key
    merged_df: pd.DataFrame = pd.merge(yahoo_keyed, shiller_keyed, on="ym_key", how="inner")

    # Assert that the merge produces exactly 288 matched rows
    if len(merged_df) != 288:
        # If the count deviates, compute the symmetric difference to identify the exact missing months
        yahoo_keys: Set[pd.Period] = set(yahoo_keyed["ym_key"])
        shiller_keys: Set[pd.Period] = set(shiller_keyed["ym_key"])
        unmatched_keys: Set[pd.Period] = yahoo_keys ^ shiller_keys

        # Raise a highly descriptive error naming the unmatched months
        raise ValueError(
            f"Alignment failed: Expected exactly 288 matched rows, got {len(merged_df)}. "
            f"Unmatched months detected: {sorted(list(unmatched_keys))}"
        )

    # Extract temporary arrays to compute the deterministic SHA-256 fingerprint
    # Explicitly cast to float64 to ensure byte-level consistency across environments
    temp_price_array: np.ndarray = merged_df["Adj Close"].to_numpy(dtype="float64")
    temp_eps_array: np.ndarray = merged_df["Earnings"].to_numpy(dtype="float64")

    # Concatenate the byte representations in a fixed order: price first, then EPS
    concatenated_bytes: bytes = temp_price_array.tobytes() + temp_eps_array.tobytes()

    # Generate the SHA-256 hex digest
    dataset_fingerprint: str = hashlib.sha256(concatenated_bytes).hexdigest()

    logger.info(f"Dataset aligned successfully. SHA-256 Fingerprint: {dataset_fingerprint}")

    return merged_df, dataset_fingerprint

# -------------------------------------------------------------------------------------------------------------------------------
# Task 4, Step 2: Construct canonical DataFrame and enforce zero-tolerance conditions
# -------------------------------------------------------------------------------------------------------------------------------
def construct_canonical_dataset(merged_df: pd.DataFrame) -> Tuple[pd.DataFrame, np.ndarray, np.ndarray]:
    """
    Constructs the canonical 3-column DataFrame, assigns the 1-based t-index,
    enforces zero-tolerance data quality assertions, and extracts the NumPy arrays.

    Parameters
    ----------
    merged_df : pd.DataFrame
        The raw merged DataFrame resulting from the inner join.

    Returns
    -------
    Tuple[pd.DataFrame, np.ndarray, np.ndarray]
        The canonical DataFrame, the 0-based price array, and the 0-based EPS array.

    Raises
    ------
    ValueError
        If any NaNs, negative values, or duplicate t-indices are detected.
    """
    # Explicitly sort by the monthly period key in ascending order
    # This is the critical step that guarantees chronological integrity before assigning t
    merged_df.sort_values("ym_key", ascending=True, inplace=True, ignore_index=True)

    # Construct the canonical DataFrame with the required columns
    canonical_df: pd.DataFrame = pd.DataFrame()

    # Assign the authoritative 1-based t-index (1 through 288)
    # Explicitly cast to int64 to prevent data type drift
    canonical_df["t"] = pd.Series(range(1, 289), dtype="int64")

    # Map the source columns to the canonical mathematical variable names
    canonical_df["P_t"] = merged_df["Adj Close"].astype("float64")
    canonical_df["EPS_t"] = merged_df["Earnings"].astype("float64")

    # Retain the ym_key temporarily for boundary certification in Step 3
    canonical_df["ym_key"] = merged_df["ym_key"]

    # Zero-tolerance assertion 1: Zero NaNs in P_t
    if canonical_df["P_t"].isna().sum() > 0:
        raise ValueError("Zero-tolerance failure: NaN values detected in P_t.")

    # Zero-tolerance assertion 2: Zero NaNs in EPS_t
    if canonical_df["EPS_t"].isna().sum() > 0:
        raise ValueError("Zero-tolerance failure: NaN values detected in EPS_t.")

    # Zero-tolerance assertion 3: Zero negative or zero values in P_t
    if (canonical_df["P_t"] <= 0).sum() > 0:
        raise ValueError("Zero-tolerance failure: Non-positive values detected in P_t.")

    # Zero-tolerance assertion 4: Zero negative or zero values in EPS_t
    if (canonical_df["EPS_t"] <= 0).sum() > 0:
        raise ValueError("Zero-tolerance failure: Non-positive values detected in EPS_t.")

    # Zero-tolerance assertion 5: No duplicate t values
    if not canonical_df["t"].is_unique:
        raise ValueError("Zero-tolerance failure: Duplicate t-indices detected.")

    # Extract the canonical columns as 0-based NumPy arrays for O(1) positional access
    # The paper's 1-based index t maps to Python array index t-1
    price_array: np.ndarray = canonical_df["P_t"].to_numpy(dtype="float64")
    eps_array: np.ndarray = canonical_df["EPS_t"].to_numpy(dtype="float64")

    return canonical_df, price_array, eps_array

# -------------------------------------------------------------------------------------------------------------------------------
# Task 4, Step 3: Confirm calendar-to-t mapping and certify pre-episode buffers
# -------------------------------------------------------------------------------------------------------------------------------
def certify_boundaries_and_buffers(canonical_df: pd.DataFrame) -> None:
    """
    Confirms the calendar-to-t mapping at the four episode boundaries and certifies
    that sufficient pre-episode history exists to prevent out-of-bounds access.

    Parameters
    ----------
    canonical_df : pd.DataFrame
        The canonical DataFrame containing the 't' and 'ym_key' columns.

    Raises
    ------
    ValueError
        If any boundary mapping is incorrect or if pre-episode buffers are insufficient.
    """
    # Define the expected calendar mappings for the four critical episode boundaries
    # Format: (t_index, expected_period_string, description)
    expected_boundaries: List[Tuple[int, str, str]] = [
        (121, "1996-01", "Dot-com start"),
        (192, "2001-12", "Dot-com end"),
        (217, "2004-01", "GFC start"),
        (288, "2009-12", "GFC end")
    ]

    # Verify each boundary mapping declaratively
    for t_val, expected_str, desc in expected_boundaries:
        # Locate the period key corresponding to the specific t-index
        actual_period: pd.Period = canonical_df.loc[canonical_df["t"] == t_val, "ym_key"].iloc[0]
        expected_period: pd.Period = pd.Period(expected_str, freq="M")

        # Assert exact equality between the actual and expected periods
        if actual_period != expected_period:
            raise ValueError(
                f"Boundary certification failed at {desc} (t={t_val}): "
                f"Expected {expected_period}, got {actual_period}."
            )

    # Certify pre-episode buffers for the most demanding window size (W=60)
    # Formula: t_0(1) = tau_start - W + 1 - 1 = tau_start - W
    w_max: int = 60

    # Calculate the earliest required index for the Dot-com episode
    t0_dotcom_w60: int = TAU_START_DOTCOM - w_max
    if t0_dotcom_w60 < 1:
        raise ValueError(f"Insufficient pre-episode buffer for Dot-com W=60: t_0={t0_dotcom_w60} < 1.")

    # Calculate the earliest required index for the GFC episode
    t0_gfc_w60: int = TAU_START_GFC - w_max
    if t0_gfc_w60 < 1:
        raise ValueError(f"Insufficient pre-episode buffer for GFC W=60: t_0={t0_gfc_w60} < 1.")

    # Assert the total dataset length is exactly 288, ensuring the upper bound (t=288) is safe
    if len(canonical_df) != 288:
        raise ValueError(f"Dataset length certification failed: Expected 288, got {len(canonical_df)}.")

    # Drop the temporary ym_key column as it is no longer needed
    canonical_df.drop(columns=["ym_key"], inplace=True)

    logger.info("All episode boundaries and pre-episode buffers certified successfully.")

# -------------------------------------------------------------------------------------------------------------------------------
# Task 4, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def align_integrate_and_cleanse(
    yahoo_container: Any,
    shiller_container: Any
) -> MergedDataContainer:
    """
    Orchestrates the alignment, integration, and cleansing of the Yahoo and Shiller datasets.

    This function sequentially performs an inner join on a monthly key, constructs the
    canonical 3-column DataFrame, enforces zero-tolerance data quality checks, extracts
    high-performance NumPy arrays, and certifies all temporal boundaries.

    Parameters
    ----------
    yahoo_container : YahooDataContainer
        The validated container holding the Yahoo Finance price data.
    shiller_container : ShillerDataContainer
        The validated container holding the Shiller EPS data.

    Returns
    -------
    MergedDataContainer
        A structured container holding the canonical DataFrame, the extracted
        NumPy arrays, and the combined provenance metadata including the SHA-256 fingerprint.
    """
    # Step 1: Inner-join the series and compute the immutable SHA-256 fingerprint
    merged_df, fingerprint = align_and_integrate_series(
        yahoo_df=yahoo_container.df,
        shiller_df=shiller_container.df
    )

    # Step 2: Construct the canonical DataFrame, enforce quality, and extract arrays
    canonical_df, price_array, eps_array = construct_canonical_dataset(merged_df)

    # Step 3: Confirm calendar-to-t mapping and certify pre-episode buffers
    certify_boundaries_and_buffers(canonical_df)

    # Combine metadata from both sources and append the new fingerprint
    combined_metadata: Dict[str, Any] = {
        "yahoo_metadata": yahoo_container.metadata,
        "shiller_metadata": shiller_container.metadata,
        "dataset_fingerprint_sha256": fingerprint
    }

    # Return the fully populated and certified MergedDataContainer
    return MergedDataContainer(
        df=canonical_df,
        price_array=price_array,
        eps_array=eps_array,
        metadata=combined_metadata
    )
````

````python
# Task 5 — Compute and Validate Buy-and-Hold Calibration Benchmarks

# ==============================================================================
# Task 5: Compute and Validate Buy-and-Hold Calibration Benchmarks
# ==============================================================================

# Define authoritative module-level constants for episode start indices
TAU_START_DOTCOM: int = 121
TAU_START_GFC: int = 217

# Define the fixed order for episode concatenation to ensure reproducibility
EPISODE_ORDER: List[str] = ["dotcom", "gfc"]

class CalibrationError(Exception):
    """
    Custom exception raised when the computed buy-and-hold Sharpe ratios
    deviate from the study's reported targets by more than the allowed tolerance.
    This indicates a fatal data version discrepancy (e.g., Yahoo Finance retroactive revisions).
    """
    pass

@dataclass
class CalibrationReport:
    """
    A structured container for the results of the buy-and-hold calibration check.

    Attributes
    ----------
    computed_sharpes : Dict[str, float]
        The computed annualized Sharpe ratios for 'dotcom', 'gfc', and 'combined'.
    target_sharpes : Dict[str, float]
        The target Sharpe ratios defined in the study configuration.
    deviations : Dict[str, float]
        The absolute deviations between the computed and target Sharpe ratios.
    passed : bool
        True if all deviations are within the allowed tolerance (<= 0.02).
    """
    computed_sharpes: Dict[str, float] = field(default_factory=dict)
    target_sharpes: Dict[str, float] = field(default_factory=dict)
    deviations: Dict[str, float] = field(default_factory=dict)
    passed: bool = False

# -------------------------------------------------------------------------------------------------------------------------------
# Helper Function: Centralized Sharpe Ratio Computation
# -------------------------------------------------------------------------------------------------------------------------------
def compute_annualized_sharpe(
    return_series: np.ndarray,
    annualization_factor: float
) -> Tuple[float, float, float]:
    """
    Computes the arithmetic mean, Bessel-corrected sample standard deviation,
    and annualized Sharpe ratio for a given sequence of returns.

    This helper centralizes the logic to ensure ddof=1 and degeneracy guards
    are applied consistently across Tasks 5, 13, and 14.

    Parameters
    ----------
    return_series : np.ndarray
        A 1-dimensional float64 NumPy array of portfolio returns.
    annualization_factor : float
        The constant used to annualize the Sharpe ratio (e.g., sqrt(12)).

    Returns
    -------
    Tuple[float, float, float]
        A tuple containing (mean_return, sample_std_dev, annualized_sharpe).
    """
    # Compute the simple arithmetic mean of the returns
    mean_return: float = float(np.mean(return_series))

    # Compute the sample standard deviation using Bessel's correction (denominator n-1)
    # ddof=1 is mandatory; ddof=0 would systematically inflate the Sharpe ratio
    std_dev: float = float(np.std(return_series, ddof=1))

    # Degeneracy guard: If standard deviation is exactly zero (or extremely close),
    # the Sharpe ratio is undefined. Set to 0.0 to prevent division by zero.
    if math.isclose(std_dev, 0.0, abs_tol=1e-12):
        logger.warning("Degenerate standard deviation (0.0) detected. Setting Sharpe ratio to 0.0.")
        return mean_return, std_dev, 0.0

    # Compute the annualized Sharpe ratio: SR = sqrt(12) * (mean / std_dev)
    sharpe_ratio: float = annualization_factor * (mean_return / std_dev)

    return mean_return, std_dev, sharpe_ratio

# -------------------------------------------------------------------------------------------------------------------------------
# Task 5, Step 1: Compute the 72-element realized monthly return sequence for each episode
# -------------------------------------------------------------------------------------------------------------------------------
def compute_episode_raw_returns(price_array: np.ndarray) -> Dict[str, np.ndarray]:
    """
    Computes the 72-element realized monthly return sequence for the dot-com
    and GFC episodes using raw, non-normalized prices.

    Implements the corrected return-index convention derived from the forecast target:
    r_tau = (P_{tau_start + tau - 1} - P_{tau_start + tau - 2}) / P_{tau_start + tau - 2}

    Parameters
    ----------
    price_array : np.ndarray
        The full 288-element 0-based float64 NumPy array of adjusted closing prices.

    Returns
    -------
    Dict[str, np.ndarray]
        A dictionary containing the 72-element return arrays keyed by 'dotcom' and 'gfc'.
    """
    episode_returns: Dict[str, np.ndarray] = {}

    # Define the start indices for the two episodes
    episodes = {
        "dotcom": TAU_START_DOTCOM,
        "gfc": TAU_START_GFC
    }

    for ep_name, tau_start in episodes.items():
        # The forecast target indices (1-based) for tau=1 to 72 are:
        # tau_start to tau_start + 71
        # In 0-based Python indexing, this is:
        # current_month_idx = tau_start - 1 to tau_start + 70
        # previous_month_idx = tau_start - 2 to tau_start + 69

        # Define the slice boundaries for the current month (P_{t_forecast})
        curr_start_idx: int = tau_start - 1
        curr_end_idx: int = tau_start + 71 # Exclusive upper bound

        # Define the slice boundaries for the previous month (P_{t_prev})
        prev_start_idx: int = tau_start - 2
        prev_end_idx: int = tau_start + 70 # Exclusive upper bound

        # Extract the overlapping 72-element slices
        p_curr: np.ndarray = price_array[curr_start_idx:curr_end_idx]
        p_prev: np.ndarray = price_array[prev_start_idx:prev_end_idx]

        # Compute the simple arithmetic return: r = (P_curr - P_prev) / P_prev
        # Vectorized operation retains full float64 precision
        r_tau: np.ndarray = (p_curr - p_prev) / p_prev

        # Store the 72-element return array
        episode_returns[ep_name] = r_tau

    return episode_returns

# -------------------------------------------------------------------------------------------------------------------------------
# Task 5, Step 2: Compute Buy-and-Hold Portfolio Returns and Sharpe Ratios
# -------------------------------------------------------------------------------------------------------------------------------
def compute_buy_and_hold_sharpes(
    episode_returns: Dict[str, np.ndarray],
    config: Dict[str, Any]
) -> Dict[str, float]:
    """
    Computes the annualized Sharpe ratios for the buy-and-hold strategy (s_t = 1.0)
    for individual episodes and the combined sequence.

    Parameters
    ----------
    episode_returns : Dict[str, np.ndarray]
        Dictionary containing the 72-element raw return arrays for 'dotcom' and 'gfc'.
    config : Dict[str, Any]
        The master study configuration dictionary containing the annualization factor.

    Returns
    -------
    Dict[str, float]
        A dictionary containing the computed Sharpe ratios keyed by 'dotcom', 'gfc', and 'combined'.
    """
    # Retrieve the shared annualization factor (sqrt(12)) from the configuration
    # This ensures bit-for-bit consistency across all Sharpe calculations
    ann_factor: float = config["sharpe_params"]["sqrt_annualization_factor"]

    computed_sharpes: Dict[str, float] = {}

    # 1. Compute Sharpe ratios for individual episodes
    for ep_name in EPISODE_ORDER:
        r_raw: np.ndarray = episode_returns[ep_name]

        # Buy-and-hold strategy: s_t = 1.0, so r_port = r_raw
        r_port: np.ndarray = r_raw.copy()

        # Compute statistics using the centralized helper
        _, _, sharpe = compute_annualized_sharpe(r_port, ann_factor)
        computed_sharpes[ep_name] = sharpe

    # 2. Compute Sharpe ratio for the combined sequence
    # Concatenate the two 72-element arrays in the fixed EPISODE_ORDER
    r_combined_raw: np.ndarray = np.concatenate([
        episode_returns[EPISODE_ORDER[0]],
        episode_returns[EPISODE_ORDER[1]]
    ])

    # Buy-and-hold strategy: s_t = 1.0
    r_combined_port: np.ndarray = r_combined_raw.copy()

    # Recompute statistics over the full 144-element sequence
    # This correctly applies the n-1=143 denominator, avoiding the mathematical error of averaging
    _, _, combined_sharpe = compute_annualized_sharpe(r_combined_port, ann_factor)
    computed_sharpes["combined"] = combined_sharpe

    return computed_sharpes

# -------------------------------------------------------------------------------------------------------------------------------
# Task 5, Step 3: Gate the Pipeline on Calibration Targets
# -------------------------------------------------------------------------------------------------------------------------------
def validate_calibration_targets(
    computed_sharpes: Dict[str, float],
    config: Dict[str, Any]
) -> CalibrationReport:
    """
    Validates the computed buy-and-hold Sharpe ratios against the study's targets
    using a strict +/- 0.02 absolute tolerance.

    Parameters
    ----------
    computed_sharpes : Dict[str, float]
        The computed Sharpe ratios for 'dotcom', 'gfc', and 'combined'.
    config : Dict[str, Any]
        The master study configuration dictionary containing the target values.

    Returns
    -------
    CalibrationReport
        A structured report documenting the calibration results.

    Raises
    ------
    CalibrationError
        If any computed Sharpe ratio deviates from its target by more than 0.02.
    """
    # Extract the target Sharpe ratios from the configuration
    target_sharpes: Dict[str, float] = {
        "dotcom": config["sharpe_params"]["bah_sharpe_dotcom"],
        "gfc": config["sharpe_params"]["bah_sharpe_gfc"],
        "combined": config["sharpe_params"]["bah_sharpe_combined"]
    }

    deviations: Dict[str, float] = {}
    errors: List[str] = []

    # Define the absolute tolerance specified in the study
    absolute_tolerance: float = 0.02

    # Evaluate each target independently
    for key in ["dotcom", "gfc", "combined"]:
        computed_val: float = computed_sharpes[key]
        target_val: float = target_sharpes[key]

        # Calculate the absolute deviation
        deviation: float = abs(computed_val - target_val)
        deviations[key] = deviation

        # Check if the deviation exceeds the allowed tolerance
        if deviation > absolute_tolerance:
            errors.append(
                f"Episode '{key}': Computed SR = {computed_val:.4f}, "
                f"Target = {target_val:.4f}, Deviation = {deviation:.4f} > {absolute_tolerance}"
            )

    # If any assertions failed, raise the custom CalibrationError and halt the pipeline
    if errors:
        error_msg: str = (
            "FATAL DATA VERSION DISCREPANCY: Buy-and-hold calibration targets failed.\n"
            + "\n".join(errors) + "\n"
            "Yahoo Finance Adj Close was likely downloaded on a date different from the original study. "
            "Archive the raw CSV and re-download a matching snapshot. "
            "Downstream comparisons to reported Tables 1 and 3 are invalid until calibration passes."
        )
        raise CalibrationError(error_msg)

    logger.info("Buy-and-hold calibration passed successfully. Data version is consistent.")

    # Return the structured report on success
    return CalibrationReport(
        computed_sharpes=computed_sharpes,
        target_sharpes=target_sharpes,
        deviations=deviations,
        passed=True
    )

# -------------------------------------------------------------------------------------------------------------------------------
# Task 5, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def compute_and_validate_calibration_benchmarks(
    merged_container: Any,
    config: Dict[str, Any]
) -> CalibrationReport:
    """
    Orchestrates the computation and validation of the buy-and-hold Sharpe ratio benchmarks.

    This function sequentially extracts the raw returns, computes the portfolio statistics
    for the individual and combined episodes, and strictly gates the pipeline against
    the study's calibration targets to ensure data version consistency.

    Parameters
    ----------
    merged_container : MergedDataContainer
        The validated container holding the 0-based float64 price array.
    config : Dict[str, Any]
        The master study configuration dictionary.

    Returns
    -------
    CalibrationReport
        A structured report documenting the successful calibration results.

    Raises
    ------
    CalibrationError
        If the data version is inconsistent and the calibration targets are not met.
    """
    # Extract the 0-based price array from the container
    price_array: np.ndarray = merged_container.price_array

    # Step 1: Compute the 72-element realized monthly return sequence for each episode
    episode_returns: Dict[str, np.ndarray] = compute_episode_raw_returns(price_array)

    # Step 2: Compute Buy-and-Hold Portfolio Returns and Sharpe Ratios
    computed_sharpes: Dict[str, float] = compute_buy_and_hold_sharpes(episode_returns, config)

    # Step 3: Gate the Pipeline on Calibration Targets
    # This will raise a CalibrationError if the data version is incorrect
    calibration_report: CalibrationReport = validate_calibration_targets(computed_sharpes, config)

    return calibration_report
````

````python
# Task 6 — Implement the NORMALIZE Algorithm (Single Evaluation Step)

# ==============================================================================
# Task 6: Implement the NORMALIZE Algorithm (Single Evaluation Step)
# ==============================================================================

# Define authoritative module-level constants for episode start indices
TAU_START_DOTCOM: int = 121
TAU_START_GFC: int = 217

@dataclass
class NormalizedDataContainer:
    """
    A structured container for the normalized price and EPS arrays for a single evaluation step.

    Attributes
    ----------
    p_prime_full : np.ndarray
        The full-precision normalized price array (P'_k).
    eps_prime_full : np.ndarray
        The full-precision normalized EPS array (EPS'_k).
    p_prime_rounded : np.ndarray
        The normalized price array rounded to 2 decimal places for serialization.
    eps_prime_rounded : np.ndarray
        The normalized EPS array rounded to 2 decimal places for serialization.
    t0 : int
        The 1-based window start index used for this normalization step.
    p0 : float
        The shared scalar base price (P_0) used as the divisor.
    """
    p_prime_full: np.ndarray = field(default_factory=lambda: np.array([]))
    eps_prime_full: np.ndarray = field(default_factory=lambda: np.array([]))
    p_prime_rounded: np.ndarray = field(default_factory=lambda: np.array([]))
    eps_prime_rounded: np.ndarray = field(default_factory=lambda: np.array([]))
    t0: int = 0
    p0: float = 0.0

# -------------------------------------------------------------------------------------------------------------------------------
# Helper Function: Centralized Window-Start Index Computation
# -------------------------------------------------------------------------------------------------------------------------------
def compute_window_start_index(tau: int, episode: str, w: int) -> int:
    """
    Computes the 1-based window-start index t_0(tau) using the governing formula.

    This helper centralizes the index arithmetic to ensure Tasks 6, 9, 11, and 16
    all use the exact same logic, preventing silent misalignment.

    Formula: t_0(tau) = tau_e,start - W + tau - 1

    Parameters
    ----------
    tau : int
        The 1-based evaluation step index (1 to 72).
    episode : str
        The episode identifier ('dotcom' or 'gfc').
    w : int
        The context window length (36 or 60).

    Returns
    -------
    int
        The 1-based window-start index t_0.

    Raises
    ------
    ValueError
        If an invalid episode identifier is provided.
    """
    if episode == "dotcom":
        tau_start: int = TAU_START_DOTCOM
    elif episode == "gfc":
        tau_start: int = TAU_START_GFC
    else:
        raise ValueError(f"Invalid episode identifier: '{episode}'. Must be 'dotcom' or 'gfc'.")

    # Apply the governing formula exactly as specified
    t0: int = tau_start - w + tau - 1

    return t0

# -------------------------------------------------------------------------------------------------------------------------------
# Task 6, Step 1: Compute the Window-Start Index and Extract Sub-Windows
# -------------------------------------------------------------------------------------------------------------------------------
def extract_raw_sub_windows(
    tau: int,
    episode: str,
    w: int,
    price_array: np.ndarray,
    eps_array: np.ndarray
) -> Tuple[int, float, np.ndarray, np.ndarray]:
    """
    Computes the window-start index, extracts the W-length raw price and EPS
    sub-windows, and fixes the shared normalization base P_0.

    Parameters
    ----------
    tau : int
        The 1-based evaluation step index (1 to 72).
    episode : str
        The episode identifier ('dotcom' or 'gfc').
    w : int
        The context window length (36 or 60).
    price_array : np.ndarray
        The full 288-element 0-based float64 NumPy array of raw prices.
    eps_array : np.ndarray
        The full 288-element 0-based float64 NumPy array of raw EPS values.

    Returns
    -------
    Tuple[int, float, np.ndarray, np.ndarray]
        A tuple containing (t0, P_0, price_sub_window, eps_sub_window).

    Raises
    ------
    ValueError
        If the extracted sub-windows do not have exactly length W.
    """
    # Compute the 1-based window-start index using the centralized helper
    t0: int = compute_window_start_index(tau, episode, w)

    # Define the 0-based slice boundaries
    # The 1-based window occupies t in {t0, t0+1, ..., t0+W-1}
    # Under the t-1 convention, this is the 0-based slice [t0-1 : t0-1+W]
    slice_start: int = t0 - 1
    slice_end: int = t0 - 1 + w

    # Extract the W-length sub-windows
    price_sub_window: np.ndarray = price_array[slice_start:slice_end]
    eps_sub_window: np.ndarray = eps_array[slice_start:slice_end]

    # Assert that the slice lengths are exactly W to guard against boundary errors
    if len(price_sub_window) != w or len(eps_sub_window) != w:
        raise ValueError(f"Sub-window extraction failed: Expected length {w}, got {len(price_sub_window)}.")

    # Fix the shared normalization base P_0
    # P_0 is the price at the first month of the context window (index 0 of the slice)
    p0: float = float(price_sub_window[0])

    return t0, p0, price_sub_window, eps_sub_window

# -------------------------------------------------------------------------------------------------------------------------------
# Task 6, Step 2: Apply Normalization Equations and Verify Invariants
# -------------------------------------------------------------------------------------------------------------------------------
def apply_normalization_and_verify(
    p0: float,
    price_sub_window: np.ndarray,
    eps_sub_window: np.ndarray
) -> Tuple[np.ndarray, np.ndarray]:
    """
    Applies the normalization equations N1 and N2 using the shared base P_0,
    and mathematically verifies the two required invariants.

    Parameters
    ----------
    p0 : float
        The shared scalar base price (P_0) used as the divisor.
    price_sub_window : np.ndarray
        The W-length raw price sub-window.
    eps_sub_window : np.ndarray
        The W-length raw EPS sub-window.

    Returns
    -------
    Tuple[np.ndarray, np.ndarray]
        A tuple containing the full-precision normalized arrays (P'_k, EPS'_k).

    Raises
    ------
    AssertionError
        If either invariant (P'_1 == 100.0 or exact P/E preservation) fails.
    """
    # Apply Equation N1: P'_k = 100 * P_{t0+k-1} / P_0
    # Vectorized operation across the entire W-length array
    p_prime_full: np.ndarray = 100.0 * price_sub_window / p0

    # Apply Equation N2: EPS'_k = 100 * EPS_{t0+k-1} / P_0
    # CRITICAL: EPS is divided by the shared base P_0, NOT by EPS_{t0}
    eps_prime_full: np.ndarray = 100.0 * eps_sub_window / p0

    # Invariant (i): P'_1 = 100.0 exactly by construction
    # This is an algebraic identity, so exact float equality is required
    if p_prime_full[0] != 100.0:
        raise AssertionError(f"Invariant (i) failed: P'_1 = {p_prime_full[0]}, expected exactly 100.0.")

    # Invariant (ii): Exact P/E preservation
    # P'_k / EPS'_k must equal P_{t0+k-1} / EPS_{t0+k-1} for all k
    # Use numpy.allclose with a tight absolute tolerance (1e-10) for floating-point round-off
    normalized_pe: np.ndarray = p_prime_full / eps_prime_full
    raw_pe: np.ndarray = price_sub_window / eps_sub_window

    if not np.allclose(normalized_pe, raw_pe, rtol=0, atol=1e-10):
        raise AssertionError("Invariant (ii) failed: P/E ratio was not preserved exactly during normalization.")

    return p_prime_full, eps_prime_full

# -------------------------------------------------------------------------------------------------------------------------------
# Task 6, Step 3: Produce Serialization-Ready Arrays and Enforce Boundaries
# -------------------------------------------------------------------------------------------------------------------------------
def finalize_normalization_and_enforce_bounds(
    tau: int,
    episode: str,
    w: int,
    t0: int,
    p0: float,
    p_prime_full: np.ndarray,
    eps_prime_full: np.ndarray
) -> NormalizedDataContainer:
    """
    Rounds the normalized arrays to 2 decimal places for serialization, enforces
    the upper-boundary forecast-index constraint, and packages the results.

    Parameters
    ----------
    tau : int
        The 1-based evaluation step index.
    episode : str
        The episode identifier.
    w : int
        The context window length.
    t0 : int
        The 1-based window start index.
    p0 : float
        The shared scalar base price.
    p_prime_full : np.ndarray
        The full-precision normalized price array.
    eps_prime_full : np.ndarray
        The full-precision normalized EPS array.

    Returns
    -------
    NormalizedDataContainer
        The structured container holding all four arrays and metadata.

    Raises
    ------
    IndexError
        If the forecast target index exceeds the dataset size (288).
    """
    # Enforce the upper-boundary forecast-index constraint: t_0(tau) + W <= 288
    # This is a dynamic defense-in-depth check, binding for GFC at tau=72
    forecast_target_index: int = t0 + w
    max_dataset_index: int = 288

    if forecast_target_index > max_dataset_index:
        raise IndexError(
            f"Forecast target t0+W={forecast_target_index} exceeds dataset size "
            f"{max_dataset_index} for tau={tau}, episode={episode}, W={w}."
        )

    # Produce serialization-ready arrays rounded to exactly 2 decimal places
    # Use vectorized numpy.round, NOT Python scalar round
    p_prime_rounded: np.ndarray = np.round(p_prime_full, 2)
    eps_prime_rounded: np.ndarray = np.round(eps_prime_full, 2)

    # Package the dual-precision arrays and metadata into the container
    return NormalizedDataContainer(
        p_prime_full=p_prime_full,
        eps_prime_full=eps_prime_full,
        p_prime_rounded=p_prime_rounded,
        eps_prime_rounded=eps_prime_rounded,
        t0=t0,
        p0=p0
    )

# -------------------------------------------------------------------------------------------------------------------------------
# Task 6, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def execute_normalization_step(
    tau: int,
    episode: str,
    w: int,
    merged_container: Any
) -> NormalizedDataContainer:
    """
    Orchestrates the NORMALIZE algorithm for a single evaluation step.

    This function sequentially computes the window-start index, extracts the raw
    sub-windows, applies the normalization equations with a shared base, verifies
    mathematical invariants, and produces dual-precision arrays while enforcing
    boundary constraints.

    Parameters
    ----------
    tau : int
        The 1-based evaluation step index (1 to 72).
    episode : str
        The episode identifier ('dotcom' or 'gfc').
    w : int
        The context window length (36 or 60).
    merged_container : MergedDataContainer
        The validated container holding the full 288-element raw price and EPS arrays.

    Returns
    -------
    NormalizedDataContainer
        A structured container holding the full-precision and rounded normalized
        arrays, along with the t0 index and P0 base used for the step.
    """
    # Extract the raw arrays from the merged container
    price_array: np.ndarray = merged_container.price_array
    eps_array: np.ndarray = merged_container.eps_array

    # Step 1: Compute the Window-Start Index and Extract Sub-Windows
    t0, p0, price_sub_window, eps_sub_window = extract_raw_sub_windows(
        tau, episode, w, price_array, eps_array
    )

    # Step 2: Apply Normalization Equations and Verify Invariants
    p_prime_full, eps_prime_full = apply_normalization_and_verify(
        p0, price_sub_window, eps_sub_window
    )

    # Step 3: Produce Serialization-Ready Arrays and Enforce Boundaries
    normalized_container: NormalizedDataContainer = finalize_normalization_and_enforce_bounds(
        tau, episode, w, t0, p0, p_prime_full, eps_prime_full
    )

    return normalized_container
````

````python
# Task 7 — Implement User Message Serialization and System Prompt Assembly

# ==============================================================================
# Task 7: Implement User Message Serialization and System Prompt Assembly
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 7, Step 1: Render the Normalized Series into a Fixed-Width Plain-Text Table
# -------------------------------------------------------------------------------------------------------------------------------
def serialize_user_message(
    condition: str,
    p_prime_rounded: np.ndarray,
    eps_prime_rounded: np.ndarray,
    config: Dict[str, Any]
) -> str:
    """
    Renders the normalized price and EPS arrays into a strict fixed-width plain-text
    table for the LLM user message payload.

    Parameters
    ----------
    condition : str
        The experimental condition ('A', 'B', 'C', or 'D').
    p_prime_rounded : np.ndarray
        The 1D float64 array of normalized prices, pre-rounded to 2 decimal places.
    eps_prime_rounded : np.ndarray
        The 1D float64 array of normalized EPS values, pre-rounded to 2 decimal places.
    config : Dict[str, Any]
        The master study configuration dictionary containing formatting parameters.

    Returns
    -------
    str
        The fully formatted, multi-line plain-text table string.

    Raises
    ------
    ValueError
        If the input arrays do not have the same length.
    """
    if len(p_prime_rounded) != len(eps_prime_rounded):
        raise ValueError("Serialization failed: Price and EPS arrays must have identical lengths.")

    # Extract formatting parameters from the configuration
    params: Dict[str, Any] = config["user_message_params"]
    idx_width: int = params["index_col_width"]
    val_width: int = params["value_col_width"]

    # Determine the correct column headers based on the condition
    # Condition A uses only Price; B, C, D use both Price and EPS
    if condition == "A":
        headers: List[str] = params["condition_A_columns"]
    elif condition in ["B", "C", "D"]:
        headers: List[str] = params["conditions_BCD_columns"]
    else:
        raise ValueError(f"Invalid condition: '{condition}'. Must be A, B, C, or D.")

    # Construct the header row
    # "t" is left-aligned; numeric headers are right-aligned to match the data
    header_row: str = f"{headers[0]:<{idx_width}}"
    for header in headers[1:]:
        header_row += f"{header:>{val_width}}"

    # Construct the separator line of dashes matching the total width
    total_width: int = idx_width + (len(headers) - 1) * val_width
    separator_row: str = "-" * total_width

    # Initialize the list of rows with the header and separator
    table_rows: List[str] = [header_row, separator_row]

    # Iterate over the arrays to construct the data rows
    w: int = len(p_prime_rounded)
    for i in range(w):
        # The display index k runs from 1 to W (anonymized time step)
        k: int = i + 1

        # Format the index left-aligned
        row_str: str = f"{k:<{idx_width}}"

        # Format the price value right-aligned, strictly enforcing 2 decimal places
        # .2f prevents trailing zeros from being dropped (e.g., 100.0 -> 100.00)
        row_str += f"{p_prime_rounded[i]:>{val_width}.2f}"

        # If condition B, C, or D, append the formatted EPS value
        if condition in ["B", "C", "D"]:
            row_str += f"{eps_prime_rounded[i]:>{val_width}.2f}"

        table_rows.append(row_str)

    # Join all rows with newline characters to form the final table string
    return "\n".join(table_rows)

# -------------------------------------------------------------------------------------------------------------------------------
# Task 7, Step 2: Assemble the Condition-Specific System Prompt
# -------------------------------------------------------------------------------------------------------------------------------
def assemble_system_prompts(config: Dict[str, Any]) -> Dict[str, str]:
    """
    Pre-assembles the system prompt strings for all four conditions based strictly
    on the boolean composition map defined in the configuration.

    Parameters
    ----------
    config : Dict[str, Any]
        The master study configuration dictionary.

    Returns
    -------
    Dict[str, str]
        A dictionary mapping condition strings ('A', 'B', 'C', 'D') to their
        fully assembled system prompt strings.
    """
    prompts_cache: Dict[str, str] = {}

    blocks: Dict[str, str] = config["prompt_blocks"]
    comp_map: Dict[str, Dict[str, Any]] = config["experiment_config"]["condition_block_map"]

    # Define the strict, verbatim order of potential blocks
    block_sequence: List[str] = [
        "block_1", "block_2", "block_3", "block_4", "block_5"
    ]

    # Map the block keys to their actual text content in the config
    block_text_map: Dict[str, str] = {
        "block_1": blocks["block_1_role_price"],
        "block_2": blocks["block_2_eps"],
        "block_3": blocks["block_3_stage_definitions"],
        "block_4": blocks["block_4_reflexivity"],
        "block_5": blocks["block_5_stage_reflexivity_map"]
    }

    # Assemble the prompt for each condition
    for condition in ["A", "B", "C", "D"]:
        condition_flags: Dict[str, Any] = comp_map[condition]
        assembled_parts: List[str] = []

        # 1. Append the theoretical blocks based on the boolean flags
        for block_key in block_sequence:
            if condition_flags.get(block_key, False):
                assembled_parts.append(block_text_map[block_key])

        # 2. Append the condition-specific task instruction
        task_instruction_key: str = f"task_instruction_{condition}"
        assembled_parts.append(blocks[task_instruction_key])

        # 3. Append the verbatim output schema (always included)
        assembled_parts.append(blocks["output_schema_verbatim"])

        # Join all parts with a double newline to preserve paragraph structure
        final_prompt: str = "\n\n".join(assembled_parts)
        prompts_cache[condition] = final_prompt

    return prompts_cache

# -------------------------------------------------------------------------------------------------------------------------------
# Task 7, Step 3: Verify System Prompt Size and Enforce Anonymization
# -------------------------------------------------------------------------------------------------------------------------------
def verify_prompts_and_anonymization(
    prompts_cache: Dict[str, str],
    config: Dict[str, Any]
) -> None:
    """
    Verifies the byte size of the assembled system prompts against targets and
    performs a strict regex scan to ensure no forbidden tokens leak identity or
    calendar information.

    Parameters
    ----------
    prompts_cache : Dict[str, str]
        The dictionary of pre-assembled system prompts.
    config : Dict[str, Any]
        The master study configuration dictionary.

    Raises
    ------
    ValueError
        If a prompt size deviates by >20% from the target, or if any forbidden
        token is detected in the text.
    """
    comp_map: Dict[str, Dict[str, Any]] = config["experiment_config"]["condition_block_map"]

    # 1. Verify Byte Size Targets
    for condition, prompt_text in prompts_cache.items():
        # Measure actual UTF-8 byte size in kilobytes
        actual_kb: float = len(prompt_text.encode("utf-8")) / 1024.0
        target_kb: float = comp_map[condition]["approx_size_kb"]

        # Calculate 20% tolerance bounds
        lower_bound: float = target_kb * 0.80
        upper_bound: float = target_kb * 1.20

        if not (lower_bound <= actual_kb <= upper_bound):
            raise ValueError(
                f"Prompt size verification failed for Condition {condition}: "
                f"Actual size {actual_kb:.2f} kB is outside the ±20% tolerance "
                f"band of target {target_kb:.2f} kB."
            )

    # 2. Enforce Anonymization Protocol
    # Compile a list of forbidden regex patterns (case-insensitive)
    forbidden_patterns: List[Pattern] = [
        re.compile(r"\bS&P\s*500\b", re.IGNORECASE),
        re.compile(r"\b\^GSPC\b", re.IGNORECASE),
        re.compile(r"\bdot-?com\b", re.IGNORECASE),
        re.compile(r"\bGFC\b", re.IGNORECASE),
        re.compile(r"\bfinancial crisis\b", re.IGNORECASE),
        re.compile(r"\bShiller\b", re.IGNORECASE),
        re.compile(r"\bYahoo\b", re.IGNORECASE),
        # Match any 4-digit year in the study range (1986-2009)
        re.compile(r"\b(19[8-9][0-9]|200[0-9])\b"),
        # Match English calendar months
        re.compile(r"\b(January|February|March|April|May|June|July|August|September|October|November|December)\b", re.IGNORECASE)
    ]

    # Scan all assembled system prompts for forbidden tokens
    for condition, prompt_text in prompts_cache.items():
        for pattern in forbidden_patterns:
            match = pattern.search(prompt_text)
            if match:
                raise ValueError(
                    f"FATAL ANONYMIZATION LEAK: Forbidden token '{match.group(0)}' "
                    f"detected in system prompt for Condition {condition}."
                )

    logger.info("System prompts assembled, sized, and anonymization verified successfully.")

def verify_user_message_anonymization(user_message: str) -> None:
    """
    Performs a strict regex scan on a serialized user message to ensure no
    forbidden tokens leak identity or calendar information.

    Parameters
    ----------
    user_message : str
        The serialized plain-text table string.

    Raises
    ------
    ValueError
        If any forbidden token is detected in the text.
    """
    forbidden_patterns: List[Pattern] = [
        re.compile(r"\bS&P\s*500\b", re.IGNORECASE),
        re.compile(r"\b\^GSPC\b", re.IGNORECASE),
        re.compile(r"\bdot-?com\b", re.IGNORECASE),
        re.compile(r"\bGFC\b", re.IGNORECASE),
        re.compile(r"\b(19[8-9][0-9]|200[0-9])\b"),
        re.compile(r"\b(January|February|March|April|May|June|July|August|September|October|November|December)\b", re.IGNORECASE)
    ]

    for pattern in forbidden_patterns:
        match = pattern.search(user_message)
        if match:
            raise ValueError(
                f"FATAL ANONYMIZATION LEAK: Forbidden token '{match.group(0)}' "
                f"detected in serialized user message."
            )

# -------------------------------------------------------------------------------------------------------------------------------
# Task 7, Orchestrator Functions
# -------------------------------------------------------------------------------------------------------------------------------
def pre_assemble_and_validate_prompts(config: Dict[str, Any]) -> Dict[str, str]:
    """
    Orchestrates the pre-assembly and validation of the system prompts.

    This function should be called exactly once before the main evaluation loop
    to cache the invariant system prompts and verify their integrity and anonymization.

    Parameters
    ----------
    config : Dict[str, Any]
        The master study configuration dictionary.

    Returns
    -------
    Dict[str, str]
        The validated dictionary cache mapping conditions to system prompt strings.
    """
    # Step 2: Assemble the prompts based on the composition map
    prompts_cache: Dict[str, str] = assemble_system_prompts(config)

    # Step 3a: Verify sizes and enforce anonymization on the system prompts
    verify_prompts_and_anonymization(prompts_cache, config)

    return prompts_cache

def generate_validated_user_message(
    condition: str,
    normalized_container: Any,
    config: Dict[str, Any]
) -> str:
    """
    Orchestrates the serialization and validation of the user message for a single step.

    This function is called inside the hot loop at every evaluation step tau.

    Parameters
    ----------
    condition : str
        The experimental condition ('A', 'B', 'C', or 'D').
    normalized_container : NormalizedDataContainer
        The container holding the rounded normalized arrays.
    config : Dict[str, Any]
        The master study configuration dictionary.

    Returns
    -------
    str
        The fully formatted and anonymization-verified plain-text table string.
    """
    # Extract the rounded arrays from the container
    p_prime_rounded: np.ndarray = normalized_container.p_prime_rounded
    eps_prime_rounded: np.ndarray = normalized_container.eps_prime_rounded

    # Step 1: Render the arrays into the fixed-width table
    user_message: str = serialize_user_message(
        condition, p_prime_rounded, eps_prime_rounded, config
    )

    # Step 3b: Enforce anonymization on the generated user message
    verify_user_message_anonymization(user_message)

    return user_message
````

````python
# Task 8 — Configure the Three LLM API Clients

# ==============================================================================
# Task 8: Configure the Three LLM API Clients
# ==============================================================================

@dataclass
class LLMClientConfig:
    """
    A structured container for an instantiated LLM client and its standardized adapter.

    Attributes
    ----------
    client : Any
        The instantiated provider-specific client object (e.g., openai.OpenAI).
    model_id : str
        The exact API model identifier string used for the calls.
    adapter : Callable[[str, str], Any]
        A standardized function that takes (system_prompt, user_message) and
        returns the provider-specific API response object.
    """
    client: Any
    model_id: str
    adapter: Callable[[str, str], Any]

# -------------------------------------------------------------------------------------------------------------------------------
# Task 8, Step 1: Configure the OpenAI GPT-5 Client
# -------------------------------------------------------------------------------------------------------------------------------
def configure_gpt5_client(config: Dict[str, Any]) -> LLMClientConfig:
    """
    Instantiates the OpenAI client, enforces verbatim GPT-5 parameters (temp=1.0),
    and constructs the standardized adapter for the Chat Completions API.

    Parameters
    ----------
    config : Dict[str, Any]
        The master study configuration dictionary.

    Returns
    -------
    LLMClientConfig
        The configured client and its standardized adapter function.

    Raises
    ------
    EnvironmentError
        If the OPENAI_API_KEY environment variable is missing.
    ImportError
        If the openai SDK is not installed or is an incompatible version.
    """
    # Retrieve the API key from the environment; never hardcode secrets
    api_key: str = os.environ.get("OPENAI_API_KEY")
    if not api_key:
        raise EnvironmentError("Missing required environment variable: OPENAI_API_KEY")

    try:
        import openai
        # Ensure the modern SDK (>= 1.0.0) is used, as the pre-1.0 API is incompatible
        if int(openai.__version__.split(".")[0]) < 1:
            raise ImportError(f"openai SDK version >= 1.0.0 required. Found {openai.__version__}")
    except ImportError as e:
        raise ImportError(f"Failed to import openai SDK: {e}")

    # Instantiate the OpenAI client
    client = openai.OpenAI(api_key=api_key)

    # Extract verbatim parameters from the configuration
    gpt5_config: Dict[str, Any] = config["llm_api_config"]["gpt5"]
    model_id: str = gpt5_config["api_model_id"]
    temperature: float = gpt5_config["temperature"]
    max_tokens: int = gpt5_config["max_tokens"]

    # Emit the mandatory stochasticity notice for reproducibility tracking
    logger.warning(
        "REPRODUCIBILITY NOTICE: GPT-5 temperature=1.0, no seed; outputs are single "
        "stochastic draws; exact numerical reproduction is categorically impossible."
    )

    # Define the standardized adapter function for GPT-5
    def gpt5_adapter(system_prompt: str, user_message: str) -> Any:
        """
        Executes the OpenAI Chat Completions API call with the correct message structure.
        """
        # Construct the payload: system prompt is an in-array system turn
        messages = [
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_message}
        ]

        # Execute the API call using the verbatim parameters
        response = client.chat.completions.create(
            model=model_id,
            messages=messages,
            temperature=temperature,
            max_tokens=max_tokens
        )
        return response

    return LLMClientConfig(client=client, model_id=model_id, adapter=gpt5_adapter)

# -------------------------------------------------------------------------------------------------------------------------------
# Task 8, Step 2: Configure the Anthropic Claude Sonnet 4.6 Client
# -------------------------------------------------------------------------------------------------------------------------------
def configure_claude_client(config: Dict[str, Any]) -> LLMClientConfig:
    """
    Instantiates the Anthropic client, enforces verbatim parameters (temp=0),
    and constructs the standardized adapter for the Messages API.

    Parameters
    ----------
    config : Dict[str, Any]
        The master study configuration dictionary.

    Returns
    -------
    LLMClientConfig
        The configured client and its standardized adapter function.

    Raises
    ------
    EnvironmentError
        If the ANTHROPIC_API_KEY environment variable is missing.
    ImportError
        If the anthropic SDK is not installed.
    """
    # Retrieve the API key from the environment
    api_key: str = os.environ.get("ANTHROPIC_API_KEY")
    if not api_key:
        raise EnvironmentError("Missing required environment variable: ANTHROPIC_API_KEY")

    try:
        import anthropic
    except ImportError as e:
        raise ImportError(f"Failed to import anthropic SDK: {e}")

    # Instantiate the Anthropic client
    # The SDK automatically handles the required 'anthropic-version: 2023-06-01' header
    client = anthropic.Anthropic(api_key=api_key)

    # Extract verbatim parameters from the configuration
    claude_config: Dict[str, Any] = config["llm_api_config"]["claude_sonnet_4_6"]
    model_id: str = claude_config["api_model_id"]
    temperature: float = float(claude_config["temperature"])
    max_tokens: int = claude_config["max_tokens"]

    # Define the standardized adapter function for Claude
    def claude_adapter(system_prompt: str, user_message: str) -> Any:
        """
        Executes the Anthropic Messages API call with the correct top-level system parameter.
        """
        # Construct the payload: messages list contains ONLY the user turn
        messages = [
            {"role": "user", "content": user_message}
        ]

        # Execute the API call, passing the system prompt as a TOP-LEVEL parameter
        response = client.messages.create(
            model=model_id,
            system=system_prompt,
            messages=messages,
            temperature=temperature,
            max_tokens=max_tokens
        )
        return response

    return LLMClientConfig(client=client, model_id=model_id, adapter=claude_adapter)

# -------------------------------------------------------------------------------------------------------------------------------
# Task 8, Step 3: Configure the Google Gemini 3 Pro Client
# -------------------------------------------------------------------------------------------------------------------------------
def configure_gemini_client(config: Dict[str, Any]) -> LLMClientConfig:
    """
    Instantiates the Google GenerativeAI client, handles model deprecation,
    enforces verbatim parameters (temp=0), and constructs the standardized adapter.

    Parameters
    ----------
    config : Dict[str, Any]
        The master study configuration dictionary.

    Returns
    -------
    LLMClientConfig
        The configured client and its standardized adapter function.

    Raises
    ------
    EnvironmentError
        If the GOOGLE_API_KEY environment variable is missing.
    ImportError
        If the google-generativeai SDK is not installed.
    """
    # Retrieve the API key from the environment
    api_key: str = os.environ.get("GOOGLE_API_KEY")
    if not api_key:
        raise EnvironmentError("Missing required environment variable: GOOGLE_API_KEY")

    try:
        import google.generativeai as genai
    except ImportError as e:
        raise ImportError(f"Failed to import google-generativeai SDK: {e}")

    # Configure the global GenerativeAI client
    genai.configure(api_key=api_key)

    # Extract configuration and handle documented deprecation
    gemini_config: Dict[str, Any] = config["llm_api_config"]["gemini_3_pro"]
    original_model_id: str = gemini_config["api_model_id"]
    successor_model_id: str = gemini_config["successor_model_id"]

    # The original model was deprecated on 2026-03-09; we must use the successor
    active_model_id: str = successor_model_id

    logger.warning(
        f"MODEL VERSION DRIFT: Original model '{original_model_id}' is deprecated. "
        f"Substituting successor '{active_model_id}'. Results may differ from the study."
    )

    temperature: float = float(gemini_config["temperature"])
    # Note the Google-specific parameter name: max_output_tokens
    max_output_tokens: int = gemini_config["max_output_tokens"]

    # Define the standardized adapter function for Gemini
    def gemini_adapter(system_prompt: str, user_message: str) -> Any:
        """
        Executes the Google GenerativeAI API call with the correct nested structure.
        """
        # Instantiate the model object, passing the system prompt via system_instruction
        model = genai.GenerativeModel(
            model_name=active_model_id,
            system_instruction=system_prompt
        )

        # Define the generation configuration
        generation_config = genai.types.GenerationConfig(
            temperature=temperature,
            max_output_tokens=max_output_tokens
        )

        # Construct the payload using Google's nested parts/text structure
        contents = [
            {"role": "user", "parts": [{"text": user_message}]}
        ]

        # Execute the API call
        response = model.generate_content(
            contents=contents,
            generation_config=generation_config
        )
        return response

    return LLMClientConfig(client=genai, model_id=active_model_id, adapter=gemini_adapter)

# -------------------------------------------------------------------------------------------------------------------------------
# Task 8, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def configure_all_llm_clients(config: Dict[str, Any]) -> Dict[str, LLMClientConfig]:
    """
    Orchestrates the instantiation and configuration of all three LLM API clients.

    This function sequentially configures the OpenAI, Anthropic, and Google clients,
    ensuring that all required environment variables are present and that the
    standardized adapter functions are ready for the evaluation loop.

    Parameters
    ----------
    config : Dict[str, Any]
        The master study configuration dictionary.

    Returns
    -------
    Dict[str, LLMClientConfig]
        A dictionary mapping the internal model keys ('gpt5', 'claude_sonnet_4_6',
        'gemini_3_pro') to their respective configured client objects.

    Raises
    ------
    EnvironmentError
        If any of the required API keys are missing from the environment.
    ImportError
        If any of the required provider SDKs are missing or incompatible.
    """
    logger.info("Initializing LLM API clients...")

    clients: Dict[str, LLMClientConfig] = {}

    # Step 1: Configure GPT-5
    clients["gpt5"] = configure_gpt5_client(config)

    # Step 2: Configure Claude Sonnet 4.6
    clients["claude_sonnet_4_6"] = configure_claude_client(config)

    # Step 3: Configure Gemini 3 Pro
    clients["gemini_3_pro"] = configure_gemini_client(config)

    logger.info("All LLM API clients configured successfully.")

    return clients
````

````python
# Task 9 — Implement the Single-Step LLM API Call with Exponential Backoff

# ==============================================================================
# Task 9: Implement the Single-Step LLM API Call with Exponential Backoff
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 9, Step 1: Perform Non-Network Preparation Deterministically
# -------------------------------------------------------------------------------------------------------------------------------
def prepare_api_call_deterministically(
    m: str,
    c: str,
    w: int,
    e: str,
    tau: int,
    merged_container: Any,
    config: Dict[str, Any],
    prompts_cache: Dict[str, str],
    clients: Dict[str, Any]
) -> Tuple[str, str, Any]:
    """
    Executes all deterministic data preparation steps prior to any network I/O.

    This function strictly separates logic errors (e.g., out-of-bounds indices,
    anonymization leaks) from transient network errors, ensuring that bugs are
    not masked by the retry loop.

    Parameters
    ----------
    m : str
        The model identifier ('gpt5', 'claude_sonnet_4_6', 'gemini_3_pro').
    c : str
        The experimental condition ('A', 'B', 'C', 'D').
    w : int
        The context window length (36 or 60).
    e : str
        The episode identifier ('dotcom' or 'gfc').
    tau : int
        The 1-based evaluation step index (1 to 72).
    merged_container : MergedDataContainer
        The validated container holding the raw price and EPS arrays.
    config : Dict[str, Any]
        The master study configuration dictionary.
    prompts_cache : Dict[str, str]
        The pre-assembled and validated system prompts cache.
    clients : Dict[str, LLMClientConfig]
        The dictionary of configured LLM clients.

    Returns
    -------
    Tuple[str, str, LLMClientConfig]
        A tuple containing the system_prompt, user_message, and the client config.

    Raises
    ------
    KeyError
        If the model or condition keys are not found in the caches.
    Exception
        Any deterministic error raised by the normalization or serialization routines.
    """
    # (i) Call the NORMALIZE routine to obtain the rounded normalized arrays
    # This will raise an IndexError if the forecast target exceeds the dataset boundary
    normalized_container = execute_normalization_step(tau, e, w, merged_container)

    # (ii) Serialize the user message and enforce anonymization
    # This will raise a ValueError if any forbidden tokens are detected
    user_message: str = generate_validated_user_message(c, normalized_container, config)

    # (iii) Retrieve the pre-assembled system prompt from the cache
    if c not in prompts_cache:
        raise KeyError(f"System prompt for condition '{c}' not found in cache.")
    system_prompt: str = prompts_cache[c]

    # (iv) Retrieve the pre-configured LLM client
    if m not in clients:
        raise KeyError(f"LLM client for model '{m}' not found in clients dictionary.")
    client_config = clients[m]

    return system_prompt, user_message, client_config

# -------------------------------------------------------------------------------------------------------------------------------
# Task 9, Step 2: Submit Prompt Pair with Exponential Backoff
# -------------------------------------------------------------------------------------------------------------------------------
def is_transient_error(e: Exception, provider: str) -> bool:
    """
    Determines if an exception is a transient network/server error that should be retried,
    based on the specific exception hierarchies of the different SDKs.

    Parameters
    ----------
    e : Exception
        The exception raised during the API call.
    provider : str
        The provider identifier ('openai', 'anthropic', 'google').

    Returns
    -------
    bool
        True if the error is transient, False if it is permanent.
    """
    error_str = str(e).lower()

    if provider == "openai":
        import openai
        # Retry on RateLimitError, APIConnectionError, InternalServerError (500), APIStatusError (503)
        if isinstance(e, (openai.RateLimitError, openai.APIConnectionError, openai.InternalServerError)):
            return True
        if isinstance(e, openai.APIStatusError) and e.status_code in [500, 502, 503, 504]:
            return True

    elif provider == "anthropic":
        import anthropic
        # Retry on RateLimitError, APIConnectionError, InternalServerError
        if isinstance(e, (anthropic.RateLimitError, anthropic.APIConnectionError, anthropic.InternalServerError)):
            return True
        if isinstance(e, anthropic.APIStatusError) and e.status_code in [500, 502, 503, 504]:
            return True

    elif provider == "google":
        from google.api_core import exceptions as google_exceptions
        # Retry on ResourceExhausted (429), ServiceUnavailable (503), DeadlineExceeded
        if isinstance(e, (google_exceptions.ResourceExhausted, google_exceptions.ServiceUnavailable, google_exceptions.DeadlineExceeded)):
            return True

    # If the error does not match known transient signatures, assume it is permanent (e.g., 400, 401)
    return False

def extract_raw_string_from_response(response: Any, provider: str) -> str:
    """
    Extracts the raw text string from the provider-specific response object.
    """
    try:
        if provider == "openai":
            return response.choices[0].message.content
        elif provider == "anthropic":
            return response.content[0].text
        elif provider == "google":
            return response.text
    except Exception as e:
        logger.error(f"Failed to extract text from {provider} response object: {e}")
        return ""
    return ""

def execute_api_call_with_backoff(
    system_prompt: str,
    user_message: str,
    client_config: Any,
    provider: str
) -> Tuple[Optional[str], int]:
    """
    Submits the prompt pair to the LLM API using a bounded exponential backoff
    with jitter for transient failures.

    Parameters
    ----------
    system_prompt : str
        The fully assembled system prompt.
    user_message : str
        The serialized user message table.
    client_config : LLMClientConfig
        The configured client and adapter.
    provider : str
        The provider identifier for exception handling.

    Returns
    -------
    Tuple[Optional[str], int]
        A tuple containing the raw response string (or None if all attempts fail)
        and the total number of attempts made.

    Raises
    ------
    Exception
        If a permanent, non-transient error occurs (e.g., 401 Unauthorized).
    """
    max_attempts: int = 5

    for attempt in range(1, max_attempts + 1):
        try:
            # Execute the API call using the standardized adapter
            response = client_config.adapter(system_prompt, user_message)

            # Extract the raw string from the provider-specific response object
            raw_string: str = extract_raw_string_from_response(response, provider)

            # Return the successful response and the attempt count
            return raw_string, attempt

        except Exception as e:
            # Check if the error is transient based on the provider's taxonomy
            if is_transient_error(e, provider):
                if attempt < max_attempts:
                    # Calculate exponential backoff with jitter: wait = 2^(a-1) * (1 + u)
                    u: float = random.uniform(0, 1)
                    wait_time: float = (2 ** (attempt - 1)) * (1 + u)

                    logger.warning(
                        f"Transient error on attempt {attempt}/{max_attempts} for {provider}: {e}. "
                        f"Retrying in {wait_time:.2f} seconds..."
                    )
                    time.sleep(wait_time)
                else:
                    logger.error(f"All {max_attempts} attempts failed for {provider} due to transient errors.")
                    return None, attempt
            else:
                # Permanent error (e.g., 400 Bad Request, 401 Unauthorized)
                # Do not retry; raise immediately to halt the pipeline
                logger.error(f"Permanent error encountered for {provider}: {e}. Halting.")
                raise e

    return None, max_attempts

# -------------------------------------------------------------------------------------------------------------------------------
# Task 9, Step 3: Persist Outcomes and Advance Loop
# -------------------------------------------------------------------------------------------------------------------------------
def get_completed_tuples(log_file_path: str) -> Set[Tuple[str, str, int, str, int]]:
    """
    Reads the persistent JSONL log file to build a set of already completed 5-tuples,
    enabling crash-recoverability and idempotency.
    """
    completed: Set[Tuple[str, str, int, str, int]] = set()
    if not os.path.exists(log_file_path):
        return completed

    with open(log_file_path, 'r', encoding='utf-8') as f:
        for line in f:
            if not line.strip():
                continue
            try:
                record = json.loads(line)
                # Reconstruct the 5-tuple identifier
                tup = (
                    record["model"],
                    record["condition"],
                    record["window"],
                    record["episode"],
                    record["tau"]
                )
                # Both SUCCESS and NULL outcomes indicate the step was fully processed
                if record.get("outcome") in ["SUCCESS", "NULL"]:
                    completed.add(tup)
            except json.JSONDecodeError:
                continue
    return completed

def persist_api_outcome(
    m: str,
    c: str,
    w: int,
    e: str,
    tau: int,
    attempt_count: int,
    raw_response: Optional[str],
    log_file_path: str
) -> None:
    """
    Appends the API outcome to a persistent JSONL log file, ensuring absolute
    durability via synchronous file I/O (fsync).

    Parameters
    ----------
    m, c, w, e, tau : str/int
        The 5-tuple identifier coordinates.
    attempt_count : int
        The number of API attempts made.
    raw_response : Optional[str]
        The raw response string, or None if all attempts failed.
    log_file_path : str
        The filesystem path to the JSONL log file.
    """
    # Determine the outcome status
    outcome: str = "SUCCESS" if raw_response is not None else "NULL"

    # Construct the log record
    record: Dict[str, Any] = {
        "timestamp_utc": datetime.now(timezone.utc).isoformat(),
        "model": m,
        "condition": c,
        "window": w,
        "episode": e,
        "tau": tau,
        "attempt_count": attempt_count,
        "outcome": outcome,
        "raw_response": raw_response
    }

    # Serialize the record to a JSON string
    record_str: str = json.dumps(record) + "\n"

    # Append to the log file and guarantee durability
    with open(log_file_path, 'a', encoding='utf-8') as f:
        f.write(record_str)
        f.flush()
        os.fsync(f.fileno())

# -------------------------------------------------------------------------------------------------------------------------------
# Task 9, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def execute_single_api_step(
    m: str,
    c: str,
    w: int,
    e: str,
    tau: int,
    merged_container: Any,
    config: Dict[str, Any],
    prompts_cache: Dict[str, str],
    clients: Dict[str, Any],
    log_file_path: str
) -> Tuple[Optional[str], int]:
    """
    Orchestrates the complete single-step LLM API call process.

    This function sequentially performs deterministic data preparation, executes the
    network call with exponential backoff, and durably persists the outcome to a log file.

    Parameters
    ----------
    m, c, w, e, tau : str/int
        The 5-tuple identifier coordinates.
    merged_container : MergedDataContainer
        The validated container holding the raw price and EPS arrays.
    config : Dict[str, Any]
        The master study configuration dictionary.
    prompts_cache : Dict[str, str]
        The pre-assembled and validated system prompts cache.
    clients : Dict[str, LLMClientConfig]
        The dictionary of configured LLM clients.
    log_file_path : str
        The filesystem path to the JSONL log file.

    Returns
    -------
    Tuple[Optional[str], int]
        A tuple containing the raw response string (or None) and the attempt count.
    """
    # Determine the provider string for exception handling
    provider_map = {
        "gpt5": "openai",
        "claude_sonnet_4_6": "anthropic",
        "gemini_3_pro": "google"
    }
    provider: str = provider_map.get(m, "unknown")

    # Step 1: Perform Non-Network Preparation Deterministically
    # Any exception here halts the pipeline immediately
    system_prompt, user_message, client_config = prepare_api_call_deterministically(
        m, c, w, e, tau, merged_container, config, prompts_cache, clients
    )

    # Step 2: Submit Prompt Pair with Exponential Backoff
    raw_response, attempt_count = execute_api_call_with_backoff(
        system_prompt, user_message, client_config, provider
    )

    # Step 3: Persist Outcomes and Advance Loop
    # Guarantees crash-recoverability via fsync
    persist_api_outcome(
        m, c, w, e, tau, attempt_count, raw_response, log_file_path
    )

    return raw_response, attempt_count
````

````python
# Task 10 — Implement JSON Response Parsing and Direction Extraction

# ==============================================================================
# Task 10: Implement JSON Response Parsing and Direction Extraction
# ==============================================================================

# Define the verbatim allowed set for the direction field
ALLOWED_DIRECTIONS = frozenset({"up", "neutral", "down"})
# Define the conservative fallback value
FALLBACK_DIRECTION = "neutral"

# -------------------------------------------------------------------------------------------------------------------------------
# Task 10, Step 1: Robustly Locate the JSON Object within the Raw Response
# -------------------------------------------------------------------------------------------------------------------------------
def locate_json_candidate(
    raw_response: Optional[str],
    m: str, c: str, w: int, e: str, tau: int
) -> Tuple[Optional[str], bool]:
    """
    Implements a three-tier extraction strategy to locate the JSON object within
    the raw LLM response, handling nulls, Markdown fences, and conversational text.

    Parameters
    ----------
    raw_response : Optional[str]
        The raw string response from the LLM API, or None if the call failed.
    m, c, w, e, tau : str/int
        The 5-tuple identifier coordinates for logging.

    Returns
    -------
    Tuple[Optional[str], bool]
        A tuple containing the candidate JSON string (or None if extraction fails)
        and a boolean indicating if a fallback was triggered.
    """
    # Short-circuit: If response is None or empty whitespace, apply fallback immediately
    if raw_response is None or not raw_response.strip():
        logger.warning(f"NO_JSON_OBJECT: Null or empty response for ({m}, {c}, {w}, {e}, {tau}).")
        return None, True

    cleaned_response: str = raw_response.strip()

    # Tier 1: Attempt to parse the stripped raw string directly
    # This is the safest path and avoids the nested-brace hazard
    try:
        json.loads(cleaned_response)
        return cleaned_response, False
    except json.JSONDecodeError:
        pass # Proceed to Tier 2

    # Tier 2: Strip Markdown code fences (```json ... ```) and retry
    # re.DOTALL ensures the regex matches across newlines
    fence_pattern = re.compile(r'^```(?:json)?\s*(.*?)\s*```$', re.DOTALL | re.IGNORECASE)
    match = fence_pattern.search(cleaned_response)
    if match:
        candidate: str = match.group(1).strip()
        try:
            json.loads(candidate)
            return candidate, False
        except json.JSONDecodeError:
            pass # Proceed to Tier 3

    # Tier 3: Last resort - extract substring between first '{' and last '}'
    # WARNING: This is susceptible to the nested-brace hazard if the rationale contains '}'
    first_brace: int = cleaned_response.find('{')
    last_brace: int = cleaned_response.rfind('}')

    if first_brace != -1 and last_brace != -1 and last_brace > first_brace:
        candidate = cleaned_response[first_brace:last_brace + 1]
        return candidate, False

    # If all tiers fail, log the failure and trigger fallback
    logger.warning(f"NO_JSON_OBJECT: Could not locate JSON structure for ({m}, {c}, {w}, {e}, {tau}).")
    return None, True

# -------------------------------------------------------------------------------------------------------------------------------
# Task 10, Step 2: Parse and Extract the `direction` Field
# -------------------------------------------------------------------------------------------------------------------------------
def parse_and_extract_direction(
    candidate_json: Optional[str],
    m: str, c: str, w: int, e: str, tau: int
) -> Tuple[Any, bool]:
    """
    Parses the candidate JSON string and extracts the value associated with the
    'direction' key, applying the conservative fallback on any structural failure.

    Parameters
    ----------
    candidate_json : Optional[str]
        The candidate JSON string located in Step 1.
    m, c, w, e, tau : str/int
        The 5-tuple identifier coordinates for logging.

    Returns
    -------
    Tuple[Any, bool]
        A tuple containing the extracted raw direction value (or the fallback string)
        and a boolean indicating if a fallback was triggered.
    """
    # If Step 1 already failed, propagate the fallback
    if candidate_json is None:
        return FALLBACK_DIRECTION, True

    try:
        # Parse the JSON string into a Python dictionary
        parsed_obj: Dict[str, Any] = json.loads(candidate_json)
    except json.JSONDecodeError as err:
        # Handle malformed JSON (e.g., missing quotes, trailing commas)
        logger.warning(f"JSON_PARSE_ERROR: {err} for ({m}, {c}, {w}, {e}, {tau}).")
        return FALLBACK_DIRECTION, True

    # Ensure the parsed object is a dictionary
    if not isinstance(parsed_obj, dict):
        logger.warning(f"JSON_PARSE_ERROR: Parsed JSON is not an object for ({m}, {c}, {w}, {e}, {tau}).")
        return FALLBACK_DIRECTION, True

    # Extract the 'direction' key
    if "direction" not in parsed_obj:
        logger.warning(f"MISSING_DIRECTION_KEY: 'direction' key absent for ({m}, {c}, {w}, {e}, {tau}).")
        return FALLBACK_DIRECTION, True

    raw_direction_value: Any = parsed_obj["direction"]

    return raw_direction_value, False

# -------------------------------------------------------------------------------------------------------------------------------
# Task 10, Step 3: Validate the Extracted Direction Against the Allowed Set
# -------------------------------------------------------------------------------------------------------------------------------
def validate_direction_value(
    raw_direction_value: Any,
    fallback_triggered_upstream: bool,
    m: str, c: str, w: int, e: str, tau: int
) -> Tuple[str, bool]:
    """
    Validates the extracted direction against the exact allowed set using
    case-sensitive equality, applying the fallback on any non-membership.

    Parameters
    ----------
    raw_direction_value : Any
        The raw value extracted from the JSON object.
    fallback_triggered_upstream : bool
        True if a fallback was already triggered in Step 1 or Step 2.
    m, c, w, e, tau : str/int
        The 5-tuple identifier coordinates for logging.

    Returns
    -------
    Tuple[str, bool]
        A tuple containing the validated direction string (guaranteed to be in
        the allowed set) and a boolean indicating if a fallback was triggered.
    """
    # If a fallback was already triggered, return it immediately
    if fallback_triggered_upstream:
        return FALLBACK_DIRECTION, True

    # Ensure the extracted value is a string before attempting string methods
    if not isinstance(raw_direction_value, str):
        logger.warning(
            f"INVALID_DIRECTION_VALUE: Expected string, got {type(raw_direction_value).__name__} "
            f"for ({m}, {c}, {w}, {e}, {tau}). Value: {raw_direction_value}"
        )
        return FALLBACK_DIRECTION, True

    # Strip surrounding whitespace (e.g., " up " -> "up")
    cleaned_direction: str = raw_direction_value.strip()

    # Strict, case-sensitive membership test against the verbatim allowed set
    # DO NOT lowercase the string; "Up" is a non-compliant response and must fail
    if cleaned_direction not in ALLOWED_DIRECTIONS:
        logger.warning(
            f"INVALID_DIRECTION_VALUE: '{cleaned_direction}' not in allowed set "
            f"for ({m}, {c}, {w}, {e}, {tau})."
        )
        return FALLBACK_DIRECTION, True

    # The value is valid and compliant
    return cleaned_direction, False

# -------------------------------------------------------------------------------------------------------------------------------
# Task 10, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def parse_and_validate_llm_response(
    raw_response: Optional[str],
    m: str, c: str, w: int, e: str, tau: int
) -> Tuple[str, bool]:
    """
    Orchestrates the robust parsing and validation of the LLM JSON response.

    This function sequentially locates the JSON candidate, extracts the direction
    field, and strictly validates it against the allowed set, applying a conservative
    fallback ("neutral") if any structural or compliance failure occurs.

    Parameters
    ----------
    raw_response : Optional[str]
        The raw string response from the LLM API.
    m, c, w, e, tau : str/int
        The 5-tuple identifier coordinates for logging.

    Returns
    -------
    Tuple[str, bool]
        A tuple containing the validated direction string (guaranteed to be "up",
        "neutral", or "down") and a boolean indicating if a fallback was applied.
    """
    # Step 1: Robustly Locate the JSON Object
    candidate_json, fallback_1 = locate_json_candidate(raw_response, m, c, w, e, tau)

    # Step 2: Parse and Extract the `direction` Field
    raw_direction, fallback_2 = parse_and_extract_direction(candidate_json, m, c, w, e, tau)

    # Combine fallback flags
    fallback_triggered: bool = fallback_1 or fallback_2

    # Step 3: Validate the Extracted Direction Against the Allowed Set
    validated_direction, final_fallback_flag = validate_direction_value(
        raw_direction, fallback_triggered, m, c, w, e, tau
    )

    return validated_direction, final_fallback_flag
````

````python
# Task 11 — Implement the Realized Monthly Return and Direction Label Computer

# ==============================================================================
# Task 11: Implement the Realized Monthly Return and Direction Label Computer
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 11, Step 1: Compute Forecast Target and Previous Month Indices
# -------------------------------------------------------------------------------------------------------------------------------
def compute_return_indices(tau: int, e: str, w: int) -> Tuple[int, int]:
    """
    Computes the 1-based indices for the forecast target month and the previous month,
    asserting that the target does not exceed the dataset boundary.

    Parameters
    ----------
    tau : int
        The 1-based evaluation step index (1 to 72).
    e : str
        The episode identifier ('dotcom' or 'gfc').
    w : int
        The context window length (36 or 60).

    Returns
    -------
    Tuple[int, int]
        A tuple containing the 1-based indices (t_forecast, t_prev).

    Raises
    ------
    IndexError
        If the forecast target index exceeds the maximum dataset size (288).
    """
    # Call the shared utility to get the 1-based window start index t_0(tau)
    # This guarantees alignment between the normalized prompt and the realized outcome
    t0: int = compute_window_start_index(tau, e, w)

    # Calculate the 1-based forecast target index (the month being predicted)
    # t_forecast = t_0(tau) + W
    t_forecast: int = t0 + w

    # Calculate the 1-based previous month index (the last month of the context window)
    # t_prev = t_0(tau) + W - 1
    t_prev: int = t0 + w - 1

    # Assert the upper boundary constraint
    # The dataset contains exactly 288 rows (Jan 1986 to Dec 2009)
    max_dataset_index: int = 288
    if t_forecast > max_dataset_index:
        raise IndexError(
            f"Realized return calculation failed: Forecast target index {t_forecast} "
            f"exceeds dataset boundary {max_dataset_index} for tau={tau}, episode={e}, W={w}."
        )

    return t_forecast, t_prev

# -------------------------------------------------------------------------------------------------------------------------------
# Task 11, Step 2: Compute the Realized Simple Monthly Return
# -------------------------------------------------------------------------------------------------------------------------------
def compute_raw_simple_return(t_forecast: int, t_prev: int, price_array: np.ndarray) -> float:
    """
    Computes the realized simple monthly return using raw, non-normalized prices
    at full float64 precision.

    Parameters
    ----------
    t_forecast : int
        The 1-based index of the forecast target month.
    t_prev : int
        The 1-based index of the previous month.
    price_array : np.ndarray
        The full 288-element 0-based float64 NumPy array of raw adjusted closing prices.

    Returns
    -------
    float
        The simple arithmetic monthly return (r_tau).
    """
    # Convert 1-based mathematical indices to 0-based Python array indices
    idx_forecast: int = t_forecast - 1
    idx_prev: int = t_prev - 1

    # Extract the raw prices from the array
    # Explicitly using raw prices, NOT normalized prices, per the study specification
    p_forecast: float = float(price_array[idx_forecast])
    p_prev: float = float(price_array[idx_prev])

    # Compute the simple arithmetic return: r_tau = (P_{t_forecast} - P_{t_prev}) / P_{t_prev}
    # A simple return (not log return) is mandatory to preserve linear portfolio scaling
    r_tau: float = (p_forecast - p_prev) / p_prev

    return r_tau

# -------------------------------------------------------------------------------------------------------------------------------
# Task 11, Step 3: Apply the Sign-Bucket Rule for Direction Labels
# -------------------------------------------------------------------------------------------------------------------------------
def bucket_return_to_direction_label(r_tau: float) -> str:
    """
    Converts the continuous realized return into a discrete ternary direction label
    using a strict sign-bucket rule with an exact-zero threshold.

    Parameters
    ----------
    r_tau : float
        The realized simple monthly return.

    Returns
    -------
    str
        The realized direction label ('up', 'down', or 'neutral').
    """
    # Apply the strict sign-bucket rule
    # No epsilon tolerance band is permitted around zero
    if r_tau > 0.0:
        return "up"
    elif r_tau < 0.0:
        return "down"
    else:
        # Exact floating-point equality to zero
        # While empirically rare for monthly S&P 500 returns, this ensures totality
        return "neutral"

# -------------------------------------------------------------------------------------------------------------------------------
# Task 11, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def compute_realized_return_and_label(
    tau: int,
    e: str,
    w: int,
    merged_container: Any
) -> Tuple[float, str]:
    """
    Orchestrates the computation of the realized monthly return and its corresponding
    direction label for a single evaluation step.

    This function sequentially calculates the aligned temporal indices, computes the
    simple arithmetic return using raw prices, and applies the strict sign-bucket rule.

    Parameters
    ----------
    tau : int
        The 1-based evaluation step index (1 to 72).
    e : str
        The episode identifier ('dotcom' or 'gfc').
    w : int
        The context window length (36 or 60).
    merged_container : MergedDataContainer
        The validated container holding the raw price array.

    Returns
    -------
    Tuple[float, str]
        A tuple containing the continuous realized return (r_tau) and the
        discrete realized direction label (d_tau).
    """
    # Extract the raw 0-based price array from the container
    price_array: np.ndarray = merged_container.price_array

    # Step 1: Compute Forecast Target and Previous Month Indices
    # This guarantees alignment with the normalized prompt generated in Task 6
    t_forecast, t_prev = compute_return_indices(tau, e, w)

    # Step 2: Compute the Realized Simple Monthly Return
    # Uses raw prices and maintains full float64 precision
    r_tau: float = compute_raw_simple_return(t_forecast, t_prev, price_array)

    # Step 3: Apply the Sign-Bucket Rule for Direction Labels
    # Converts the continuous return to 'up', 'down', or 'neutral'
    d_tau: str = bucket_return_to_direction_label(r_tau)

    return r_tau, d_tau
````

````python
# Task 12 — Implement Portfolio Position Mapping and Per-Step Return Accumulator

# ==============================================================================
# Task 12: Implement Portfolio Position Mapping and Per-Step Return Accumulator
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 12, Step 1: Map direction forecasts to portfolio positions
# -------------------------------------------------------------------------------------------------------------------------------
def map_direction_to_position(d_hat_tau: str, config: Dict[str, Any]) -> float:
    """
    Maps the parsed direction forecast to a portfolio position scalar using the
    verbatim deterministic position map defined in the study configuration.

    Parameters
    ----------
    d_hat_tau : str
        The validated direction forecast string ('up', 'neutral', or 'down').
    config : Dict[str, Any]
        The master study configuration dictionary containing the position map.

    Returns
    -------
    float
        The portfolio position scalar (1.0, 0.5, or 0.0).

    Raises
    ------
    KeyError
        If the required configuration keys are missing.
    TypeError
        If the input direction is not a string.
    """
    # Validate that the input direction is a string
    if not isinstance(d_hat_tau, str):
        # Raise a TypeError if the input is invalid
        raise TypeError(f"Expected string for d_hat_tau, got {type(d_hat_tau).__name__}")

    # Retrieve the position mapping dictionary from the configuration
    # This ensures we use the single source of truth for the mapping logic
    position_map: Dict[str, float] = config["sharpe_params"]["position_map"]

    # Perform a safe dictionary lookup with a conservative fallback to 0.5 ('neutral')
    # Task 10 guarantees validity, but this provides defense-in-depth
    s_tau_raw: float = position_map.get(d_hat_tau, 0.5)

    # Explicitly cast the retrieved value to a float to guarantee type stability
    s_tau: float = float(s_tau_raw)

    # Return the mapped portfolio position scalar
    return s_tau

# -------------------------------------------------------------------------------------------------------------------------------
# Task 12, Step 2: Compute the per-step portfolio return
# -------------------------------------------------------------------------------------------------------------------------------
def compute_portfolio_return(s_tau: float, r_tau: float) -> float:
    """
    Computes the per-step portfolio return by scaling the realized simple monthly
    return by the portfolio position scalar.

    Equation: r^port_tau = s_tau * r_tau

    Parameters
    ----------
    s_tau : float
        The portfolio position scalar (1.0, 0.5, or 0.0).
    r_tau : float
        The realized simple monthly return at full precision.

    Returns
    -------
    float
        The computed portfolio return at full precision.

    Raises
    ------
    TypeError
        If either input is not a float or int.
    """
    # Validate that the position scalar is a numeric type
    if not isinstance(s_tau, (float, int)):
        # Raise a TypeError to prevent invalid arithmetic operations
        raise TypeError(f"Expected float for s_tau, got {type(s_tau).__name__}")

    # Validate that the realized return is a numeric type
    if not isinstance(r_tau, (float, int)):
        # Raise a TypeError to prevent invalid arithmetic operations
        raise TypeError(f"Expected float for r_tau, got {type(r_tau).__name__}")

    # Execute the portfolio return equation: r^port_tau = s_tau * r_tau
    # This naturally handles the cash position (0.0 * r_tau = 0.0)
    r_port_tau: float = float(s_tau * r_tau)

    # Return the computed portfolio return
    return r_port_tau

# -------------------------------------------------------------------------------------------------------------------------------
# Task 12, Step 3: Accumulate per-step outputs into fixed-length arrays
# -------------------------------------------------------------------------------------------------------------------------------
@dataclass
class CellAccumulator:
    """
    A structured container that pre-allocates and holds the accumulated time series
    data for a single experimental cell (model, condition, window, episode).

    Attributes
    ----------
    r_port_array : np.ndarray
        A pre-allocated length-72 float64 NumPy array for portfolio returns.
    d_hat_list : List[Optional[str]]
        A pre-allocated length-72 Python list for predicted direction strings.
    d_list : List[Optional[str]]
        A pre-allocated length-72 Python list for realized direction strings.
    """
    # Pre-allocate a NumPy array of length 72, initialized with NaNs to detect missing writes
    r_port_array: np.ndarray = field(default_factory=lambda: np.full(72, np.nan, dtype=np.float64))
    # Pre-allocate a Python list of length 72, initialized with Nones for predicted directions
    d_hat_list: List[Optional[str]] = field(default_factory=lambda: [None] * 72)
    # Pre-allocate a Python list of length 72, initialized with Nones for realized directions
    d_list: List[Optional[str]] = field(default_factory=lambda: [None] * 72)

def initialize_cell_accumulator() -> CellAccumulator:
    """
    Initializes and returns a fresh, pre-allocated CellAccumulator instance.

    Returns
    -------
    CellAccumulator
        A new accumulator instance ready to receive 72 steps of data.
    """
    # Instantiate and return the dataclass, which automatically handles pre-allocation
    return CellAccumulator()

def update_cell_accumulator(
    accumulator: CellAccumulator,
    tau: int,
    r_port_tau: float,
    d_hat_tau: str,
    d_tau: str
) -> None:
    """
    Writes the per-step outputs into the pre-allocated arrays at the specific
    positional index (tau - 1), ensuring order independence.

    Parameters
    ----------
    accumulator : CellAccumulator
        The active accumulator instance for the current experimental cell.
    tau : int
        The 1-based evaluation step index (1 to 72).
    r_port_tau : float
        The computed portfolio return for the current step.
    d_hat_tau : str
        The predicted direction string for the current step.
    d_tau : str
        The realized direction string for the current step.

    Raises
    ------
    IndexError
        If tau is outside the valid range of 1 to 72.
    """
    # Enforce strict bounds checking on the 1-based tau index
    if not (1 <= tau <= 72):
        # Raise an IndexError to prevent memory corruption or out-of-bounds writes
        raise IndexError(f"Evaluation step tau must be between 1 and 72, got {tau}")

    # Convert the 1-based tau index to a 0-based array index
    index: int = tau - 1

    # Write the portfolio return to the NumPy array at the specific index
    accumulator.r_port_array[index] = r_port_tau

    # Write the predicted direction to the Python list at the specific index
    accumulator.d_hat_list[index] = d_hat_tau

    # Write the realized direction to the Python list at the specific index
    accumulator.d_list[index] = d_tau

def finalize_cell_accumulation(
    accumulator: CellAccumulator,
    m: str,
    c: str,
    w: int,
    e: str,
    results_dict: Dict[Tuple[str, str, int, str], CellAccumulator]
) -> None:
    """
    Validates that the accumulator is fully populated and stores it in the global
    results dictionary under the 4-tuple cell key.

    Parameters
    ----------
    accumulator : CellAccumulator
        The fully populated accumulator instance.
    m, c, w, e : str/int
        The 4-tuple identifier coordinates (model, condition, window, episode).
    results_dict : Dict[Tuple[str, str, int, str], CellAccumulator]
        The global dictionary storing all completed cell accumulators.

    Raises
    ------
    ValueError
        If any NaNs or Nones remain in the accumulator, indicating skipped steps.
    """
    # Check for any remaining NaNs in the portfolio return array
    if np.isnan(accumulator.r_port_array).any():
        # Raise a ValueError if the array is incomplete
        raise ValueError(f"Accumulator for cell ({m}, {c}, {w}, {e}) contains missing portfolio returns.")

    # Check for any remaining Nones in the predicted direction list
    if None in accumulator.d_hat_list:
        # Raise a ValueError if the list is incomplete
        raise ValueError(f"Accumulator for cell ({m}, {c}, {w}, {e}) contains missing predicted directions.")

    # Check for any remaining Nones in the realized direction list
    if None in accumulator.d_list:
        # Raise a ValueError if the list is incomplete
        raise ValueError(f"Accumulator for cell ({m}, {c}, {w}, {e}) contains missing realized directions.")

    # Construct the 4-tuple cell key
    cell_key: Tuple[str, str, int, str] = (m, c, w, e)

    # Store the validated accumulator in the global results dictionary
    results_dict[cell_key] = accumulator

# -------------------------------------------------------------------------------------------------------------------------------
# Task 12, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def execute_portfolio_mapping_and_accumulation(
    tau: int,
    d_hat_tau: str,
    r_tau: float,
    d_tau: str,
    config: Dict[str, Any],
    accumulator: CellAccumulator
) -> float:
    """
    Orchestrates the portfolio position mapping, return computation, and state
    accumulation for a single evaluation step.

    This function sequentially maps the forecast to a position, computes the
    portfolio return, and mutates the provided CellAccumulator in place.

    Parameters
    ----------
    tau : int
        The 1-based evaluation step index (1 to 72).
    d_hat_tau : str
        The predicted direction string from the LLM.
    r_tau : float
        The realized simple monthly return.
    d_tau : str
        The realized direction string.
    config : Dict[str, Any]
        The master study configuration dictionary.
    accumulator : CellAccumulator
        The active accumulator instance for the current experimental cell.

    Returns
    -------
    float
        The computed portfolio return for the current step (r^port_tau).
    """
    # Step 1: Map the predicted direction to a portfolio position scalar
    s_tau: float = map_direction_to_position(d_hat_tau, config)

    # Step 2: Compute the per-step portfolio return
    r_port_tau: float = compute_portfolio_return(s_tau, r_tau)

    # Step 3: Accumulate the outputs into the pre-allocated arrays
    # This mutates the accumulator instance in place
    update_cell_accumulator(accumulator, tau, r_port_tau, d_hat_tau, d_tau)

    # Return the computed portfolio return for potential logging or debugging
    return r_port_tau
````

````python
# Task 13 — Implement Per-Cell Directional Accuracy and Sharpe Ratio Computation

# ==============================================================================
# Task 13: Implement Per-Cell Directional Accuracy and Sharpe Ratio Computation
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Data Structure: Container for computed cell metrics
# -------------------------------------------------------------------------------------------------------------------------------
@dataclass
class CellMetrics:
    """
    A structured container holding the computed scalar metrics for a single
    experimental cell (model, condition, window, episode).

    Attributes
    ----------
    accuracy : float
        The directional accuracy (fraction of exact matches), Acc in [0, 1].
    sharpe_ratio : float
        The annualized Sharpe ratio of the implied long/cash portfolio.
    """
    accuracy: float
    sharpe_ratio: float

# -------------------------------------------------------------------------------------------------------------------------------
# Task 13, Step 1: Compute per-cell directional accuracy
# -------------------------------------------------------------------------------------------------------------------------------
def compute_directional_accuracy(
    d_hat_list: List[Optional[str]],
    d_list: List[Optional[str]]
) -> float:
    """
    Computes the directional accuracy as the fraction of months where the predicted
    direction exactly matches the realized direction.

    Equation: Acc = (1/72) * sum_{tau=1}^{72} 1[d_hat_tau == d_tau]

    Parameters
    ----------
    d_hat_list : List[Optional[str]]
        The length-72 list of predicted direction strings.
    d_list : List[Optional[str]]
        The length-72 list of realized direction strings.

    Returns
    -------
    float
        The computed directional accuracy scalar in the range [0.0, 1.0].

    Raises
    ------
    ValueError
        If the lists do not have exactly 72 elements or contain None values.
    """
    # Enforce strict length constraints as defined by the study (n=72 per episode)
    if len(d_hat_list) != 72 or len(d_list) != 72:
        raise ValueError(f"Accuracy computation requires exactly 72 elements. Got {len(d_hat_list)} and {len(d_list)}.")

    # Defense-in-depth: Ensure no None values exist, which would indicate skipped steps
    if None in d_hat_list or None in d_list:
        raise ValueError("Cannot compute accuracy: lists contain None values indicating incomplete accumulation.")

    # Convert Python lists to NumPy arrays with dtype=object to preserve exact string representations
    # This prevents any silent truncation or coercion that might occur with fixed-length string dtypes
    d_hat_array: np.ndarray = np.array(d_hat_list, dtype=object)
    d_array: np.ndarray = np.array(d_list, dtype=object)

    # Perform vectorized element-wise equality comparison
    # The '==' operator enforces exact, case-sensitive string matching
    matches: np.ndarray = (d_hat_array == d_array)

    # Compute the mean of the boolean array, which mathematically equals the fraction of True values
    # This perfectly implements the (1/n) * sum(...) equation
    accuracy: float = float(np.mean(matches))

    return accuracy

# -------------------------------------------------------------------------------------------------------------------------------
# Task 13, Step 2 & 3: Compute portfolio statistics and store the results
# -------------------------------------------------------------------------------------------------------------------------------
def compute_and_store_cell_sharpe(
    r_port_array: np.ndarray,
    accuracy: float,
    m: str,
    c: str,
    w: int,
    e: str,
    config: Dict[str, Any],
    metrics_dict: Dict[Tuple[str, str, int, str], CellMetrics]
) -> None:
    """
    Computes the annualized Sharpe ratio using the centralized helper (enforcing
    Bessel correction and degeneracy guards) and stores the (Acc, SR) pair.

    Equation: SR_hat = sqrt(12) * (r_bar^port / sigma_hat^port)

    Parameters
    ----------
    r_port_array : np.ndarray
        The length-72 float64 NumPy array of portfolio returns.
    accuracy : float
        The computed directional accuracy from Step 1.
    m, c, w, e : str/int
        The 4-tuple identifier coordinates (model, condition, window, episode).
    config : Dict[str, Any]
        The master study configuration dictionary.
    metrics_dict : Dict[Tuple[str, str, int, str], CellMetrics]
        The global dictionary storing all computed cell metrics.

    Raises
    ------
    ValueError
        If the return array does not have exactly 72 elements or contains NaNs.
    KeyError
        If the annualization factor is missing from the configuration.
    """
    # Enforce strict length constraints
    if len(r_port_array) != 72:
        raise ValueError(f"Sharpe computation requires exactly 72 elements. Got {len(r_port_array)}.")

    # Defense-in-depth: Ensure no NaN values exist
    if np.isnan(r_port_array).any():
        raise ValueError("Cannot compute Sharpe ratio: return array contains NaN values.")

    # Retrieve the shared annualization factor (sqrt(12)) from the configuration
    # This guarantees bit-for-bit consistency with the buy-and-hold calibration (Task 5)
    try:
        ann_factor: float = config["sharpe_params"]["sqrt_annualization_factor"]
    except KeyError as err:
        raise KeyError(f"Missing required configuration key for annualization factor: {err}")

    # Invoke the centralized helper function defined in Task 5
    # This helper strictly enforces ddof=1 (n-1=71 denominator) and handles zero-variance degeneracy
    mean_return, std_dev, sharpe_ratio = compute_annualized_sharpe(r_port_array, ann_factor)

    # Pack the computed scalars into the strongly-typed dataclass container
    cell_metrics = CellMetrics(
        accuracy=accuracy,
        sharpe_ratio=sharpe_ratio
    )

    # Construct the 4-tuple cell key
    cell_key: Tuple[str, str, int, str] = (m, c, w, e)

    # Store the metrics in the global dictionary
    metrics_dict[cell_key] = cell_metrics

# -------------------------------------------------------------------------------------------------------------------------------
# Task 13, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def execute_per_cell_metric_computation(
    accumulator: Any, # Type is CellAccumulator from Task 12
    m: str,
    c: str,
    w: int,
    e: str,
    config: Dict[str, Any],
    metrics_dict: Dict[Tuple[str, str, int, str], CellMetrics]
) -> None:
    """
    Orchestrates the computation of directional accuracy and the annualized Sharpe
    ratio for a single completed experimental cell.

    This function sequentially extracts the accumulated arrays, computes the accuracy,
    computes the Sharpe ratio, and persists the results in the global metrics dictionary.

    Parameters
    ----------
    accumulator : CellAccumulator
        The fully populated accumulator instance for the current cell.
    m, c, w, e : str/int
        The 4-tuple identifier coordinates (model, condition, window, episode).
    config : Dict[str, Any]
        The master study configuration dictionary.
    metrics_dict : Dict[Tuple[str, str, int, str], CellMetrics]
        The global dictionary storing all computed cell metrics.
    """
    # Extract the accumulated lists and arrays from the container
    d_hat_list: List[Optional[str]] = accumulator.d_hat_list
    d_list: List[Optional[str]] = accumulator.d_list
    r_port_array: np.ndarray = accumulator.r_port_array

    # Step 1: Compute per-cell directional accuracy
    # This evaluates the exact string matches between predictions and realizations
    accuracy: float = compute_directional_accuracy(d_hat_list, d_list)

    # Step 2 & 3: Compute portfolio statistics and store the results
    # This calculates the Bessel-corrected Sharpe ratio and updates the global dictionary
    compute_and_store_cell_sharpe(
        r_port_array=r_port_array,
        accuracy=accuracy,
        m=m,
        c=c,
        w=w,
        e=e,
        config=config,
        metrics_dict=metrics_dict
    )
````

````python
# Task 14 — Implement Combined Episode Metrics Computation

# ==============================================================================
# Task 14: Implement Combined Episode Metrics Computation
# ==============================================================================

# Define the fixed order for episode concatenation to ensure reproducibility
EPISODE_ORDER: List[str] = ["dotcom", "gfc"]

# -------------------------------------------------------------------------------------------------------------------------------
# Task 14, Step 1: Build the 144-element combined portfolio-return series
# -------------------------------------------------------------------------------------------------------------------------------
def build_combined_return_series(
    m: str,
    c: str,
    w: int,
    accumulators_dict: Dict[Tuple[str, str, int, str], Any] # Type is CellAccumulator
) -> np.ndarray:
    """
    Retrieves the per-episode portfolio return arrays for a specific (m, c, W) triple
    and concatenates them in a fixed, deterministic order to form the 144-element
    combined sequence.

    Parameters
    ----------
    m : str
        The model identifier.
    c : str
        The experimental condition.
    w : int
        The context window length.
    accumulators_dict : Dict[Tuple[str, str, int, str], CellAccumulator]
        The global dictionary storing all completed cell accumulators.

    Returns
    -------
    np.ndarray
        A 1D float64 NumPy array of length 144 containing the concatenated returns.

    Raises
    ------
    KeyError
        If the accumulator for either the dot-com or GFC episode is missing.
    ValueError
        If the concatenated array does not have exactly 144 elements.
    """
    arrays_to_concat: List[np.ndarray] = []

    # Iterate through the fixed episode order to guarantee deterministic concatenation
    for ep_name in EPISODE_ORDER:
        # Construct the 4-tuple key for the specific episode
        cell_key: Tuple[str, str, int, str] = (m, c, w, ep_name)

        # Retrieve the accumulator; raise KeyError if it hasn't been populated
        if cell_key not in accumulators_dict:
            raise KeyError(f"Missing accumulator for cell {cell_key}. Cannot compute combined metrics.")

        accumulator = accumulators_dict[cell_key]

        # Extract the 72-element float64 return array
        r_port_array: np.ndarray = accumulator.r_port_array
        arrays_to_concat.append(r_port_array)

    # Concatenate the arrays end-to-end
    # The resulting array will have length 72 + 72 = 144
    combined_returns: np.ndarray = np.concatenate(arrays_to_concat)

    # Enforce strict length validation
    if len(combined_returns) != 144:
        raise ValueError(f"Combined return series must have exactly 144 elements. Got {len(combined_returns)}.")

    return combined_returns

# -------------------------------------------------------------------------------------------------------------------------------
# Task 14, Step 2: Compute the combined-episode Sharpe ratio
# -------------------------------------------------------------------------------------------------------------------------------
def compute_combined_sharpe_ratio(
    combined_returns: np.ndarray,
    config: Dict[str, Any]
) -> float:
    """
    Computes the combined-episode Sharpe ratio over the full 144-element series.

    This strictly enforces the prohibition against averaging per-episode Sharpe ratios
    by applying the Bessel correction (n-1=143) to the concatenated array.

    Parameters
    ----------
    combined_returns : np.ndarray
        The 144-element float64 NumPy array of concatenated portfolio returns.
    config : Dict[str, Any]
        The master study configuration dictionary.

    Returns
    -------
    float
        The annualized combined Sharpe ratio.

    Raises
    ------
    KeyError
        If the annualization factor is missing from the configuration.
    """
    # Retrieve the shared annualization factor (sqrt(12)) from the configuration
    try:
        ann_factor: float = config["sharpe_params"]["sqrt_annualization_factor"]
    except KeyError as err:
        raise KeyError(f"Missing required configuration key for annualization factor: {err}")

    # Invoke the centralized helper function defined in Task 5
    # Because the input array has length 144, the helper's internal np.std(..., ddof=1)
    # automatically applies the correct n-1 = 143 Bessel correction.
    mean_return, std_dev, combined_sharpe = compute_annualized_sharpe(combined_returns, ann_factor)

    return combined_sharpe

# -------------------------------------------------------------------------------------------------------------------------------
# Task 14, Step 3: Compute combined directional accuracy and store metrics
# -------------------------------------------------------------------------------------------------------------------------------
def compute_and_store_combined_metrics(
    m: str,
    c: str,
    w: int,
    combined_sharpe: float,
    accumulators_dict: Dict[Tuple[str, str, int, str], Any], # Type is CellAccumulator
    metrics_dict: Dict[Tuple[str, str, int, str], Any] # Type is CellMetrics
) -> None:
    """
    Computes the combined directional accuracy over the concatenated 144 label pairs
    and stores the (Acc, SR) pair in the global metrics dictionary.

    Parameters
    ----------
    m : str
        The model identifier.
    c : str
        The experimental condition.
    w : int
        The context window length.
    combined_sharpe : float
        The computed combined Sharpe ratio from Step 2.
    accumulators_dict : Dict[Tuple[str, str, int, str], CellAccumulator]
        The global dictionary storing all completed cell accumulators.
    metrics_dict : Dict[Tuple[str, str, int, str], CellMetrics]
        The global dictionary storing all computed cell metrics.

    Raises
    ------
    ValueError
        If the concatenated label lists do not have exactly 144 elements.
    """
    combined_d_hat: List[Optional[str]] = []
    combined_d: List[Optional[str]] = []

    # Iterate through the fixed episode order to concatenate the label lists
    for ep_name in EPISODE_ORDER:
        cell_key: Tuple[str, str, int, str] = (m, c, w, ep_name)
        accumulator = accumulators_dict[cell_key]

        # Extend the combined lists with the 72 elements from the current episode
        combined_d_hat.extend(accumulator.d_hat_list)
        combined_d.extend(accumulator.d_list)

    # Enforce strict length validation
    if len(combined_d_hat) != 144 or len(combined_d) != 144:
        raise ValueError("Combined label lists must have exactly 144 elements.")

    # Convert the concatenated lists to NumPy object arrays to preserve exact strings
    d_hat_array: np.ndarray = np.array(combined_d_hat, dtype=object)
    d_array: np.ndarray = np.array(combined_d, dtype=object)

    # Perform vectorized element-wise equality comparison and compute the mean
    # This perfectly implements the (1/144) * sum(...) equation
    matches: np.ndarray = (d_hat_array == d_array)
    combined_accuracy: float = float(np.mean(matches))

    # Pack the computed scalars into the strongly-typed dataclass container
    # (Assuming CellMetrics is available in the environment from Task 13)
    cell_metrics = CellMetrics(
        accuracy=combined_accuracy,
        sharpe_ratio=combined_sharpe
    )

    # Construct the 4-tuple key using the explicit literal "combined"
    combined_key: Tuple[str, str, int, str] = (m, c, w, "combined")

    # Store the metrics in the global dictionary
    metrics_dict[combined_key] = cell_metrics

# -------------------------------------------------------------------------------------------------------------------------------
# Task 14, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def execute_combined_metric_computation(
    m: str,
    c: str,
    w: int,
    config: Dict[str, Any],
    accumulators_dict: Dict[Tuple[str, str, int, str], Any],
    metrics_dict: Dict[Tuple[str, str, int, str], Any]
) -> None:
    """
    Orchestrates the computation of the combined-episode metrics for a specific
    (model, condition, window) triple.

    This function sequentially builds the concatenated return series, computes the
    combined Sharpe ratio using the correct Bessel correction, computes the combined
    accuracy, and persists the results.

    Parameters
    ----------
    m : str
        The model identifier.
    c : str
        The experimental condition.
    w : int
        The context window length.
    config : Dict[str, Any]
        The master study configuration dictionary.
    accumulators_dict : Dict[Tuple[str, str, int, str], CellAccumulator]
        The global dictionary storing all completed cell accumulators.
    metrics_dict : Dict[Tuple[str, str, int, str], CellMetrics]
        The global dictionary storing all computed cell metrics.
    """
    # Step 1: Build the 144-element combined portfolio-return series
    # This guarantees the fixed dot-com -> GFC concatenation order
    combined_returns: np.ndarray = build_combined_return_series(
        m, c, w, accumulators_dict
    )

    # Step 2: Compute the combined-episode Sharpe ratio
    # This enforces the n-1=143 denominator and prohibits averaging
    combined_sharpe: float = compute_combined_sharpe_ratio(
        combined_returns, config
    )

    # Step 3: Compute combined directional accuracy and store metrics
    # This evaluates exact string matches over the 144 concatenated labels
    compute_and_store_combined_metrics(
        m, c, w, combined_sharpe, accumulators_dict, metrics_dict
    )
````

````python
# Task 15 — Implement Results Tables 1 and 3 Assembly

# ==============================================================================
# Task 15: Implement Results Tables 1 and 3 Assembly
# ==============================================================================

# Define the strict, verbatim ordering for rows and columns
ORDERED_EPISODES: List[str] = ["dotcom", "gfc", "combined"]
ORDERED_CONDITIONS: List[str] = ["A", "B", "C", "D"]
# W=60 must appear before W=36 per the verbatim table layout
ORDERED_WINDOWS: List[int] = [60, 36]
ORDERED_MODELS: List[str] = ["gpt5", "claude_sonnet_4_6", "gemini_3_pro"]

# Mapping from internal model keys to verbatim display names
MODEL_DISPLAY_NAMES: Dict[str, str] = {
    "gpt5": "GPT-5",
    "claude_sonnet_4_6": "Claude 4.6",
    "gemini_3_pro": "Gemini 3 Pro"
}

# Mapping from internal episode keys to verbatim display names
EPISODE_DISPLAY_NAMES: Dict[str, str] = {
    "dotcom": "Dot-com\n1996--2001",
    "gfc": "GFC\n2004--2009",
    "combined": "Overall\ncombined"
}

# -------------------------------------------------------------------------------------------------------------------------------
# Helper Function: Initialize the empty MultiIndex DataFrame
# -------------------------------------------------------------------------------------------------------------------------------
def initialize_results_dataframe() -> pd.DataFrame:
    """
    Constructs an empty 12x6 pandas DataFrame with the exact verbatim MultiIndex
    row and column structures required for Tables 1 and 3.

    Returns
    -------
    pd.DataFrame
        An empty DataFrame initialized with np.nan, featuring hierarchical indices.
    """
    # Construct the row MultiIndex: (Episode, Condition)
    row_tuples: List[Tuple[str, str]] = [
        (EPISODE_DISPLAY_NAMES[ep], cond)
        for ep in ORDERED_EPISODES
        for cond in ORDERED_CONDITIONS
    ]
    row_index = pd.MultiIndex.from_tuples(row_tuples, names=["Episode", "Condition"])

    # Construct the column MultiIndex: (Window, Model)
    # Note: W=60 appears before W=36
    col_tuples: List[Tuple[str, str]] = [
        (f"W = {w}", MODEL_DISPLAY_NAMES[mod])
        for w in ORDERED_WINDOWS
        for mod in ORDERED_MODELS
    ]
    col_index = pd.MultiIndex.from_tuples(col_tuples, names=["Window", "Model"])

    # Instantiate the DataFrame filled with NaNs
    df = pd.DataFrame(np.nan, index=row_index, columns=col_index)
    return df

# -------------------------------------------------------------------------------------------------------------------------------
# Task 15, Step 1: Assemble Table 1 (Directional Accuracy)
# -------------------------------------------------------------------------------------------------------------------------------
def assemble_accuracy_table(
    metrics_dict: Dict[Tuple[str, str, int, str], Any] # Type is CellMetrics
) -> pd.DataFrame:
    """
    Populates the 12x6 MultiIndex DataFrame with the directional accuracy scalars
    from the global metrics dictionary, rounding to 3 decimal places.

    Parameters
    ----------
    metrics_dict : Dict[Tuple[str, str, int, str], CellMetrics]
        The global dictionary storing all computed cell metrics.

    Returns
    -------
    pd.DataFrame
        The populated and rounded Table 1 DataFrame.

    Raises
    ------
    ValueError
        If any cell in the DataFrame remains NaN after population.
    """
    # Initialize the empty structured DataFrame
    df_acc: pd.DataFrame = initialize_results_dataframe()

    # Iterate through the metrics dictionary and populate the DataFrame
    for (m, c, w, e), metrics in metrics_dict.items():
        # Map internal keys to display names for coordinate-based assignment
        ep_display: str = EPISODE_DISPLAY_NAMES[e]
        mod_display: str = MODEL_DISPLAY_NAMES[m]
        win_display: str = f"W = {w}"

        # Assign the accuracy scalar to the exact (row, column) coordinate
        df_acc.loc[(ep_display, c), (win_display, mod_display)] = metrics.accuracy

    # Enforce completeness: ensure no NaNs remain
    if df_acc.isna().any().any():
        raise ValueError("Table 1 assembly failed: Missing accuracy data for one or more cells.")

    # Round the entire DataFrame to exactly 3 decimal places per the verbatim specification
    df_acc = df_acc.round(3)

    return df_acc

# -------------------------------------------------------------------------------------------------------------------------------
# Task 15, Step 2: Assemble Table 3 (Sharpe Ratios)
# -------------------------------------------------------------------------------------------------------------------------------
def assemble_sharpe_table(
    metrics_dict: Dict[Tuple[str, str, int, str], Any], # Type is CellMetrics
    config: Dict[str, Any]
) -> pd.DataFrame:
    """
    Populates the 12x6 MultiIndex DataFrame with the Sharpe ratio scalars,
    rounds to 3 decimal places, and attaches the buy-and-hold benchmarks as metadata.

    Parameters
    ----------
    metrics_dict : Dict[Tuple[str, str, int, str], CellMetrics]
        The global dictionary storing all computed cell metrics.
    config : Dict[str, Any]
        The master study configuration dictionary.

    Returns
    -------
    pd.DataFrame
        The populated, rounded, and annotated Table 3 DataFrame.

    Raises
    ------
    ValueError
        If any cell in the DataFrame remains NaN after population.
    """
    # Initialize the empty structured DataFrame
    df_sharpe: pd.DataFrame = initialize_results_dataframe()

    # Iterate through the metrics dictionary and populate the DataFrame
    for (m, c, w, e), metrics in metrics_dict.items():
        # Map internal keys to display names
        ep_display: str = EPISODE_DISPLAY_NAMES[e]
        mod_display: str = MODEL_DISPLAY_NAMES[m]
        win_display: str = f"W = {w}"

        # Assign the Sharpe ratio scalar to the exact coordinate
        df_sharpe.loc[(ep_display, c), (win_display, mod_display)] = metrics.sharpe_ratio

    # Enforce completeness
    if df_sharpe.isna().any().any():
        raise ValueError("Table 3 assembly failed: Missing Sharpe data for one or more cells.")

    # Round to 3 decimal places
    df_sharpe = df_sharpe.round(3)

    # Attach the buy-and-hold benchmarks as non-destructive metadata
    bah_benchmarks: Dict[str, float] = {
        "dotcom": config["sharpe_params"]["bah_sharpe_dotcom"],
        "gfc": config["sharpe_params"]["bah_sharpe_gfc"],
        "combined": config["sharpe_params"]["bah_sharpe_combined"]
    }
    df_sharpe.attrs["bah_benchmarks"] = bah_benchmarks

    return df_sharpe

# -------------------------------------------------------------------------------------------------------------------------------
# Task 15, Step 3: Apply the verbatim bolding rule and verify structural identity
# -------------------------------------------------------------------------------------------------------------------------------
def apply_bolding_rule(df: pd.DataFrame) -> pd.io.formats.style.Styler:
    """
    Applies the verbatim bolding rule: "best per episode x model x context window".
    This bolds the maximum value among the 4 conditions (A, B, C, D) within each
    specific column and episode block.

    Parameters
    ----------
    df : pd.DataFrame
        The populated 12x6 DataFrame (Table 1 or Table 3).

    Returns
    -------
    pd.io.formats.style.Styler
        The styled DataFrame object with CSS bolding applied.
    """
    def highlight_max_per_block(data: pd.DataFrame) -> pd.DataFrame:
        """
        Internal styling function that returns a DataFrame of CSS strings
        with the same shape as the input data.
        """
        # Initialize an empty DataFrame of strings to hold the CSS styles
        styles = pd.DataFrame('', index=data.index, columns=data.columns)

        # Iterate over each column (Window, Model)
        for col in data.columns:
            # Iterate over each episode block
            for ep_key in EPISODE_DISPLAY_NAMES.values():
                # Extract the 4-element Series for this specific block and column
                # .loc[(episode, slice(None))] selects all conditions for the episode
                block_series = data.loc[(ep_key, slice(None)), col]

                # Find the maximum value in this 4-element block
                max_val = block_series.max()

                # Identify all indices where the value equals the maximum (handles ties)
                is_max = block_series == max_val

                # Apply the bold CSS style to the identified indices
                for cond in is_max[is_max].index.get_level_values('Condition'):
                    styles.loc[(ep_key, cond), col] = 'font-weight: bold'

        return styles

    # Apply the styling function to the DataFrame
    styled_df = df.style.apply(highlight_max_per_block, axis=None)
    return styled_df

def verify_structural_identity(df: pd.DataFrame, table_name: str) -> None:
    """
    Verifies that the assembled DataFrame has exactly the 12x6 shape required.
    (Detailed numerical verification against reported_results is deferred to Task 17).

    Parameters
    ----------
    df : pd.DataFrame
        The assembled DataFrame.
    table_name : str
        The name of the table (e.g., "Table 1") for error logging.

    Raises
    ------
    ValueError
        If the shape is not exactly (12, 6).
    """
    expected_shape: Tuple[int, int] = (12, 6)
    if df.shape != expected_shape:
        raise ValueError(f"Structural verification failed for {table_name}: Expected shape {expected_shape}, got {df.shape}.")
    logger.info(f"Structural identity verified for {table_name}.")

# -------------------------------------------------------------------------------------------------------------------------------
# Task 15, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def assemble_results_tables(
    metrics_dict: Dict[Tuple[str, str, int, str], Any],
    config: Dict[str, Any]
) -> Tuple[pd.io.formats.style.Styler, pd.io.formats.style.Styler]:
    """
    Orchestrates the assembly, formatting, and styling of Results Tables 1 and 3.

    This function sequentially builds the MultiIndex DataFrames, populates them
    with the computed metrics, applies the block-wise bolding rule, and verifies
    their structural integrity.

    Parameters
    ----------
    metrics_dict : Dict[Tuple[str, str, int, str], CellMetrics]
        The global dictionary storing all computed cell metrics.
    config : Dict[str, Any]
        The master study configuration dictionary.

    Returns
    -------
    Tuple[pd.io.formats.style.Styler, pd.io.formats.style.Styler]
        A tuple containing the styled Table 1 (Accuracy) and Table 3 (Sharpe).
    """
    logger.info("Assembling Results Tables 1 and 3...")

    # Step 1: Assemble Table 1 (Directional Accuracy)
    df_table_1: pd.DataFrame = assemble_accuracy_table(metrics_dict)
    verify_structural_identity(df_table_1, "Table 1 (Accuracy)")

    # Step 2: Assemble Table 3 (Sharpe Ratios)
    df_table_3: pd.DataFrame = assemble_sharpe_table(metrics_dict, config)
    verify_structural_identity(df_table_3, "Table 3 (Sharpe)")

    # Step 3: Apply the verbatim bolding rule to both tables
    styled_table_1: pd.io.formats.style.Styler = apply_bolding_rule(df_table_1)
    styled_table_3: pd.io.formats.style.Styler = apply_bolding_rule(df_table_3)

    logger.info("Results Tables assembled and styled successfully.")

    return styled_table_1, styled_table_3
````

````python
# Task 16 — Implement the End-to-End Research Pipeline Orchestrator

# ==============================================================================
# Task 16: Implement the End-to-End Research Pipeline Orchestrator
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 16, Step 1: Define the Orchestrator Interface and Execute the Calibration Gate
# -------------------------------------------------------------------------------------------------------------------------------
def execute_calibration_gate(
    # The validated container holding the 0-based price and EPS arrays.
    merged_container: Any,
    # The master study configuration dictionary containing targets.
    config: Dict[str, Any]
) -> None:
    """
    Executes the buy-and-hold calibration check as an absolute gate before any
    API budget is expended.

    Parameters
    ----------
    merged_container : MergedDataContainer
        The validated container holding the raw price array.
    config : Dict[str, Any]
        The master study configuration dictionary.

    Returns
    -------
    None

    Raises
    ------
    CalibrationError
        If the computed Sharpe ratios deviate from the targets by > 0.02.
    TypeError
        If the input arguments are not of the expected types.
    """
    # Verify that the config argument is a dictionary.
    if not isinstance(config, dict):
        # Raise a TypeError if the config type is invalid.
        raise TypeError(f"Expected dict for config, got {type(config).__name__}")

    # Log the initiation of the upfront calibration check.
    logger.info("Executing upfront buy-and-hold calibration gate...")
    # Invoke the Task 5 orchestrator to compute and validate the benchmarks.

    # We allow any CalibrationError to propagate up to the top-level orchestrator.
    compute_and_validate_calibration_benchmarks(merged_container, config)

    # Log the successful validation of the calibration targets.
    logger.info("Calibration gate passed successfully. Proceeding to evaluation loop.")


def validate_client_dictionary(
    # The dictionary of configured LLM client configurations.
    clients: Dict[str, Any]
) -> None:
    """
    Asserts that the provided clients dictionary contains exactly the required
    model identifiers, preventing KeyErrors deep within the evaluation loop.

    Parameters
    ----------
    clients : Dict[str, LLMClientConfig]
        The dictionary of configured LLM clients.

    Returns
    -------
    None

    Raises
    ------
    KeyError
        If any required client identifier is missing.
    TypeError
        If the input clients argument is not a dictionary.
    """
    # Verify that the clients argument is a dictionary.
    if not isinstance(clients, dict):
        # Raise a TypeError if the input is not a dictionary.
        raise TypeError(f"Expected dict for clients, got {type(clients).__name__}")

    # Define the strict set of required client keys based on the study.
    required_keys: Set[str] = {"gpt5", "claude_sonnet_4_6", "gemini_3_pro"}

    # Extract the actual keys present in the clients dictionary.
    actual_keys: Set[str] = set(clients.keys())

    # Perform a set-theoretic check to verify all required keys are present.
    if not required_keys.issubset(actual_keys):
        # Calculate the exact set of missing keys.
        missing_keys: Set[str] = required_keys - actual_keys
        # Raise a specific KeyError listing the missing model identifiers.
        raise KeyError(f"Fatal configuration error: Missing required LLM clients: {missing_keys}")

# -------------------------------------------------------------------------------------------------------------------------------
# Task 16, Step 2: Execute the Nested Evaluation Loop with Exact State Reconstruction
# -------------------------------------------------------------------------------------------------------------------------------
def load_completed_responses(
    # The filesystem path to the JSONL log file.
    log_file_path: str
) -> Dict[Tuple[str, str, int, str, int], Optional[str]]:
    """
    Reads the persistent JSONL log file to build a mapping of completed 5-tuples
    to their raw LLM responses. This enables true crash-recoverability by allowing
    the pipeline to reconstruct the exact mathematical state of the accumulators
    without duplicating API calls.

    Parameters
    ----------
    log_file_path : str
        The filesystem path to the persistent JSONL log file.

    Returns
    -------
    Dict[Tuple[str, str, int, str, int], Optional[str]]
        A dictionary mapping the 5-tuple (m, c, W, e, tau) to the raw response string.

    Raises
    ------
    TypeError
        If the input log_file_path is not a string.
    """
    # Verify that the log_file_path is a string.
    if not isinstance(log_file_path, str):
        # Raise a TypeError if the file path is not a string.
        raise TypeError(f"Expected str for log_file_path, got {type(log_file_path).__name__}")

    # Initialize an empty dictionary to store the completed responses mapping.
    completed_responses: Dict[Tuple[str, str, int, str, int], Optional[str]] = {}
    # Check if the log file exists on the filesystem.
    if not os.path.exists(log_file_path):
        # Return the empty dictionary if the file does not exist yet.
        return completed_responses

    # Open the log file in read-only mode with explicit UTF-8 encoding.
    with open(log_file_path, "r", encoding="utf-8") as f:
        # Iterate over each line in the JSONL log file.
        for line in f:
            # Skip any empty or whitespace-only lines.
            if not line.strip():
                # Continue to the next line.
                continue

            # Parse the line inside a try-except block to handle file corruption.
            try:
                # Deserialize the JSON line into a Python dictionary.
                record: Dict[str, Any] = json.loads(line)
                # Construct the 5-tuple coordinate key.
                tup: Tuple[str, str, int, str, int] = (
                    # The model identifier.
                    str(record["model"]),
                    # The prompting condition.
                    str(record["condition"]),
                    # The context window length.
                    int(record["window"]),
                    # The episode identifier.
                    str(record["episode"]),
                    # The 1-based evaluation step.
                    int(record["tau"])
                )
                # Extract the recorded API outcome.
                outcome: str = str(record.get("outcome", ""))
                # Verify if the step represents a fully completed execution.
                if outcome in ["SUCCESS", "NULL"]:
                    # Store the raw response string in the dictionary under the 5-tuple key.
                    completed_responses[tup] = record.get("raw_response")
            # Catch JSON decoding exceptions caused by mid-write crashes.
            except json.JSONDecodeError:
                # Log a debug warning and silently ignore the corrupted line.
                logger.debug("Corrupted line detected in JSONL log; skipping.")
                # Continue to the next line.
                continue

    # Return the completed responses dictionary mapping.
    return completed_responses

def execute_nested_evaluation_loop(
    # The validated container holding the raw price and EPS arrays.
    merged_container: Any,
    # The master study configuration dictionary.
    config: Dict[str, Any],
    # The dictionary of configured LLM clients.
    clients: Dict[str, Any],
    # The filesystem path to the persistent JSONL log file.
    log_file_path: str
) -> Tuple[Dict[Tuple[str, str, int, str], Any], int]:
    """
    Executes the 3,456-call nested evaluation loop, orchestrating normalization,
    prompt assembly, API calls, parsing, and accumulation for each step.

    Parameters
    ----------
    merged_container : MergedDataContainer
        The validated container holding the raw price and EPS arrays.
    config : Dict[str, Any]
        The master study configuration dictionary.
    clients : Dict[str, LLMClientConfig]
        The dictionary of configured LLM clients.
    log_file_path : str
        The filesystem path to the persistent JSONL log file.

    Returns
    -------
    Tuple[Dict[Tuple[str, str, int, str], CellAccumulator], int]
        A tuple containing the dictionary of fully populated accumulators for all 48 cells,
        and the count of parsing fallbacks triggered during live API calls in this session.

    Raises
    ------
    ValueError
        If any data validation or state finalization check fails during execution.
    IndexError
        If an out-of-bounds array access is attempted.
    """
    # Pre-assemble and validate all condition system prompts to cache them in memory.
    prompts_cache: Dict[str, str] = pre_assemble_and_validate_prompts(config)
    # Load completed responses from the JSONL log file for state reconstruction.
    completed_responses: Dict[Tuple[str, str, int, str, int], Optional[str]] = load_completed_responses(log_file_path)
    # Initialize the global dictionary to store completed cell accumulators.
    accumulators_dict: Dict[Tuple[str, str, int, str], Any] = {}
    # Initialize the session fallback counter to track parsing failures.
    session_fallbacks: int = 0

    # Define the ordered list of model identifiers to iterate.
    models: List[str] = ["gpt5", "claude_sonnet_4_6", "gemini_3_pro"]
    # Define the ordered list of prompting conditions to iterate.
    conditions: List[str] = ["A", "B", "C", "D"]
    # Retrieve the context window lengths list from the configuration.
    windows: List[int] = config["experiment_config"]["context_windows"]
    # Define the ordered list of historical episodes to iterate.
    episodes: List[str] = ["dotcom", "gfc"]

    # Iterate over each model identifier in the outer loop.
    for m in models:
        # Iterate over each prompting condition.
        for c in conditions:
            # Iterate over each context window length.
            for w in windows:
                # Iterate over each historical episode.
                for e in episodes:
                    # Initialize a fresh accumulator instance for the current cell.
                    accumulator: Any = initialize_cell_accumulator()
                    # Iterate over each evaluation step tau from 1 to 72.
                    for tau in range(1, 73):
                        # Construct the unique 5-tuple identifier for this step.
                        step_tuple: Tuple[str, str, int, str, int] = (m, c, w, e, tau)

                        # (a) NORMALIZE: Produce rounded and full normalized arrays.
                        normalized_container: Any = execute_normalization_step(tau, e, w, merged_container)

                        # Check if this step's raw response already exists in the log.
                        if step_tuple in completed_responses:
                            # Reconstruct the response from the persistent log to bypass the API.
                            raw_response: Optional[str] = completed_responses[step_tuple]
                        # If not completed, execute the live API call.
                        else:
                            # (b) Serialize User Message: Render the table string.
                            user_message: str = generate_validated_user_message(c, normalized_container, config)

                            # (c) API Call: Submit prompt pair and log the outcome.
                            raw_response, _ = execute_single_api_step(
                                m, c, w, e, tau, merged_container, config, prompts_cache, clients, log_file_path
                            )

                        # (d) Parse JSON: Extract and validate the direction forecast.
                        d_hat_tau, fallback_flag = parse_and_validate_llm_response(raw_response, m, c, w, e, tau)
                        # Check if a fallback was triggered on a live API call in this session.
                        if fallback_flag and step_tuple not in completed_responses:
                            # Increment the local session fallback counter.
                            session_fallbacks += 1

                        # (e) Realized Return: Compute simple return and realized label.
                        r_tau, d_tau = compute_realized_return_and_label(tau, e, w, merged_container)

                        # (f) Accumulator: Map position, compute return, and write to array.
                        # This mutates the accumulator instance in place.
                        execute_portfolio_mapping_and_accumulation(
                            tau, d_hat_tau, r_tau, d_tau, config, accumulator
                        )

                    # Finalize the accumulator and store it in the dictionary.
                    # This raises a ValueError if any of the 72 steps are missing.
                    finalize_cell_accumulation(accumulator, m, c, w, e, accumulators_dict)

                    # Log the successful completion of the current cell.
                    logger.info(f"Completed evaluation and accumulation for cell: ({m}, {c}, {w}, {e})")

    # Return the completed accumulators dictionary and the session fallbacks count.
    return accumulators_dict, session_fallbacks

# -------------------------------------------------------------------------------------------------------------------------------
# Task 16, Step 3: Finalize Metrics, Assemble Tables, and Emit Summary
# -------------------------------------------------------------------------------------------------------------------------------
def compute_all_metrics_and_tables(
    # The dictionary of fully completed cell accumulators.
    accumulators_dict: Dict[Tuple[str, str, int, str], Any],
    # The master study configuration dictionary.
    config: Dict[str, Any]
) -> Tuple[Any, Any]:
    """
    Computes per-cell and combined metrics, and assembles the final styled tables.

    Parameters
    ----------
    accumulators_dict : Dict[Tuple[str, str, int, str], CellAccumulator]
        The dictionary of completed cell accumulators.
    config : Dict[str, Any]
        The master study configuration dictionary.

    Returns
    -------
    Tuple[pd.io.formats.style.Styler, pd.io.formats.style.Styler]
        A tuple containing the styled Table 1 (Accuracy) and Table 3 (Sharpe).

    Raises
    ------
    ValueError
        If any metric calculation or table assembly step fails.
    KeyError
        If required configuration parameters are missing.
    """
    # Initialize an empty dictionary to hold the computed cell metrics.
    metrics_dict: Dict[Tuple[str, str, int, str], Any] = {}

    # 1. Compute per-cell metrics for all 48 individual episode cells.
    for (m, c, w, e), accumulator in accumulators_dict.items():
        # Invoke the Task 13 orchestrator to compute accuracy and Sharpe for each cell.
        # This will write the results directly to metrics_dict.
        execute_per_cell_metric_computation(accumulator, m, c, w, e, config, metrics_dict)

    # Define the model identifiers loop.
    models: List[str] = ["gpt5", "claude_sonnet_4_6", "gemini_3_pro"]
    # Define the prompting conditions loop.
    conditions: List[str] = ["A", "B", "C", "D"]
    # Retrieve the window lengths list from the configuration.
    windows: List[int] = config["experiment_config"]["context_windows"]

    # 2. Compute combined-episode metrics for each of the 24 (m, c, W) triples.
    for m in models:
        # Loop over conditions.
        for c in conditions:
            # Loop over windows.
            for w in windows:
                # Invoke the Task 14 orchestrator to compute combined metrics.
                # This concatenates arrays first and avoids the mathematical error of averaging.
                execute_combined_metric_computation(m, c, w, config, accumulators_dict, metrics_dict)

    # 3. Assemble and style Results Tables 1 and 3.
    # This pivots the metrics and applies the block-wise bolding rule.
    styled_table_1, styled_table_3 = assemble_results_tables(metrics_dict, config)

    # Return the two styled tables as the output tuple.
    return styled_table_1, styled_table_3

def generate_run_summary(
    # The filesystem path to the JSONL log file.
    log_file_path: str,
    # The wall-clock start time of the pipeline execution.
    start_time: float,
    # The count of parsing fallbacks triggered during live API calls in this session.
    session_fallbacks: int
) -> Dict[str, float]:
    """
    Parses the persistent JSONL log file to generate accurate execution summary statistics.

    Parameters
    ----------
    log_file_path : str
        The filesystem path to the JSONL log file.
    start_time : float
        The wall-clock start time of the pipeline.
    session_fallbacks : int
        The count of parsing fallbacks triggered during live API calls in this session.

    Returns
    -------
    Dict[str, float]
        A dictionary containing the summary statistics.

    Raises
    ------
    TypeError
        If the input types are invalid.
    """
    # Verify that the log_file_path is a string.
    if not isinstance(log_file_path, str):
        # Raise a TypeError if the log path is invalid.
        raise TypeError(f"Expected str for log_file_path, got {type(log_file_path).__name__}")
    # Verify that the start_time is a float.
    if not isinstance(start_time, (float, int)):
        # Raise a TypeError if the start time is invalid.
        raise TypeError(f"Expected float for start_time, got {type(start_time).__name__}")
    # Verify that the session_fallbacks is an integer.
    if not isinstance(session_fallbacks, int):
        # Raise a TypeError if the fallback count is invalid.
        raise TypeError(f"Expected int for session_fallbacks, got {type(session_fallbacks).__name__}")

    # Initialize the count of successful API calls.
    total_api_calls: int = 0
    # Initialize the count of hard API failures (NULL outcomes in log).
    total_null_logs: int = 0
    # Initialize the count of total retries.
    total_retries: int = 0

    # Check if the log file exists before attempting to read.
    if os.path.exists(log_file_path):
        # Open the log file in read-only mode with UTF-8 encoding.
        with open(log_file_path, "r", encoding="utf-8") as f:
            # Iterate over each line in the log.
            for line in f:
                # Skip whitespace-only lines.
                if not line.strip():
                    # Continue to the next line.
                    continue
                # Parse the JSON line inside a try-except block.
                try:
                    # Deserialize the JSON line.
                    record: Dict[str, Any] = json.loads(line)

                    # Extract the outcome string.
                    outcome: str = str(record.get("outcome", ""))

                    # Extract the attempt count.
                    attempt_count: int = int(record.get("attempt_count", 1))

                    # Increment the appropriate API status counter.
                    if outcome == "SUCCESS":
                        # Add to the successful call count.
                        total_api_calls += 1
                    elif outcome == "NULL":
                        # Add to the hard failure count.
                        total_null_logs += 1
                    # Retries represent attempts beyond the first one.
                    total_retries += max(0, attempt_count - 1)

                # Catch JSON decoding exceptions.
                except json.JSONDecodeError:
                    # Silently skip the corrupted line.
                    continue

    # Record the wall-clock end time.
    end_time: float = time.time()

    # Compute the total elapsed wall-clock seconds.
    wall_clock_seconds: float = float(end_time - start_time)

    # Total fallbacks is the sum of hard API failures and session-level parsing failures.
    total_fallbacks: int = total_null_logs + session_fallbacks

    # Construct the final summary statistics dictionary.
    summary: Dict[str, float] = {
        "wall_clock_seconds": wall_clock_seconds,
        "total_api_calls_executed": float(total_api_calls),
        "total_fallback_responses": float(total_fallbacks),
        "total_retry_events": float(total_retries)
    }

    # Log the summary statistics at the INFO level.
    logger.info("=== Pipeline Execution Summary ===")

    # Loop over key-value pairs in the summary dictionary.
    for k, v in summary.items():
        # Log each statistic.
        logger.info(f"{k}: {v}")

    # Return the summary dictionary.
    return summary

# -------------------------------------------------------------------------------------------------------------------------------
# Task 16, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def run_end_to_end_pipeline(
    # The validated container holding the raw price and EPS arrays.
    merged_container: Any,
    # The master study configuration dictionary.
    config: Dict[str, Any],
    # The dictionary of configured LLM clients.
    clients: Dict[str, Any],
    # The filesystem path to the persistent JSONL log file.
    log_file_path: str
) -> Tuple[Any, Any, Dict[str, float]]:
    """
    Orchestrates the complete end-to-end research evaluation pipeline.

    This function sequentially validates clients, executes the upfront calibration
    gate, runs the 3,456-call evaluation loop (with exact state reconstruction),
    calculates all cell and combined metrics, assembles Results Tables 1 and 3,
    and generates the execution summary.

    Parameters
    ----------
    merged_container : MergedDataContainer
        The validated container holding the raw price and EPS arrays.
    config : Dict[str, Any]
        The master study configuration dictionary.
    clients : Dict[str, LLMClientConfig]
        The dictionary of configured LLM clients.
    log_file_path : str
        The filesystem path to the persistent JSONL log file.

    Returns
    -------
    Tuple[pd.io.formats.style.Styler, pd.io.formats.style.Styler, Dict[str, float]]
        A tuple containing styled Table 1, styled Table 3, and the summary statistics.

    Raises
    ------
    CalibrationError
        If the data version is inconsistent and buy-and-hold benchmarks fail.
    ValueError
        If any data validation or state finalization check fails during execution.
    """
    # Record the precise wall-clock start time of the pipeline execution.
    start_time: float = float(time.time())

    # Log the initiation of the end-to-end evaluation pipeline.
    logger.info("Entering End-to-End Research Pipeline (Task 16)...")

    # Step 1a: Validate the presence of all required LLM clients.
    validate_client_dictionary(clients)

    # Step 1b: Execute the non-bypassable upfront calibration gate.
    # Any CalibrationError will propagate up naturally to halt execution.
    execute_calibration_gate(merged_container, config)

    # Step 2: Execute the 3,456-call nested evaluation loop with state reconstruction.
    accumulators_dict, session_fallbacks = execute_nested_evaluation_loop(
        merged_container, config, clients, log_file_path
    )

    # Step 3a: Compute per-cell and combined metrics and assemble the tables.
    styled_table_1, styled_table_3 = compute_all_metrics_and_tables(
        accumulators_dict, config
    )

    # Step 3b: Parse the persistent log file to generate the run summary.
    summary_stats: Dict[str, float] = generate_run_summary(
        log_file_path, start_time, session_fallbacks
    )

    # Return the assembled styled tables and execution summary statistics.
    return styled_table_1, styled_table_3, summary_stats
````

````python
# Task 17 — Verify Reproduced Results Against the Paper's Reported Tables

# ==============================================================================
# Task 17: Verify Reproduced Results Against the Paper's Reported Tables
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Data Structures and Mappings
# -------------------------------------------------------------------------------------------------------------------------------

@dataclass
class DeviationRecord:
    """
    A structured container representing a single cell's deviation from the reported results.

    Attributes
    ----------
    episode : str
        The episode identifier (e.g., 'dotcom', 'gfc', 'combined').
    condition : str
        The experimental condition ('A', 'B', 'C', 'D').
    model : str
        The internal model identifier (e.g., 'gpt5').
    window : int
        The context window length (36 or 60).
    metric : str
        The metric being compared ('accuracy' or 'sharpe').
    reproduced_value : float
        The value computed by the current pipeline.
    reported_value : float
        The value reported in the original study.
    absolute_deviation : float
        The absolute difference between reproduced and reported values.
    classification : str
        The diagnostic conclusion (e.g., 'MODEL_VERSION_DRIFT').
    """
    episode: str
    condition: str
    model: str
    window: int
    metric: str
    reproduced_value: float
    reported_value: float
    absolute_deviation: float
    classification: str = ""

# Mappings to translate between MultiIndex display names and internal config keys
DISPLAY_TO_INTERNAL_MODEL: Dict[str, str] = {
    "GPT-5": "gpt5",
    "Claude 4.6": "claude_sonnet_4_6",
    "Gemini 3 Pro": "gemini_3_pro"
}

INTERNAL_TO_REPORTED_MODEL: Dict[str, str] = {
    "gpt5": "gpt5",
    "claude_sonnet_4_6": "claude",
    "gemini_3_pro": "gemini"
}

DISPLAY_TO_INTERNAL_EPISODE: Dict[str, str] = {
    "Dot-com\n1996--2001": "dotcom",
    "GFC\n2004--2009": "gfc",
    "Overall\ncombined": "combined"
}

# -------------------------------------------------------------------------------------------------------------------------------
# Task 17, Step 1: Compute per-cell absolute deviations for directional accuracy
# -------------------------------------------------------------------------------------------------------------------------------
def compute_accuracy_deviations(
    df_reproduced: pd.DataFrame,
    config: Dict[str, Any]
) -> Tuple[pd.DataFrame, List[DeviationRecord]]:
    """
    Computes the absolute deviation between reproduced and reported directional accuracy
    for all 72 cells, flagging deterministic models that exceed the 0.010 threshold.

    Equation: delta_acc = |Acc_reproduced - Acc_reported|

    Parameters
    ----------
    df_reproduced : pd.DataFrame
        The reproduced Table 1 (Accuracy) with MultiIndex rows and columns.
    config : Dict[str, Any]
        The master study configuration dictionary containing reported_results.

    Returns
    -------
    Tuple[pd.DataFrame, List[DeviationRecord]]
        A DataFrame of the same shape containing the absolute deviations, and a list
        of flagged DeviationRecords for deterministic models.
    """
    # Initialize an empty DataFrame with the exact same hierarchical structure
    df_deviations = pd.DataFrame(np.nan, index=df_reproduced.index, columns=df_reproduced.columns)
    flagged_cells: List[DeviationRecord] = []

    reported_results = config["reported_results"]

    # Iterate over every cell coordinate in the MultiIndex DataFrame
    for (ep_display, cond) in df_reproduced.index:
        for (win_display, mod_display) in df_reproduced.columns:

            # Map display names to internal keys
            ep_internal: str = DISPLAY_TO_INTERNAL_EPISODE[ep_display]
            mod_internal: str = DISPLAY_TO_INTERNAL_MODEL[mod_display]
            mod_reported: str = INTERNAL_TO_REPORTED_MODEL[mod_internal]

            # Extract the integer window length from the display string (e.g., "W = 60" -> 60)
            w_int: int = int(win_display.split("=")[1].strip())

            # Construct the key to access the reported results dictionary
            table_key: str = f"table_1_accuracy_W{w_int}"

            # Retrieve the reported value
            try:
                reported_val: float = reported_results[table_key][ep_internal][cond][mod_reported]
            except KeyError as e:
                raise KeyError(f"Missing reported accuracy for {table_key}, {ep_internal}, {cond}, {mod_reported}: {e}")

            # Retrieve the reproduced value
            reproduced_val: float = df_reproduced.loc[(ep_display, cond), (win_display, mod_display)]

            # Compute the absolute deviation
            delta_acc: float = abs(reproduced_val - reported_val)

            # Store the deviation in the matrix
            df_deviations.loc[(ep_display, cond), (win_display, mod_display)] = delta_acc

            # Flag deterministic models exceeding the threshold
            # Note: Accuracy is quantized in steps of 1/72 (~0.01389), so > 0.010 catches any single mismatch
            if mod_internal != "gpt5" and delta_acc > 0.010:
                record = DeviationRecord(
                    episode=ep_internal,
                    condition=cond,
                    model=mod_internal,
                    window=w_int,
                    metric="accuracy",
                    reproduced_value=reproduced_val,
                    reported_value=reported_val,
                    absolute_deviation=delta_acc
                )
                flagged_cells.append(record)

    return df_deviations, flagged_cells

# -------------------------------------------------------------------------------------------------------------------------------
# Task 17, Step 2: Compute per-cell Sharpe deviations and classify GPT-5
# -------------------------------------------------------------------------------------------------------------------------------
def compute_sharpe_deviations(
    df_reproduced: pd.DataFrame,
    config: Dict[str, Any]
) -> Tuple[pd.DataFrame, List[DeviationRecord], List[DeviationRecord]]:
    """
    Computes the absolute deviation for Sharpe ratios, flagging deterministic models
    exceeding 0.020, and isolating all GPT-5 deviations as expected irreproducibility.

    Equation: delta_SR = |SR_reproduced - SR_reported|

    Parameters
    ----------
    df_reproduced : pd.DataFrame
        The reproduced Table 3 (Sharpe) with MultiIndex rows and columns.
    config : Dict[str, Any]
        The master study configuration dictionary.

    Returns
    -------
    Tuple[pd.DataFrame, List[DeviationRecord], List[DeviationRecord]]
        The deviation DataFrame, the list of flagged deterministic cells, and the
        list of GPT-5 specific deviations.
    """
    df_deviations = pd.DataFrame(np.nan, index=df_reproduced.index, columns=df_reproduced.columns)
    flagged_cells: List[DeviationRecord] = []
    gpt5_deviations: List[DeviationRecord] = []

    reported_results = config["reported_results"]

    for (ep_display, cond) in df_reproduced.index:
        for (win_display, mod_display) in df_reproduced.columns:

            ep_internal: str = DISPLAY_TO_INTERNAL_EPISODE[ep_display]
            mod_internal: str = DISPLAY_TO_INTERNAL_MODEL[mod_display]
            mod_reported: str = INTERNAL_TO_REPORTED_MODEL[mod_internal]
            w_int: int = int(win_display.split("=")[1].strip())

            table_key: str = f"table_3_sharpe_W{w_int}"

            try:
                reported_val: float = reported_results[table_key][ep_internal][cond][mod_reported]
            except KeyError as e:
                raise KeyError(f"Missing reported Sharpe for {table_key}, {ep_internal}, {cond}, {mod_reported}: {e}")

            reproduced_val: float = df_reproduced.loc[(ep_display, cond), (win_display, mod_display)]

            # Compute absolute deviation (handles negative Sharpe ratios correctly)
            delta_sr: float = abs(reproduced_val - reported_val)
            df_deviations.loc[(ep_display, cond), (win_display, mod_display)] = delta_sr

            record = DeviationRecord(
                episode=ep_internal,
                condition=cond,
                model=mod_internal,
                window=w_int,
                metric="sharpe",
                reproduced_value=reproduced_val,
                reported_value=reported_val,
                absolute_deviation=delta_sr
            )

            # GPT-5 Special Handling: All deviations are expected due to temperature=1.0
            if mod_internal == "gpt5":
                record.classification = "EXPECTED_STOCHASTICITY"
                gpt5_deviations.append(record)
            # Deterministic models: Flag if deviation exceeds the 0.020 threshold
            elif delta_sr > 0.020:
                flagged_cells.append(record)

    return df_deviations, flagged_cells, gpt5_deviations

# -------------------------------------------------------------------------------------------------------------------------------
# Task 17, Step 3: Diagnose deterministic deviations and compile the report
# -------------------------------------------------------------------------------------------------------------------------------
def diagnose_and_compile_report(
    flagged_acc: List[DeviationRecord],
    flagged_sr: List[DeviationRecord],
    gpt5_devs: List[DeviationRecord],
    calibration_passed: bool,
    dataset_fingerprint: str,
    run_summary: Dict[str, float],
    clients: Dict[str, Any],
    config: Dict[str, Any]
) -> Dict[str, Any]:
    """
    Applies the diagnostic taxonomy to flagged cells and compiles the final
    structured reproducibility report.

    Taxonomy:
    - Calibration Failed -> DATA_VERSION_DISCREPANCY
    - Calibration Passed + Deterministic Deviation -> MODEL_VERSION_DRIFT
    - GPT-5 -> EXPECTED_STOCHASTICITY

    Parameters
    ----------
    flagged_acc : List[DeviationRecord]
        Flagged accuracy deviations for deterministic models.
    flagged_sr : List[DeviationRecord]
        Flagged Sharpe deviations for deterministic models.
    gpt5_devs : List[DeviationRecord]
        All GPT-5 deviations.
    calibration_passed : bool
        True if the upfront buy-and-hold calibration succeeded.
    dataset_fingerprint : str
        The SHA-256 hash of the aligned dataset.
    run_summary : Dict[str, float]
        The execution statistics (calls, fallbacks, retries).
    clients : Dict[str, LLMClientConfig]
        The configured clients (used to extract exact model IDs).
    config : Dict[str, Any]
        The master study configuration.

    Returns
    -------
    Dict[str, Any]
        The comprehensive, structured reproducibility report.
    """
    all_flagged = flagged_acc + flagged_sr

    # Apply the diagnostic taxonomy to deterministic models
    for record in all_flagged:
        if not calibration_passed:
            # If calibration failed, the data version is the root cause
            record.classification = "DATA_VERSION_DISCREPANCY"
        else:
            # If calibration passed, the data is sound; the model weights must have drifted
            record.classification = "MODEL_VERSION_DRIFT"

    # Combine all records for the final table
    all_records = all_flagged + gpt5_devs

    # Extract exact model identifiers used during the run
    exact_model_ids = {
        "gpt5": clients["gpt5"].model_id,
        "claude_sonnet_4_6": clients["claude_sonnet_4_6"].model_id,
        "gemini_3_pro": clients["gemini_3_pro"].model_id
    }

    # Retrieve the download date from the configuration
    download_date = config["data_provenance"]["price_series"].get("download_date", "UNKNOWN")

    # Construct the final report dictionary
    report: Dict[str, Any] = {
        "metadata": {
            "dataset_fingerprint_sha256": dataset_fingerprint,
            "yahoo_finance_download_date": download_date,
            "exact_model_identifiers_used": exact_model_ids,
            "calibration_passed": calibration_passed
        },
        "execution_summary": run_summary,
        "diagnostic_taxonomy_applied": {
            "DATA_VERSION_DISCREPANCY": len([r for r in all_flagged if r.classification == "DATA_VERSION_DISCREPANCY"]),
            "MODEL_VERSION_DRIFT": len([r for r in all_flagged if r.classification == "MODEL_VERSION_DRIFT"]),
            "EXPECTED_STOCHASTICITY": len(gpt5_devs)
        },
        "cell_by_cell_deviations": [asdict(r) for r in all_records]
    }

    return report

def generate_deviation_heatmaps(
    df_acc_dev: pd.DataFrame,
    df_sr_dev: pd.DataFrame
) -> Tuple[pd.io.formats.style.Styler, pd.io.formats.style.Styler]:
    """
    Applies a background gradient to the deviation DataFrames to create visual heatmaps.
    """
    # Apply a red color map where darker red indicates higher deviation
    styled_acc = df_acc_dev.style.background_gradient(cmap='Reds', axis=None)
    styled_sr = df_sr_dev.style.background_gradient(cmap='Reds', axis=None)

    return styled_acc, styled_sr

# -------------------------------------------------------------------------------------------------------------------------------
# Task 17, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def verify_reproduced_results(
    df_table_1: pd.DataFrame,
    df_table_3: pd.DataFrame,
    calibration_passed: bool,
    dataset_fingerprint: str,
    run_summary: Dict[str, float],
    clients: Dict[str, Any],
    config: Dict[str, Any]
) -> Tuple[Dict[str, Any], pd.io.formats.style.Styler, pd.io.formats.style.Styler]:
    """
    Orchestrates the verification of reproduced results against the paper's reported tables.

    This function computes absolute deviations, flags exceedances, applies the
    diagnostic taxonomy (isolating GPT-5 stochasticity), and compiles the final
    audit-ready reproducibility report and heatmaps.

    Parameters
    ----------
    df_table_1 : pd.DataFrame
        The reproduced Table 1 (Accuracy).
    df_table_3 : pd.DataFrame
        The reproduced Table 3 (Sharpe).
    calibration_passed : bool
        True if the upfront buy-and-hold calibration succeeded.
    dataset_fingerprint : str
        The SHA-256 hash of the aligned dataset.
    run_summary : Dict[str, float]
        The execution statistics.
    clients : Dict[str, LLMClientConfig]
        The configured clients.
    config : Dict[str, Any]
        The master study configuration.

    Returns
    -------
    Tuple[Dict[str, Any], pd.io.formats.style.Styler, pd.io.formats.style.Styler]
        The structured JSON report dictionary, and the two styled deviation heatmaps.
    """
    logger.info("Verifying reproduced results against reported tables...")

    # Step 1: Compute accuracy deviations and flag deterministic models
    df_acc_dev, flagged_acc = compute_accuracy_deviations(df_table_1, config)

    # Step 2: Compute Sharpe deviations, flag deterministic models, and isolate GPT-5
    df_sr_dev, flagged_sr, gpt5_devs = compute_sharpe_deviations(df_table_3, config)

    # Step 3: Diagnose deviations and compile the final report
    report: Dict[str, Any] = diagnose_and_compile_report(
        flagged_acc, flagged_sr, gpt5_devs, calibration_passed,
        dataset_fingerprint, run_summary, clients, config
    )

    # Generate visual heatmaps for the deviations
    heatmap_acc, heatmap_sr = generate_deviation_heatmaps(df_acc_dev, df_sr_dev)

    logger.info("Reproducibility verification complete. Report generated.")

    return report, heatmap_acc, heatmap_sr
````

````python
# Top Level Orchestrator

# ==============================================================================
# Top-Level Orchestrator: Execute the Complete Research Pipeline
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Helper Function: Validate File Paths
# -------------------------------------------------------------------------------------------------------------------------------
def validate_file_paths(
    # The filesystem path to the raw Yahoo Finance CSV file.
    yahoo_path: str,
    # The filesystem path to the raw Shiller Excel workbook.
    shiller_path: str
) -> None:
    """
    Validates the physical existence of the raw input data files on disk.

    Parameters
    ----------
    yahoo_path : str
        The filesystem path to the raw Yahoo Finance CSV file.
    shiller_path : str
        The filesystem path to the raw Shiller Excel workbook.

    Returns
    -------
    None

    Raises
    ------
    FileNotFoundError
        If either of the specified files does not exist.
    TypeError
        If any path argument is not a string.
    """
    # Verify that the Yahoo Finance path argument is a string.
    if not isinstance(yahoo_path, str):
        # Raise a TypeError if the Yahoo path is not a string.
        raise TypeError(f"Expected str for yahoo_path, got {type(yahoo_path).__name__}")

    # Verify that the Shiller path argument is a string.
    if not isinstance(shiller_path, str):
        # Raise a TypeError if the Shiller path is not a string.
        raise TypeError(f"Expected str for shiller_path, got {type(shiller_path).__name__}")

    # Verify the physical existence of the Yahoo Finance CSV file on disk.
    if not os.path.exists(yahoo_path):
        # Raise FileNotFoundError if the Yahoo file is missing.
        raise FileNotFoundError(f"Fatal Error: Yahoo Finance CSV file not found at '{yahoo_path}'")

    # Verify the physical existence of the Shiller Excel workbook on disk.
    if not os.path.exists(shiller_path):
        # Raise FileNotFoundError if the Shiller file is missing.
        raise FileNotFoundError(f"Fatal Error: Shiller Excel workbook not found at '{shiller_path}'")

    # Log the successful validation of the file paths.
    logger.info("Raw input file paths validated successfully on disk.")

# -------------------------------------------------------------------------------------------------------------------------------
# Top-Level Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------
def execute_complete_research_pipeline(
    # The filesystem path to the raw Yahoo Finance CSV file.
    yahoo_csv_path: str,
    # The filesystem path to the raw Shiller Excel workbook.
    shiller_excel_path: str,
    # The ISO 8601 formatted string representing the Yahoo download date.
    yahoo_download_date: str,
    # The master study configuration dictionary.
    config: Dict[str, Any],
    # The filesystem path to the persistent JSONL log file.
    log_file_path: str
) -> Dict[str, Any]:
    """
    The apex orchestrator function that executes the entire "Reflexivity as Prompt"
    research pipeline from raw data ingestion to final reproducibility verification.

    This function sequentially binds Tasks 1 through 17, enforcing strict data
    validation, data version calibration, crash-recoverable loop execution, and
    reproducibility audit.

    Parameters
    ----------
    yahoo_csv_path : str
        The filesystem path to the raw Yahoo Finance CSV file.
    shiller_excel_path : str
        The filesystem path to the raw Shiller Excel workbook.
    yahoo_download_date : str
        The ISO 8601 formatted string representing the Yahoo download date.
    config : Dict[str, Any]
        The master study configuration dictionary (STUDY_CONFIG).
    log_file_path : str
        The filesystem path to the persistent JSONL log file.

    Returns
    -------
    Dict[str, Any]
        A comprehensive dictionary containing all terminal artifacts:
        - 'table_1_accuracy': pd.io.formats.style.Styler (or absent on failure)
        - 'table_3_sharpe': pd.io.formats.style.Styler (or absent on failure)
        - 'execution_summary': Dict[str, float] (or absent on failure)
        - 'reproducibility_report': Dict[str, Any] (or absent on failure)
        - 'heatmap_accuracy': pd.io.formats.style.Styler (or absent on failure)
        - 'heatmap_sharpe': pd.io.formats.style.Styler (or absent on failure)
        - 'status': str ("SUCCESS", "CALIBRATION_FAILED", or "FATAL_ERROR")
        - 'error_message': str (only present if status is not "SUCCESS")

    Raises
    ------
    Exception
        Re-raises any non-calibration fatal errors to ensure the environment is alerted.
    """
    # Initialize the terminal artifacts dictionary.
    artifacts: Dict[str, Any] = {
        # Set the initial state to INITIALIZING.
        "status": "INITIALIZING"
    }
    # Wrap the entire pipeline execution inside a broad try-except block.
    try:
        # Step 1: Validate the physical existence of the raw files on disk.
        validate_file_paths(yahoo_csv_path, shiller_excel_path)

        # Step 2: Validate the structural and type completeness of the master configuration.
        # This will raise a KeyError, TypeError, or ValueError on any consistency failure.
        validate_study_config(config)

        # Step 3: Ingest, clean, and validate the Yahoo Finance price series.
        # This extracts Adj Close, parses dates, and returns the verified container.
        yahoo_container: Any = ingest_and_validate_yahoo_finance_series(yahoo_csv_path, yahoo_download_date)

        # Step 4: Ingest, convert, and validate the Shiller EPS series.
        # This decodes decimal dates, extracts Earnings, and returns the verified container.
        shiller_container: Any = ingest_and_validate_shiller_eps_series(shiller_excel_path)

        # Step 5: Align, integrate, and cleanse the datasets into a single canonical series.
        # This performs Period key matching and generates the SHA-256 fingerprint.
        merged_container: Any = align_integrate_and_cleanse(yahoo_container, shiller_container)

        # Step 6: Configure OpenAI, Anthropic, and Google LLM API clients.
        # This instantiates clients and verifies that all API keys exist in the environment.
        clients: Dict[str, Any] = configure_all_llm_clients(config)

        # Step 7: Execute the Task 16 orchestrator.
        # This handles the calibration gate, evaluation loop, metrics, and table assembly.
        styled_table_1, styled_table_3, summary_stats = run_end_to_end_pipeline(
            merged_container=merged_container,
            config=config,
            clients=clients,
            log_file_path=log_file_path
        )

        # Extract the SHA-256 fingerprint from the aligned container metadata.
        dataset_fingerprint: str = str(merged_container.metadata.get("dataset_fingerprint_sha256", "UNKNOWN"))

        # Step 8: Execute the Task 17 orchestrator.
        # This computes absolute deviations, applies the taxonomy, and generates heatmaps.
        # We pass the underlying DataFrames (.data) from the Stylers for numerical comparison.
        report, heatmap_acc, heatmap_sr = verify_reproduced_results(
            df_table_1=styled_table_1.data,
            df_table_3=styled_table_3.data,
            calibration_passed=True,
            dataset_fingerprint=dataset_fingerprint,
            run_summary=summary_stats,
            clients=clients,
            config=config
        )

        # Populate the artifact payload dictionary with the successful results.
        artifacts["table_1_accuracy"] = styled_table_1

        # Add the styled Sharpe ratio table.
        artifacts["table_3_sharpe"] = styled_table_3

        # Add the run execution summary statistics.
        artifacts["execution_summary"] = summary_stats

        # Add the structured audit report.
        artifacts["reproducibility_report"] = report

        # Add the styled accuracy deviation heatmap.
        artifacts["heatmap_accuracy"] = heatmap_acc

        # Add the styled Sharpe deviation heatmap.
        artifacts["heatmap_sharpe"] = heatmap_sr

        # Set the final execution status to SUCCESS.
        artifacts["status"] = "SUCCESS"

        # Log the successful execution of the complete pipeline.
        logger.info("=== COMPLETE RESEARCH PIPELINE EXECUTED SUCCESSFULLY ===")
    # Catch CalibrationError specifically to handle data version discrepancy.
    except Exception as e:
        # Check if the exception represents a data version calibration failure.
        if type(e).__name__ == "CalibrationError":
            # Log the calibration failure at the ERROR level.
            logger.error(f"PIPELINE HALTED: {str(e)}")

            # Set the status metadata to CALIBRATION_FAILED.
            artifacts["status"] = "CALIBRATION_FAILED"

            # Save the explicit error message.
            artifacts["error_message"] = str(e)
        # Handle all other general, system, or programming exceptions.
        else:
            # Log the fatal error at the ERROR level.
            logger.error(f"PIPELINE HALTED DUE TO FATAL ERROR: {str(e)}")

            # Set the status metadata to FATAL_ERROR.
            artifacts["status"] = "FATAL_ERROR"

            # Save the explicit error message.
            artifacts["error_message"] = str(e)

            # Re-raise the exception to alert the calling environment.
            raise e

    # Return the populated, structured artifacts dictionary.
    return artifacts
````
